```python
import os

import json

import pandas as pd

import numpy as np

import networkx as nx

from rdflib import Graph, URIRef, Literal, Namespace

from rdflib.namespace import RDF, RDFS, XSD

from typing import Dict, List, Tuple, Union, Optional

import random

from tqdm import tqdm

import hashlib

import torch

from torch.utils.data import Dataset

from torchvision import transforms

from PIL import Image

import spacy

from transformers import AutoTokenizer, AutoModel

import shutil

from pathlib import Path

import os

import torch

import torch.nn as nn

import torch.nn.functional as F

import torchvision.models as models

import torchvision.transforms as transforms

from torch.utils.data import Dataset, DataLoader

from PIL import Image

import numpy as np

import pandas as pd

from transformers import AutoModel, AutoTokenizer

from tqdm import tqdm

import math

import random

from typing import Dict, List, Tuple, Optional, Union

import os

import json

import torch

import torch.nn as nn

import torch.nn.functional as F

import numpy as np

import pandas as pd

import networkx as nx

from torch.nn import Parameter

from typing import Dict, List, Tuple, Optional, Union

import math

import random

from rdflib import Graph, URIRef, Literal

from torch\_geometric.nn import RGCNConv, FastRGCNConv

from torch\_geometric.data import Data

import pyro

import pyro.distributions as dist

from torch.distributions import Normal, Independent, Beta

import os

import json

import torch

import torch.nn as nn

import torch.nn.functional as F

import numpy as np

import pandas as pd

from typing import Dict, List, Tuple, Optional, Union

import math

import random

from torch.nn import Parameter

from rdflib import Graph, URIRef, Literal

from torch.autograd import Function
import os

import json

import torch

import torch.nn as nn

import torch.nn.functional as F

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

from PIL import Image, ImageDraw, ImageFont

import cv2

from typing import Dict, List, Tuple, Optional, Union, Any

import math

import random

from sklearn.tree import DecisionTreeClassifier, export\_text

from sklearn.linear\_model import LogisticRegression

from sklearn.metrics import accuracy\_score

from rdflib import Graph, URIRef, Literal

import networkx as nx

from torch.autograd import grad

import copy

import os

import json

import torch

import torch.nn as nn

import torch.nn.functional as F

import numpy as np

import pandas as pd

from typing import Dict, List, Tuple, Optional, Union, Any

import math

import random

from transformers import (

&nbsp;   AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM,

&nbsp;   GPT2LMHeadModel, GPT2Tokenizer, GPT2Config,

&nbsp;   T5ForConditionalGeneration, T5Tokenizer,

&nbsp;   Trainer, TrainingArguments, DataCollatorForSeq2Seq

)

from torch.utils.data import Dataset, DataLoader

from rdflib import Graph, URIRef, Literal

import re

import spacy

from sklearn.metrics.pairwise import cosine\_similarity
import os

import json

import torch

import torch.nn as nn

import torch.nn.functional as F

import numpy as np

import pandas as pd

from typing import Dict, List, Tuple, Optional, Union, Any

import math

import random

from datetime import datetime

import threading

import queue

import time

from torch.utils.data import DataLoader, Dataset

from torch.optim import SGD, Adam

from torch.utils.tensorboard import SummaryWriter

import matplotlib.pyplot as plt

from collections import defaultdict

import copy

import requests

import uuid

from rdflib import Graph, URIRef, Literal

import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

import numpy as np

import networkx as nx

from matplotlib\_venn import venn3

from wordcloud import WordCloud

from matplotlib.patches import Patch

import matplotlib.dates as mdates

from datetime import datetime, timedelta

from IPython.display import HTML, IFrame

import os


























\# Load medical NLP model for text processing

try:

&nbsp;   nlp = spacy.load("en\_core\_sci\_sm")  # Medical-specific NLP model

except:

&nbsp;   nlp = spacy.load("en\_core\_web\_sm")  # Fallback to general model



\# Define namespaces for knowledge graph

MED = Namespace("http://example.org/medical/")

BREAST = Namespace("http://example.org/breast\_cancer/")



def preprocess\_cbis\_ddsm(cbis\_path, output\_path):

&nbsp;   """

&nbsp;   Preprocess CBIS-DDSM dataset to match our pipeline requirements

&nbsp;   """

&nbsp;   # Create output directories

&nbsp;   os.makedirs(os.path.join(output\_path, "mammograms"), exist\_ok=True)

&nbsp;   os.makedirs(os.path.join(output\_path, "reports"), exist\_ok=True)

&nbsp;

&nbsp;   # Load metadata files

&nbsp;   csv\_path = os.path.join(cbis\_path, "csv")

&nbsp;   mass\_train = pd.read\_csv(os.path.join(csv\_path, "mass\_case\_description\_train\_set.csv"))

&nbsp;   calc\_train = pd.read\_csv(os.path.join(csv\_path, "calc\_case\_description\_train\_set.csv"))

&nbsp;   mass\_test = pd.read\_csv(os.path.join(csv\_path, "mass\_case\_description\_test\_set.csv"))

&nbsp;   calc\_test = pd.read\_csv(os.path.join(csv\_path, "calc\_case\_description\_test\_set.csv"))

&nbsp;

&nbsp;   # Combine datasets

&nbsp;   metadata = pd.concat(\[mass\_train, calc\_train, mass\_test, calc\_test])

&nbsp;

&nbsp;   # Process each case

&nbsp;   processed\_count = 0

&nbsp;   for \_, row in metadata.iterrows():

&nbsp;       patient\_id = f"P{row\['patient\_id']}"

&nbsp;       study\_id = f"S{row\['study\_id']}"

&nbsp;

&nbsp;       # Process mammogram image

&nbsp;       img\_path = os.path.join(cbis\_path, "jpeg", row\['image file path'])

&nbsp;       if os.path.exists(img\_path):

&nbsp;           # Determine view from filename

&nbsp;           parent\_dir = os.path.basename(os.path.dirname(img\_path))

&nbsp;           view = parent\_dir.split('\_')\[-1]  # Extract CC or MLO

&nbsp;

&nbsp;           # Create standardized filename: PatientID\_StudyID\_View.jpg

&nbsp;           new\_img\_name = f"{patient\_id}\_{study\_id}\_{view}.jpg"

&nbsp;           new\_img\_path = os.path.join(output\_path, "mammograms", new\_img\_name)

&nbsp;

&nbsp;           # Copy image to output directory

&nbsp;           shutil.copy2(img\_path, new\_img\_path)

&nbsp;

&nbsp;           # Create a report file with metadata

&nbsp;           report\_path = os.path.join(output\_path, "reports", f"{patient\_id}\_{study\_id}.txt")

&nbsp;           with open(report\_path, 'w') as f:

&nbsp;               f.write(f"Patient ID: {patient\_id}\\n")

&nbsp;               f.write(f"Study ID: {study\_id}\\n")

&nbsp;               f.write(f"View: {view}\\n")

&nbsp;               f.write(f"Pathology: {row\['pathology']}\\n")

&nbsp;               f.write(f"Assessment: {row\['assessment']}\\n")

&nbsp;               f.write(f"Subtlety: {row\['subtlety']}\\n")

&nbsp;

&nbsp;           processed\_count += 1

&nbsp;           if processed\_count % 100 == 0:

&nbsp;               print(f"Processed {processed\_count} CBIS-DDSM cases...")

&nbsp;

&nbsp;   print(f"CBIS-DDSM preprocessing completed. Total cases processed: {processed\_count}")

&nbsp;   return processed\_count



def preprocess\_breakhis(breakhis\_path, output\_path):

&nbsp;   """

&nbsp;   Preprocess BreakHis dataset to match our pipeline requirements

&nbsp;   """

&nbsp;   # Create output directory

&nbsp;   histopathology\_dir = os.path.join(output\_path, "histopathology")

&nbsp;   os.makedirs(histopathology\_dir, exist\_ok=True)

&nbsp;

&nbsp;   # Create metadata file

&nbsp;   metadata\_path = os.path.join(output\_path, "breakhis\_metadata.csv")

&nbsp;   metadata\_list = \[]

&nbsp;

&nbsp;   # Traverse through the BreakHis dataset

&nbsp;   processed\_count = 0

&nbsp;   for fold\_dir in os.listdir(breakhis\_path):

&nbsp;       fold\_path = os.path.join(breakhis\_path, fold\_dir)

&nbsp;       if not os.path.isdir(fold\_path) or not fold\_dir.startswith("B\_Fold"):

&nbsp;           continue

&nbsp;

&nbsp;       for class\_name in \["benign", "malignant"]:

&nbsp;           class\_path = os.path.join(fold\_path, class\_name)

&nbsp;           if not os.path.exists(class\_path):

&nbsp;               continue

&nbsp;

&nbsp;           # Traverse through the SOB subdirectories

&nbsp;           for sob\_dir in os.listdir(class\_path):

&nbsp;               sob\_path = os.path.join(class\_path, sob\_dir)

&nbsp;               if not os.path.isdir(sob\_path):

&nbsp;                   continue

&nbsp;

&nbsp;               # Process each image in the SOB directory

&nbsp;               for img\_file in os.listdir(sob\_path):

&nbsp;                   if not img\_file.lower().endswith(('.png', '.jpg', '.jpeg')):

&nbsp;                       continue

&nbsp;

&nbsp;                   img\_path = os.path.join(sob\_path, img\_file)

&nbsp;

&nbsp;                   # Extract information from filename

&nbsp;                   parts = img\_file.split('\_')

&nbsp;                   if len(parts) < 4:

&nbsp;                       continue

&nbsp;

&nbsp;                   # The third part is the patient ID (e.g., "A-14-22549G")

&nbsp;                   patient\_id = parts\[2]

&nbsp;

&nbsp;                   # The fourth part contains magnification and image number (e.g., "40-001.png")

&nbsp;                   mag\_part = parts\[3].split('-')

&nbsp;                   if len(mag\_part) < 2:

&nbsp;                       continue

&nbsp;                   magnification = mag\_part\[0]  # e.g., "40"

&nbsp;

&nbsp;                   # Create new filename

&nbsp;                   new\_img\_name = f"{patient\_id}\_{class\_name}\_{magnification}.jpg"

&nbsp;                   new\_img\_path = os.path.join(histopathology\_dir, new\_img\_name)

&nbsp;

&nbsp;                   # Convert and save as JPG

&nbsp;                   try:

&nbsp;                       img = Image.open(img\_path)

&nbsp;                       # Convert to RGB if necessary (for PNG with transparency)

&nbsp;                       if img.mode != 'RGB':

&nbsp;                           img = img.convert('RGB')

&nbsp;                       img.save(new\_img\_path, 'JPEG')

&nbsp;

&nbsp;                       # Add to metadata

&nbsp;                       metadata\_list.append({

&nbsp;                           'patient\_id': patient\_id,

&nbsp;                           'class': class\_name,

&nbsp;                           'magnification': magnification,

&nbsp;                           'original\_file': img\_file,

&nbsp;                           'new\_file': new\_img\_name,

&nbsp;                           'fold': fold\_dir

&nbsp;                       })

&nbsp;

&nbsp;                       processed\_count += 1

&nbsp;                       if processed\_count % 100 == 0:

&nbsp;                           print(f"Processed {processed\_count} BreakHis images...")

&nbsp;                   except Exception as e:

&nbsp;                       print(f"Error processing {img\_path}: {e}")

&nbsp;

&nbsp;   # Save metadata to CSV

&nbsp;   metadata\_df = pd.DataFrame(metadata\_list)

&nbsp;   metadata\_df.to\_csv(metadata\_path, index=False)

&nbsp;

&nbsp;   print(f"BreakHis preprocessing completed. Total images processed: {processed\_count}")

&nbsp;   return processed\_count



def preprocess\_datasets(cbis\_path, breakhis\_path, output\_path):

&nbsp;   """

&nbsp;   Preprocess both CBIS-DDSM and BreakHis datasets

&nbsp;   """

&nbsp;   # Create main output directory

&nbsp;   os.makedirs(output\_path, exist\_ok=True)

&nbsp;

&nbsp;   print("Starting dataset preprocessing...")

&nbsp;   print("="\*50)

&nbsp;

&nbsp;   # Process CBIS-DDSM

&nbsp;   print("Processing CBIS-DDSM dataset...")

&nbsp;   cbis\_count = preprocess\_cbis\_ddsm(cbis\_path, output\_path)

&nbsp;

&nbsp;   print("\\nProcessing BreakHis dataset...")

&nbsp;   breakhis\_count = preprocess\_breakhis(breakhis\_path, output\_path)

&nbsp;

&nbsp;   print("\\n" + "="\*50)

&nbsp;   print("Preprocessing completed successfully!")

&nbsp;   print(f"CBIS-DDSM cases processed: {cbis\_count}")

&nbsp;   print(f"BreakHis images processed: {breakhis\_count}")

&nbsp;   print(f"All data saved to: {output\_path}")

&nbsp;

&nbsp;   # Create a summary file

&nbsp;   summary\_path = os.path.join(output\_path, "preprocessing\_summary.txt")

&nbsp;   with open(summary\_path, 'w') as f:

&nbsp;       f.write("Dataset Preprocessing Summary\\n")

&nbsp;       f.write("="\*50 + "\\n")

&nbsp;       f.write(f"CBIS-DDSM cases processed: {cbis\_count}\\n")

&nbsp;       f.write(f"BreakHis images processed: {breakhis\_count}\\n")

&nbsp;       f.write(f"Output directory: {output\_path}\\n")

&nbsp;       f.write("\\nDirectory structure:\\n")

&nbsp;       f.write("- mammograms/: Processed mammogram images from CBIS-DDSM\\n")

&nbsp;       f.write("- reports/: Metadata reports for CBIS-DDSM cases\\n")

&nbsp;       f.write("- histopathology/: Processed histopathology images from BreakHis\\n")

&nbsp;       f.write("- breakhis\_metadata.csv: Metadata for BreakHis images\\n")

&nbsp;

&nbsp;   return cbis\_count, breakhis\_count



class MultiModalHarmonizer:

&nbsp;   """

&nbsp;   Creates a unified schema for multi-modal breast cancer data

&nbsp;   """

&nbsp;   def \_\_init\_\_(self, config\_path: str = None):

&nbsp;       self.schema = {}

&nbsp;       self.config = self.\_load\_config(config\_path) if config\_path else {}

&nbsp;       self.data\_sources = \[]

&nbsp;       self.unified\_data = \[]

&nbsp;

&nbsp;   def \_load\_config(self, config\_path: str) -> Dict:

&nbsp;       """Load configuration for data harmonization"""

&nbsp;       with open(config\_path, 'r') as f:

&nbsp;           return json.load(f)

&nbsp;

&nbsp;   def register\_data\_source(self, source\_type: str, source\_path: str, metadata: Dict = None):

&nbsp;       """Register a new data source (mammogram, histopathology, clinical text)"""

&nbsp;       source\_info = {

&nbsp;           "type": source\_type,

&nbsp;           "path": source\_path,

&nbsp;           "metadata": metadata or {}

&nbsp;       }

&nbsp;       self.data\_sources.append(source\_info)

&nbsp;       print(f"Registered {source\_type} data source: {source\_path}")

&nbsp;

&nbsp;   def create\_unified\_schema(self) -> Dict:

&nbsp;       """Create a unified schema for all data sources"""

&nbsp;       # Define the unified schema structure

&nbsp;       self.schema = {

&nbsp;           "patient\_id": {"type": "string", "description": "Unique patient identifier"},

&nbsp;           "study\_id": {"type": "string", "description": "Unique study identifier"},

&nbsp;           "modality": {"type": "string", "description": "Data modality (mammogram, histopathology, text)"},

&nbsp;           "data\_path": {"type": "string", "description": "Path to the data file"},

&nbsp;           "features": {"type": "dict", "description": "Extracted features from the data"},

&nbsp;           "metadata": {"type": "dict", "description": "Additional metadata"},

&nbsp;           "labels": {"type": "dict", "description": "Available labels (if any)"},

&nbsp;           "timestamp": {"type": "datetime", "description": "Data acquisition timestamp"}

&nbsp;       }

&nbsp;

&nbsp;       # Add modality-specific fields

&nbsp;       self.schema\["mammogram\_specific"] = {

&nbsp;           "view": {"type": "string", "description": "Mammogram view (CC, MLO, etc.)"},

&nbsp;           "breast\_density": {"type": "string", "description": "Breast density category"}

&nbsp;       }

&nbsp;

&nbsp;       self.schema\["histopathology\_specific"] = {

&nbsp;           "stain\_type": {"type": "string", "description": "Stain type (H\&E, IHC, etc.)"},

&nbsp;           "magnification": {"type": "string", "description": "Microscope magnification"}

&nbsp;       }

&nbsp;

&nbsp;       self.schema\["text\_specific"] = {

&nbsp;           "report\_type": {"type": "string", "description": "Type of clinical report"},

&nbsp;           "findings": {"type": "string", "description": "Key findings from the report"}

&nbsp;       }

&nbsp;

&nbsp;       return self.schema

&nbsp;

&nbsp;   def harmonize\_data(self) -> List\[Dict]:

&nbsp;       """Harmonize all registered data sources into the unified schema"""

&nbsp;       self.unified\_data = \[]

&nbsp;

&nbsp;       for source in tqdm(self.data\_sources, desc="Harmonizing data sources"):

&nbsp;           source\_type = source\["type"]

&nbsp;           source\_path = source\["path"]

&nbsp;           metadata = source\["metadata"]

&nbsp;

&nbsp;           if source\_type == "mammogram":

&nbsp;               harmonized = self.\_harmonize\_mammograms(source\_path, metadata)

&nbsp;           elif source\_type == "histopathology":

&nbsp;               harmonized = self.\_harmonize\_histopathology(source\_path, metadata)

&nbsp;           elif source\_type == "clinical\_text":

&nbsp;               harmonized = self.\_harmonize\_clinical\_text(source\_path, metadata)

&nbsp;           else:

&nbsp;               print(f"Unknown source type: {source\_type}")

&nbsp;               continue

&nbsp;

&nbsp;           self.unified\_data.extend(harmonized)

&nbsp;

&nbsp;       print(f"Harmonized {len(self.unified\_data)} data entries")

&nbsp;       return self.unified\_data

&nbsp;

&nbsp;   def \_harmonize\_mammograms(self, path: str, metadata: Dict) -> List\[Dict]:

&nbsp;       """Harmonize mammogram data"""

&nbsp;       harmonized\_data = \[]

&nbsp;

&nbsp;       # Process mammogram files

&nbsp;       reports\_dir = os.path.join(os.path.dirname(path), "reports")

&nbsp;

&nbsp;       for root, \_, files in os.walk(path):

&nbsp;           for file in files:

&nbsp;               if file.lower().endswith(('.png', '.jpg', '.jpeg')):

&nbsp;                   file\_path = os.path.join(root, file)

&nbsp;

&nbsp;                   # Extract patient and study ID from filename

&nbsp;                   filename\_parts = file.split('\_')

&nbsp;                   if len(filename\_parts) < 3:

&nbsp;                       continue

&nbsp;

&nbsp;                   patient\_id = filename\_parts\[0]

&nbsp;                   study\_id = filename\_parts\[1]

&nbsp;                   view = filename\_parts\[2].split('.')\[0]

&nbsp;

&nbsp;                   # Try to get additional metadata from report file

&nbsp;                   report\_path = os.path.join(reports\_dir, f"{patient\_id}\_{study\_id}.txt")

&nbsp;                   pathology = ""

&nbsp;                   assessment = ""

&nbsp;                   subtlety = ""

&nbsp;

&nbsp;                   if os.path.exists(report\_path):

&nbsp;                       with open(report\_path, 'r') as f:

&nbsp;                           for line in f:

&nbsp;                               if line.startswith("Pathology:"):

&nbsp;                                   pathology = line.split(":", 1)\[1].strip()

&nbsp;                               elif line.startswith("Assessment:"):

&nbsp;                                   assessment = line.split(":", 1)\[1].strip()

&nbsp;                               elif line.startswith("Subtlety:"):

&nbsp;                                   subtlety = line.split(":", 1)\[1].strip()

&nbsp;

&nbsp;                   # Create harmonized entry

&nbsp;                   entry = {

&nbsp;                       "patient\_id": patient\_id,

&nbsp;                       "study\_id": study\_id,

&nbsp;                       "modality": "mammogram",

&nbsp;                       "data\_path": file\_path,

&nbsp;                       "features": {},

&nbsp;                       "metadata": {

&nbsp;                           \*\*metadata,

&nbsp;                           "view": view,

&nbsp;                           "pathology": pathology,

&nbsp;                           "assessment": assessment,

&nbsp;                           "subtlety": subtlety,

&nbsp;                           "breast\_density": self.\_extract\_density(pathology)

&nbsp;                       },

&nbsp;                       "labels": {

&nbsp;                           "malignancy": 1 if pathology and "MALIGNANT" in pathology.upper() else 0

&nbsp;                       },

&nbsp;                       "timestamp": self.\_extract\_timestamp(file, metadata)

&nbsp;                   }

&nbsp;

&nbsp;                   harmonized\_data.append(entry)

&nbsp;

&nbsp;       return harmonized\_data

&nbsp;

&nbsp;   def \_harmonize\_histopathology(self, path: str, metadata: Dict) -> List\[Dict]:

&nbsp;       """Harmonize histopathology data"""

&nbsp;       harmonized\_data = \[]

&nbsp;

&nbsp;       # Load metadata file

&nbsp;       metadata\_path = os.path.join(os.path.dirname(path), "breakhis\_metadata.csv")

&nbsp;       if not os.path.exists(metadata\_path):

&nbsp;           print(f"Metadata file not found: {metadata\_path}")

&nbsp;           return harmonized\_data

&nbsp;

&nbsp;       metadata\_df = pd.read\_csv(metadata\_path)

&nbsp;

&nbsp;       for root, \_, files in os.walk(path):

&nbsp;           for file in files:

&nbsp;               if file.lower().endswith(('.png', '.jpg', '.jpeg')):

&nbsp;                   file\_path = os.path.join(root, file)

&nbsp;

&nbsp;                   # Find corresponding metadata

&nbsp;                   file\_metadata = metadata\_df\[metadata\_df\['new\_file'] == file]

&nbsp;                   if file\_metadata.empty:

&nbsp;                       continue

&nbsp;

&nbsp;                   file\_metadata = file\_metadata.iloc\[0]

&nbsp;

&nbsp;                   # Extract patient and study ID

&nbsp;                   filename\_parts = file.split('\_')

&nbsp;                   if len(filename\_parts) < 3:

&nbsp;                       continue

&nbsp;

&nbsp;                   patient\_id = filename\_parts\[0]

&nbsp;                   study\_id = f"H\_{filename\_parts\[1]}\_{filename\_parts\[2].split('.')\[0]}"  # Create unique study ID

&nbsp;

&nbsp;                   # Create harmonized entry

&nbsp;                   entry = {

&nbsp;                       "patient\_id": patient\_id,

&nbsp;                       "study\_id": study\_id,

&nbsp;                       "modality": "histopathology",

&nbsp;                       "data\_path": file\_path,

&nbsp;                       "features": {},

&nbsp;                       "metadata": {

&nbsp;                           \*\*metadata,

&nbsp;                           "stain\_type": "H\&E",  # BreakHis primarily uses H\&E stains

&nbsp;                           "magnification": file\_metadata\['magnification'],

&nbsp;                           "class": file\_metadata\['class'],

&nbsp;                           "fold": file\_metadata\['fold']

&nbsp;                       },

&nbsp;                       "labels": {

&nbsp;                           "malignancy": 1 if file\_metadata\['class'] == "malignant" else 0

&nbsp;                       },

&nbsp;                       "timestamp": self.\_extract\_timestamp(file, metadata)

&nbsp;                   }

&nbsp;

&nbsp;                   harmonized\_data.append(entry)

&nbsp;

&nbsp;       return harmonized\_data

&nbsp;

&nbsp;   def \_harmonize\_clinical\_text(self, path: str, metadata: Dict) -> List\[Dict]:

&nbsp;       """Harmonize clinical text data"""

&nbsp;       harmonized\_data = \[]

&nbsp;

&nbsp;       # Process report files

&nbsp;       reports\_dir = os.path.join(os.path.dirname(path), "reports")

&nbsp;

&nbsp;       for root, \_, files in os.walk(reports\_dir):

&nbsp;           for file in files:

&nbsp;               if file.lower().endswith('.txt'):

&nbsp;                   file\_path = os.path.join(root, file)

&nbsp;

&nbsp;                   # Extract patient and study ID from filename

&nbsp;                   filename\_parts = file.split('\_')

&nbsp;                   if len(filename\_parts) < 2:

&nbsp;                       continue

&nbsp;

&nbsp;                   patient\_id = filename\_parts\[0]

&nbsp;                   study\_id = filename\_parts\[1].split('.')\[0]

&nbsp;

&nbsp;                   # Extract text content

&nbsp;                   content = self.\_extract\_text\_content(file\_path)

&nbsp;

&nbsp;                   # Create harmonized entry

&nbsp;                   entry = {

&nbsp;                       "patient\_id": patient\_id,

&nbsp;                       "study\_id": study\_id,

&nbsp;                       "modality": "clinical\_text",

&nbsp;                       "data\_path": file\_path,

&nbsp;                       "features": {},

&nbsp;                       "metadata": {

&nbsp;                           \*\*metadata,

&nbsp;                           "report\_type": "pathology\_report"

&nbsp;                       },

&nbsp;                       "labels": self.\_extract\_labels\_from\_text(content),

&nbsp;                       "timestamp": self.\_extract\_timestamp(file, metadata),

&nbsp;                       "content": content

&nbsp;                   }

&nbsp;

&nbsp;                   harmonized\_data.append(entry)

&nbsp;

&nbsp;       return harmonized\_data

&nbsp;

&nbsp;   # Helper methods for extracting information from files

&nbsp;   def \_extract\_patient\_id(self, filename: str, metadata: Dict) -> str:

&nbsp;       """Extract patient ID from filename or metadata"""

&nbsp;       if "patient\_id" in metadata:

&nbsp;           return metadata\["patient\_id"]

&nbsp;

&nbsp;       # Simple extraction from filename

&nbsp;       parts = filename.split('\_')

&nbsp;       if len(parts) > 0:

&nbsp;           return parts\[0]

&nbsp;       return "unknown"

&nbsp;

&nbsp;   def \_extract\_study\_id(self, filename: str, metadata: Dict) -> str:

&nbsp;       """Extract study ID from filename or metadata"""

&nbsp;       if "study\_id" in metadata:

&nbsp;           return metadata\["study\_id"]

&nbsp;

&nbsp;       # Simple extraction from filename

&nbsp;       parts = filename.split('\_')

&nbsp;       if len(parts) > 1:

&nbsp;           return parts\[1]

&nbsp;       return "unknown"

&nbsp;

&nbsp;   def \_extract\_view(self, filename: str) -> str:

&nbsp;       """Extract mammogram view from filename"""

&nbsp;       if "CC" in filename.upper():

&nbsp;           return "CC"

&nbsp;       elif "MLO" in filename.upper():

&nbsp;           return "MLO"

&nbsp;       return "unknown"

&nbsp;

&nbsp;   def \_extract\_density(self, pathology: str) -> str:

&nbsp;       """Extract breast density from pathology text"""

&nbsp;       # Simple keyword matching for density

&nbsp;       if not pathology:

&nbsp;           return "unknown"

&nbsp;

&nbsp;       pathology\_upper = pathology.upper()

&nbsp;       if "DENSITY D" in pathology\_upper or "DENSE" in pathology\_upper:

&nbsp;           return "D"

&nbsp;       elif "DENSITY C" in pathology\_upper:

&nbsp;           return "C"

&nbsp;       elif "DENSITY B" in pathology\_upper:

&nbsp;           return "B"

&nbsp;       elif "DENSITY A" in pathology\_upper:

&nbsp;           return "A"

&nbsp;       return "unknown"

&nbsp;

&nbsp;   def \_extract\_stain\_type(self, filename: str) -> str:

&nbsp;       """Extract stain type from filename"""

&nbsp;       if "H\&E" in filename.upper() or "HE" in filename.upper():

&nbsp;           return "H\&E"

&nbsp;       elif "IHC" in filename.upper():

&nbsp;           return "IHC"

&nbsp;       return "unknown"

&nbsp;

&nbsp;   def \_extract\_magnification(self, filename: str) -> str:

&nbsp;       """Extract magnification from filename"""

&nbsp;       # In a real implementation, this would come from metadata

&nbsp;       magnifications = \["10x", "20x", "40x"]

&nbsp;       return random.choice(magnifications)  # Simulated

&nbsp;

&nbsp;   def \_extract\_report\_type(self, filename: str) -> str:

&nbsp;       """Extract report type from filename"""

&nbsp;       if "pathology" in filename.lower():

&nbsp;           return "pathology\_report"

&nbsp;       elif "radiology" in filename.lower():

&nbsp;           return "radiology\_report"

&nbsp;       return "clinical\_note"

&nbsp;

&nbsp;   def \_extract\_labels(self, filename: str, metadata: Dict) -> Dict:

&nbsp;       """Extract labels from filename or metadata"""

&nbsp;       labels = {}

&nbsp;

&nbsp;       # Check for malignancy status

&nbsp;       if "malignant" in filename.lower() or "positive" in filename.lower():

&nbsp;           labels\["malignancy"] = 1

&nbsp;       elif "benign" in filename.lower() or "negative" in filename.lower():

&nbsp;           labels\["malignancy"] = 0

&nbsp;

&nbsp;       # Add any labels from metadata

&nbsp;       if "labels" in metadata:

&nbsp;           labels.update(metadata\["labels"])

&nbsp;

&nbsp;       return labels

&nbsp;

&nbsp;   def \_extract\_labels\_from\_text(self, content: str) -> Dict:

&nbsp;       """Extract labels from clinical text content"""

&nbsp;       labels = {}

&nbsp;

&nbsp;       if not content:

&nbsp;           return labels

&nbsp;

&nbsp;       content\_upper = content.upper()

&nbsp;

&nbsp;       # Check for malignancy status

&nbsp;       if "MALIGNANT" in content\_upper or "CANCER" in content\_upper:

&nbsp;           labels\["malignancy"] = 1

&nbsp;       elif "BENIGN" in content\_upper and "NO MALIGNANCY" in content\_upper:

&nbsp;           labels\["malignancy"] = 0

&nbsp;

&nbsp;       # Check for other findings

&nbsp;       if "MASS" in content\_upper:

&nbsp;           labels\["mass"] = 1

&nbsp;       if "CALCIFICATION" in content\_upper:

&nbsp;           labels\["calcification"] = 1

&nbsp;       if "SPICULATION" in content\_upper:

&nbsp;           labels\["spiculation"] = 1

&nbsp;

&nbsp;       return labels

&nbsp;

&nbsp;   def \_extract\_timestamp(self, filename: str, metadata: Dict) -> str:

&nbsp;       """Extract timestamp from filename or metadata"""

&nbsp;       if "timestamp" in metadata:

&nbsp;           return metadata\["timestamp"]

&nbsp;

&nbsp;       # Simulated timestamp

&nbsp;       return "2023-01-01T00:00:00"

&nbsp;

&nbsp;   def \_extract\_text\_content(self, file\_path: str) -> str:

&nbsp;       """Extract text content from a file"""

&nbsp;       try:

&nbsp;           with open(file\_path, 'r', encoding='utf-8') as f:

&nbsp;               return f.read()

&nbsp;       except Exception as e:

&nbsp;           print(f"Error reading {file\_path}: {e}")

&nbsp;           return ""

&nbsp;

&nbsp;   def save\_harmonized\_data(self, output\_path: str):

&nbsp;       """Save harmonized data to a file"""

&nbsp;       with open(output\_path, 'w') as f:

&nbsp;           json.dump(self.unified\_data, f, indent=2)

&nbsp;       print(f"Saved harmonized data to {output\_path}")





class KnowledgeGraphAugmentor:

&nbsp;   """

&nbsp;   Extracts medical relationships and builds a knowledge graph

&nbsp;   """

&nbsp;   def \_\_init\_\_(self, unified\_data: List\[Dict]):

&nbsp;       self.unified\_data = unified\_data

&nbsp;       self.kg = Graph()

&nbsp;       self.entity\_map = {}

&nbsp;       self.relationships = \[]

&nbsp;

&nbsp;       # Define namespaces

&nbsp;       self.kg.bind("med", MED)

&nbsp;       self.kg.bind("breast", BREAST)

&nbsp;

&nbsp;   def build\_knowledge\_graph(self) -> Graph:

&nbsp;       """Build a knowledge graph from the unified data"""

&nbsp;       # First, extract entities and relationships from the data

&nbsp;       self.\_extract\_entities\_and\_relationships()

&nbsp;

&nbsp;       # Then, build the RDF graph

&nbsp;       self.\_build\_rdf\_graph()

&nbsp;

&nbsp;       print(f"Built knowledge graph with {len(self.kg)} triples")

&nbsp;       return self.kg

&nbsp;

&nbsp;   def \_extract\_entities\_and\_relationships(self):

&nbsp;       """Extract medical entities and relationships from the data"""

&nbsp;       for entry in tqdm(self.unified\_data, desc="Extracting entities and relationships"):

&nbsp;           patient\_id = entry\["patient\_id"]

&nbsp;           study\_id = entry\["study\_id"]

&nbsp;           modality = entry\["modality"]

&nbsp;

&nbsp;           # Create patient entity

&nbsp;           patient\_uri = self.\_get\_entity\_uri("patient", patient\_id)

&nbsp;

&nbsp;           # Create study entity

&nbsp;           study\_uri = self.\_get\_entity\_uri("study", study\_id)

&nbsp;

&nbsp;           # Add basic relationships

&nbsp;           self.\_add\_relationship(patient\_uri, MED\["hasStudy"], study\_uri)

&nbsp;           self.\_add\_relationship(study\_uri, MED\["hasModality"], Literal(modality))

&nbsp;

&nbsp;           # Extract modality-specific entities and relationships

&nbsp;           if modality == "mammogram":

&nbsp;               self.\_extract\_mammogram\_entities(entry, patient\_uri, study\_uri)

&nbsp;           elif modality == "histopathology":

&nbsp;               self.\_extract\_histopathology\_entities(entry, patient\_uri, study\_uri)

&nbsp;           elif modality == "clinical\_text":

&nbsp;               self.\_extract\_text\_entities(entry, patient\_uri, study\_uri)

&nbsp;

&nbsp;           # Extract labels as relationships

&nbsp;           for label\_type, label\_value in entry.get("labels", {}).items():

&nbsp;               if isinstance(label\_value, (int, float, bool)):

&nbsp;                   self.\_add\_relationship(

&nbsp;                       study\_uri,

&nbsp;                       MED\[f"has{label\_type.capitalize()}"],

&nbsp;                       Literal(label\_value, datatype=XSD.boolean if isinstance(label\_value, bool) else XSD.integer)

&nbsp;                   )

&nbsp;

&nbsp;   def \_extract\_mammogram\_entities(self, entry: Dict, patient\_uri: URIRef, study\_uri: URIRef):

&nbsp;       """Extract entities and relationships specific to mammograms"""

&nbsp;       metadata = entry.get("metadata", {})

&nbsp;

&nbsp;       # Extract breast density

&nbsp;       density = metadata.get("breast\_density")

&nbsp;       if density:

&nbsp;           density\_uri = self.\_get\_entity\_uri("density", density)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasBreastDensity"], density\_uri)

&nbsp;

&nbsp;       # Extract view

&nbsp;       view = metadata.get("view")

&nbsp;       if view:

&nbsp;           view\_uri = self.\_get\_entity\_uri("view", view)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasView"], view\_uri)

&nbsp;

&nbsp;       # Extract pathology

&nbsp;       pathology = metadata.get("pathology")

&nbsp;       if pathology:

&nbsp;           pathology\_uri = self.\_get\_entity\_uri("pathology", pathology)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasPathology"], pathology\_uri)

&nbsp;

&nbsp;       # Extract assessment

&nbsp;       assessment = metadata.get("assessment")

&nbsp;       if assessment:

&nbsp;           assessment\_uri = self.\_get\_entity\_uri("assessment", assessment)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasAssessment"], assessment\_uri)

&nbsp;

&nbsp;   def \_extract\_histopathology\_entities(self, entry: Dict, patient\_uri: URIRef, study\_uri: URIRef):

&nbsp;       """Extract entities and relationships specific to histopathology"""

&nbsp;       metadata = entry.get("metadata", {})

&nbsp;

&nbsp;       # Extract stain type

&nbsp;       stain = metadata.get("stain\_type")

&nbsp;       if stain:

&nbsp;           stain\_uri = self.\_get\_entity\_uri("stain", stain)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasStainType"], stain\_uri)

&nbsp;

&nbsp;       # Extract magnification

&nbsp;       magnification = metadata.get("magnification")

&nbsp;       if magnification:

&nbsp;           mag\_uri = self.\_get\_entity\_uri("magnification", magnification)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasMagnification"], mag\_uri)

&nbsp;

&nbsp;       # Extract class

&nbsp;       class\_name = metadata.get("class")

&nbsp;       if class\_name:

&nbsp;           class\_uri = self.\_get\_entity\_uri("class", class\_name)

&nbsp;           self.\_add\_relationship(study\_uri, BREAST\["hasClass"], class\_uri)

&nbsp;

&nbsp;   def \_extract\_text\_entities(self, entry: Dict, patient\_uri: URIRef, study\_uri: URIRef):

&nbsp;       """Extract entities and relationships from clinical text"""

&nbsp;       content = entry.get("content", "")

&nbsp;       if not content:

&nbsp;           return

&nbsp;

&nbsp;       # Use NLP to extract medical entities

&nbsp;       doc = nlp(content)

&nbsp;

&nbsp;       # Extract medical entities

&nbsp;       for ent in doc.ents:

&nbsp;           if ent.label\_ in \["DISEASE", "SYMPTOM", "ANATOMY"]:

&nbsp;               entity\_uri = self.\_get\_entity\_uri(ent.label\_.lower(), ent.text)

&nbsp;               self.\_add\_relationship(study\_uri, MED\["hasFinding"], entity\_uri)

&nbsp;

&nbsp;       # Extract relationships using simple patterns

&nbsp;       self.\_extract\_text\_relationships(content, study\_uri)

&nbsp;

&nbsp;   def \_extract\_text\_relationships(self, text: str, study\_uri: URIRef):

&nbsp;       """Extract relationships from text using pattern matching"""

&nbsp;       # Simple pattern-based relationship extraction

&nbsp;       # In a real implementation, this would use more sophisticated NLP techniques

&nbsp;

&nbsp;       # Example patterns

&nbsp;       patterns = \[

&nbsp;           (r"mass located in the (\\w+)", BREAST\["hasLocation"]),

&nbsp;           (r"calcifications in the (\\w+)", BREAST\["hasLocation"]),

&nbsp;           (r"suspicious for (\\w+)", BREAST\["hasFinding"]),

&nbsp;           (r"consistent with (\\w+)", BREAST\["hasDiagnosis"])

&nbsp;       ]

&nbsp;

&nbsp;       import re

&nbsp;       for pattern, predicate in patterns:

&nbsp;           matches = re.findall(pattern, text, re.IGNORECASE)

&nbsp;           for match in matches:

&nbsp;               entity\_uri = self.\_get\_entity\_uri("finding", match)

&nbsp;               self.\_add\_relationship(study\_uri, predicate, entity\_uri)

&nbsp;

&nbsp;   def \_get\_entity\_uri(self, entity\_type: str, entity\_value: str) -> URIRef:

&nbsp;       """Get or create a URI for an entity"""

&nbsp;       # Create a unique identifier for the entity

&nbsp;       entity\_id = f"{entity\_type}\_{hashlib.md5(entity\_value.encode()).hexdigest()}"

&nbsp;

&nbsp;       if entity\_id not in self.entity\_map:

&nbsp;           # Create a new URI for the entity

&nbsp;           uri = URIRef(f"http://example.org/medical/{entity\_id}")

&nbsp;           self.entity\_map\[entity\_id] = uri

&nbsp;

&nbsp;           # Add type information

&nbsp;           self.kg.add((uri, RDF.type, MED\[entity\_type.capitalize()]))

&nbsp;           self.kg.add((uri, RDFS.label, Literal(entity\_value)))

&nbsp;

&nbsp;       return self.entity\_map\[entity\_id]

&nbsp;

&nbsp;   def \_add\_relationship(self, subject: URIRef, predicate: URIRef, obj: Union\[URIRef, Literal]):

&nbsp;       """Add a relationship to the knowledge graph"""

&nbsp;       self.kg.add((subject, predicate, obj))

&nbsp;       self.relationships.append((subject, predicate, obj))

&nbsp;

&nbsp;   def \_build\_rdf\_graph(self):

&nbsp;       """Build the RDF graph from extracted entities and relationships"""

&nbsp;       # The graph is already being built as we add entities and relationships

&nbsp;       # This method can be extended to add additional triples or infer new relationships

&nbsp;       pass

&nbsp;

&nbsp;   def save\_knowledge\_graph(self, output\_path: str):

&nbsp;       """Save the knowledge graph to a file"""

&nbsp;       self.kg.serialize(destination=output\_path, format='turtle')

&nbsp;       print(f"Saved knowledge graph to {output\_path}")

&nbsp;

&nbsp;   def get\_networkx\_graph(self) -> nx.Graph:

&nbsp;       """Convert the knowledge graph to a NetworkX graph for analysis"""

&nbsp;       G = nx.DiGraph()

&nbsp;

&nbsp;       for s, p, o in self.kg:

&nbsp;           if isinstance(o, URIRef):

&nbsp;               G.add\_edge(str(s), str(o), predicate=str(p))

&nbsp;           else:

&nbsp;               # Add literal values as node attributes

&nbsp;               if str(s) not in G.nodes:

&nbsp;                   G.add\_node(str(s))

&nbsp;               G.nodes\[str(s)]\[str(p).split('#')\[-1]] = str(o)

&nbsp;

&nbsp;       return G





class SyntheticDataGenerator:

&nbsp;   """

&nbsp;   Generates synthetic triplets validated by medical knowledge

&nbsp;   """

&nbsp;   def \_\_init\_\_(self, knowledge\_graph: Graph, unified\_data: List\[Dict]):

&nbsp;       self.kg = knowledge\_graph

&nbsp;       self.unified\_data = unified\_data

&nbsp;       self.synthetic\_triplets = \[]

&nbsp;       self.medical\_rules = self.\_define\_medical\_rules()

&nbsp;

&nbsp;   def \_define\_medical\_rules(self) -> List\[Dict]:

&nbsp;       """Define medical rules for validation"""

&nbsp;       return \[

&nbsp;           {

&nbsp;               "name": "malignancy\_association",

&nbsp;               "description": "Malignant tumors are associated with specific findings",

&nbsp;               "pattern": (None, BREAST\["hasFinding"], BREAST\["SpiculatedMass"]),

&nbsp;               "implication": (None, BREAST\["hasMalignancy"], Literal(True, datatype=XSD.boolean))

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "density\_risk",

&nbsp;               "description": "Higher breast density increases cancer risk",

&nbsp;               "pattern": (None, BREAST\["hasBreastDensity"], BREAST\["DensityD"]),

&nbsp;               "implication": (None, MED\["hasRiskFactor"], Literal("high", datatype=XSD.string))

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "calcification\_pattern",

&nbsp;               "description": "Certain calcification patterns indicate malignancy",

&nbsp;               "pattern": (None, BREAST\["hasFinding"], BREAST\["LinearCalcifications"]),

&nbsp;               "implication": (None, BREAST\["hasMalignancy"], Literal(True, datatype=XSD.boolean))

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "magnification\_detail",

&nbsp;               "description": "Higher magnification provides more detail",

&nbsp;               "pattern": (None, BREAST\["hasMagnification"], BREAST\["40x"]),

&nbsp;               "implication": (None, MED\["hasDetailLevel"], Literal("high", datatype=XSD.string))

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "pathology\_correlation",

&nbsp;               "description": "Malignant pathology should correlate with histopathology class",

&nbsp;               "pattern": (None, BREAST\["hasPathology"], BREAST\["Malignant"]),

&nbsp;               "implication": (None, BREAST\["hasClass"], BREAST\["malignant"])

&nbsp;           }

&nbsp;       ]

&nbsp;

&nbsp;   def generate\_synthetic\_triplets(self, num\_triplets: int = 100) -> List\[Tuple]:

&nbsp;       """Generate synthetic triplets validated by medical knowledge"""

&nbsp;       self.synthetic\_triplets = \[]

&nbsp;

&nbsp;       # Extract existing entities and relationships

&nbsp;       entities = set()

&nbsp;       predicates = set()

&nbsp;

&nbsp;       for s, p, o in self.kg:

&nbsp;           entities.add(s)

&nbsp;           if isinstance(o, URIRef):

&nbsp;               entities.add(o)

&nbsp;           predicates.add(p)

&nbsp;

&nbsp;       entities = list(entities)

&nbsp;       predicates = list(predicates)

&nbsp;

&nbsp;       # Generate synthetic triplets

&nbsp;       for \_ in tqdm(range(num\_triplets), desc="Generating synthetic triplets"):

&nbsp;           # Randomly select entities and predicate

&nbsp;           subject = random.choice(entities)

&nbsp;           predicate = random.choice(predicates)

&nbsp;

&nbsp;           # Determine if object should be an entity or literal

&nbsp;           if random.random() < 0.7:  # 70% chance of object being an entity

&nbsp;               obj = random.choice(entities)

&nbsp;           else:  # 30% chance of object being a literal

&nbsp;               obj = self.\_generate\_literal(predicate)

&nbsp;

&nbsp;           # Create triplet

&nbsp;           triplet = (subject, predicate, obj)

&nbsp;

&nbsp;           # Validate triplet against medical rules

&nbsp;           if self.\_validate\_triplet(triplet):

&nbsp;               self.synthetic\_triplets.append(triplet)

&nbsp;

&nbsp;       print(f"Generated {len(self.synthetic\_triplets)} validated synthetic triplets")

&nbsp;       return self.synthetic\_triplets

&nbsp;

&nbsp;   def \_generate\_literal(self, predicate: str) -> Literal:

&nbsp;       """Generate a literal value based on predicate type"""

&nbsp;       # In a real implementation, this would be more sophisticated

&nbsp;       predicate\_name = predicate.split('#')\[-1].lower()

&nbsp;

&nbsp;       if "malignancy" in predicate\_name:

&nbsp;           return Literal(random.choice(\[True, False]), datatype=XSD.boolean)

&nbsp;       elif "density" in predicate\_name:

&nbsp;           return Literal(random.choice(\["A", "B", "C", "D"]), datatype=XSD.string)

&nbsp;       elif "size" in predicate\_name:

&nbsp;           return Literal(round(random.uniform(0.5, 5.0), 1), datatype=XSD.float)

&nbsp;       elif "age" in predicate\_name:

&nbsp;           return Literal(random.randint(30, 80), datatype=XSD.integer)

&nbsp;       elif "magnification" in predicate\_name:

&nbsp;           return Literal(random.choice(\["10x", "20x", "40x"]), datatype=XSD.string)

&nbsp;       else:

&nbsp;           return Literal(f"synthetic\_{random.randint(1000, 9999)}", datatype=XSD.string)

&nbsp;

&nbsp;   def \_validate\_triplet(self, triplet: Tuple) -> bool:

&nbsp;       """Validate a triplet against medical rules"""

&nbsp;       subject, predicate, obj = triplet

&nbsp;

&nbsp;       # Check against each medical rule

&nbsp;       for rule in self.medical\_rules:

&nbsp;           # If the triplet matches the pattern, check if it implies the rule's implication

&nbsp;           if self.\_matches\_pattern(triplet, rule\["pattern"]):

&nbsp;               # Check if the implication is valid

&nbsp;               if not self.\_check\_implication(subject, rule\["implication"]):

&nbsp;                   return False

&nbsp;

&nbsp;       # Additional validation can be added here

&nbsp;       return True

&nbsp;

&nbsp;   def \_matches\_pattern(self, triplet: Tuple, pattern: Tuple) -> bool:

&nbsp;       """Check if a triplet matches a pattern"""

&nbsp;       s, p, o = triplet

&nbsp;       pattern\_s, pattern\_p, pattern\_o = pattern

&nbsp;

&nbsp;       # Check predicate

&nbsp;       if pattern\_p != p and pattern\_p is not None:

&nbsp;           return False

&nbsp;

&nbsp;       # Check object

&nbsp;       if pattern\_o != o and pattern\_o is not None:

&nbsp;           return False

&nbsp;

&nbsp;       return True

&nbsp;

&nbsp;   def \_check\_implication(self, subject: URIRef, implication: Tuple) -> bool:

&nbsp;       """Check if an implication is valid for a subject"""

&nbsp;       impl\_s, impl\_p, impl\_o = implication

&nbsp;

&nbsp;       # In a real implementation, this would check the knowledge graph

&nbsp;       # For now, we'll assume implications are valid

&nbsp;       return True

&nbsp;

&nbsp;   def integrate\_synthetic\_data(self) -> Graph:

&nbsp;       """Integrate synthetic triplets into the knowledge graph"""

&nbsp;       for triplet in self.synthetic\_triplets:

&nbsp;           self.kg.add(triplet)

&nbsp;

&nbsp;       print(f"Integrated {len(self.synthetic\_triplets)} synthetic triplets into knowledge graph")

&nbsp;       return self.kg

&nbsp;

&nbsp;   def save\_synthetic\_triplets(self, output\_path: str):

&nbsp;       """Save synthetic triplets to a file"""

&nbsp;       with open(output\_path, 'w') as f:

&nbsp;           for triplet in self.synthetic\_triplets:

&nbsp;               s, p, o = triplet

&nbsp;               f.write(f"{s} {p} {o}\\n")

&nbsp;       print(f"Saved synthetic triplets to {output\_path}")





\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Set paths for the datasets

&nbsp;   cbis\_path = "data/CBIS-DDSM"

&nbsp;   breakhis\_path = "data/BreakHis"

&nbsp;   output\_path = "output"

&nbsp;

&nbsp;   # Step 1: Preprocess datasets

&nbsp;   print("=== Step 1: Dataset Preprocessing ===")

&nbsp;   os.makedirs(output\_path, exist\_ok=True)

&nbsp;

&nbsp;   # Process datasets

&nbsp;   cbis\_count, breakhis\_count = preprocess\_datasets(cbis\_path, breakhis\_path, output\_path)

&nbsp;

&nbsp;   # Step 2: Multi-Modal Harmonization

&nbsp;   print("\\n=== Step 2: Multi-Modal Harmonization ===")

&nbsp;   harmonizer = MultiModalHarmonizer()

&nbsp;

&nbsp;   # Register data sources

&nbsp;   harmonizer.register\_data\_source("mammogram", os.path.join(output\_path, "mammograms"))

&nbsp;   harmonizer.register\_data\_source("histopathology", os.path.join(output\_path, "histopathology"))

&nbsp;   harmonizer.register\_data\_source("clinical\_text", os.path.join(output\_path, "reports"))

&nbsp;

&nbsp;   # Create unified schema

&nbsp;   schema = harmonizer.create\_unified\_schema()

&nbsp;   print(f"Created unified schema with {len(schema)} fields")

&nbsp;

&nbsp;   # Harmonize data

&nbsp;   harmonized\_data = harmonizer.harmonize\_data()

&nbsp;

&nbsp;   # Save harmonized data

&nbsp;   harmonizer.save\_harmonized\_data(os.path.join(output\_path, "harmonized\_data.json"))

&nbsp;

&nbsp;   # Step 3: Knowledge-Graph Augmentation

&nbsp;   print("\\n=== Step 3: Knowledge-Graph Augmentation ===")

&nbsp;   kg\_augmentor = KnowledgeGraphAugmentor(harmonized\_data)

&nbsp;   knowledge\_graph = kg\_augmentor.build\_knowledge\_graph()

&nbsp;

&nbsp;   # Save knowledge graph

&nbsp;   kg\_augmentor.save\_knowledge\_graph(os.path.join(output\_path, "knowledge\_graph.ttl"))

&nbsp;

&nbsp;   # Convert to NetworkX for analysis

&nbsp;   nx\_graph = kg\_augmentor.get\_networkx\_graph()

&nbsp;   print(f"Created NetworkX graph with {nx\_graph.number\_of\_nodes()} nodes and {nx\_graph.number\_of\_edges()} edges")

&nbsp;

&nbsp;   # Step 4: Synthetic Data Generation

&nbsp;   print("\\n=== Step 4: Synthetic Data Generation ===")

&nbsp;   synthetic\_generator = SyntheticDataGenerator(knowledge\_graph, harmonized\_data)

&nbsp;   synthetic\_triplets = synthetic\_generator.generate\_synthetic\_triplets(num\_triplets=50)

&nbsp;

&nbsp;   # Integrate synthetic data into knowledge graph

&nbsp;   updated\_kg = synthetic\_generator.integrate\_synthetic\_data()

&nbsp;

&nbsp;   # Save synthetic triplets

&nbsp;   synthetic\_generator.save\_synthetic\_triplets(os.path.join(output\_path, "synthetic\_triplets.txt"))

&nbsp;

&nbsp;   # Save updated knowledge graph

&nbsp;   updated\_kg.serialize(destination=os.path.join(output\_path, "updated\_knowledge\_graph.ttl"), format='turtle')

&nbsp;

&nbsp;   print("\\nDataset Processing Pipeline completed successfully!")










\# Set device

device = torch.device('cuda' if torch.cuda.is\_available() else 'cpu')



class MedicalImageAugmentation:

&nbsp;   """Domain-specific augmentations for medical images"""

&nbsp;   def \_\_init\_\_(self, modality: str = "mammogram"):

&nbsp;       self.modality = modality

&nbsp;

&nbsp;       if modality == "mammogram":

&nbsp;           # Mammogram-specific augmentations

&nbsp;           self.transform = transforms.Compose(\[

&nbsp;               transforms.RandomHorizontalFlip(p=0.5),

&nbsp;               transforms.RandomRotation(degrees=(-15, 15)),

&nbsp;               transforms.ColorJitter(brightness=0.1, contrast=0.1),

&nbsp;               transforms.RandomAffine(degrees=0, translate=(0.05, 0.05)),

&nbsp;               transforms.RandomApply(\[transforms.GaussianBlur(kernel\_size=3, sigma=(0.1, 0.5))], p=0.3),

&nbsp;               transforms.ToTensor(),

&nbsp;               transforms.Normalize(mean=\[0.485, 0.456, 0.406], std=\[0.229, 0.224, 0.225])

&nbsp;           ])

&nbsp;       elif modality == "histopathology":

&nbsp;           # Histopathology-specific augmentations

&nbsp;           self.transform = transforms.Compose(\[

&nbsp;               transforms.RandomHorizontalFlip(p=0.5),

&nbsp;               transforms.RandomVerticalFlip(p=0.5),

&nbsp;               transforms.RandomRotation(degrees=(-90, 90)),

&nbsp;               transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1, hue=0.05),

&nbsp;               transforms.RandomApply(\[transforms.GaussianBlur(kernel\_size=3, sigma=(0.1, 1.0))], p=0.5),

&nbsp;               transforms.ToTensor(),

&nbsp;               transforms.Normalize(mean=\[0.707, 0.582, 0.703], std=\[0.190, 0.226, 0.188])

&nbsp;           ])

&nbsp;       else:

&nbsp;           # Default augmentations

&nbsp;           self.transform = transforms.Compose(\[

&nbsp;               transforms.RandomHorizontalFlip(p=0.5),

&nbsp;               transforms.RandomRotation(degrees=(-10, 10)),

&nbsp;               transforms.ColorJitter(brightness=0.1, contrast=0.1),

&nbsp;               transforms.ToTensor(),

&nbsp;               transforms.Normalize(mean=\[0.485, 0.456, 0.406], std=\[0.229, 0.224, 0.225])

&nbsp;           ])

&nbsp;

&nbsp;   def \_\_call\_\_(self, img):

&nbsp;       return self.transform(img)



class ContrastiveLearningModel(nn.Module):

&nbsp;   """Self-supervised contrastive learning model (SimCLR/BYOL style)"""

&nbsp;   def \_\_init\_\_(self, base\_encoder: str = "resnet50", projection\_dim: int = 128,

&nbsp;                temperature: float = 0.07, momentum: float = 0.999):

&nbsp;       super(ContrastiveLearningModel, self).\_\_init\_\_()

&nbsp;       self.temperature = temperature

&nbsp;       self.momentum = momentum

&nbsp;

&nbsp;       # Create encoder and momentum encoder

&nbsp;       if base\_encoder == "resnet50":

&nbsp;           self.encoder = models.resnet50(pretrained=False)

&nbsp;           self.momentum\_encoder = models.resnet50(pretrained=False)

&nbsp;           feature\_dim = self.encoder.fc.in\_features

&nbsp;           self.encoder.fc = nn.Identity()  # Remove final classification layer

&nbsp;           self.momentum\_encoder.fc = nn.Identity()

&nbsp;       elif base\_encoder == "densenet121":

&nbsp;           self.encoder = models.densenet121(pretrained=False)

&nbsp;           self.momentum\_encoder = models.densenet121(pretrained=False)

&nbsp;           feature\_dim = self.encoder.classifier.in\_features

&nbsp;           self.encoder.classifier = nn.Identity()

&nbsp;           self.momentum\_encoder.classifier = nn.Identity()

&nbsp;       else:

&nbsp;           raise ValueError(f"Unsupported base encoder: {base\_encoder}")

&nbsp;

&nbsp;       # Projection head

&nbsp;       self.projection\_head = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, feature\_dim),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Linear(feature\_dim, projection\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Momentum projection head

&nbsp;       self.momentum\_projection\_head = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, feature\_dim),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Linear(feature\_dim, projection\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Initialize momentum encoder with encoder weights

&nbsp;       self.\_initialize\_momentum\_encoder()

&nbsp;

&nbsp;       # Memory bank for hard negatives

&nbsp;       self.register\_buffer("memory\_bank", torch.randn(10000, projection\_dim))

&nbsp;       self.memory\_bank = F.normalize(self.memory\_bank, dim=1)

&nbsp;       self.register\_buffer("memory\_ptr", torch.zeros(1, dtype=torch.long))

&nbsp;

&nbsp;   def \_initialize\_momentum\_encoder(self):

&nbsp;       """Initialize momentum encoder with encoder weights"""

&nbsp;       for param\_q, param\_k in zip(self.encoder.parameters(), self.momentum\_encoder.parameters()):

&nbsp;           param\_k.data.copy\_(param\_q.data)

&nbsp;           param\_k.requires\_grad = False

&nbsp;

&nbsp;       for param\_q, param\_k in zip(self.projection\_head.parameters(), self.momentum\_projection\_head.parameters()):

&nbsp;           param\_k.data.copy\_(param\_q.data)

&nbsp;           param\_k.requires\_grad = False

&nbsp;

&nbsp;   @torch.no\_grad()

&nbsp;   def \_update\_momentum\_encoder(self):

&nbsp;       """Update momentum encoder using exponential moving average"""

&nbsp;       for param\_q, param\_k in zip(self.encoder.parameters(), self.momentum\_encoder.parameters()):

&nbsp;           param\_k.data = param\_k.data \* self.momentum + param\_q.data \* (1. - self.momentum)

&nbsp;

&nbsp;       for param\_q, param\_k in zip(self.projection\_head.parameters(), self.momentum\_projection\_head.parameters()):

&nbsp;           param\_k.data = param\_k.data \* self.momentum + param\_q.data \* (1. - self.momentum)

&nbsp;

&nbsp;   @torch.no\_grad()

&nbsp;   def \_update\_memory\_bank(self, keys):

&nbsp;       """Update memory bank with new keys"""

&nbsp;       batch\_size = keys.shape\[0]

&nbsp;       ptr = int(self.memory\_ptr)

&nbsp;

&nbsp;       # Replace the keys at ptr with new keys

&nbsp;       if ptr + batch\_size > self.memory\_bank.shape\[0]:

&nbsp;           # Handle wrap-around

&nbsp;           self.memory\_bank\[ptr:] = keys\[:self.memory\_bank.shape\[0] - ptr]

&nbsp;           self.memory\_bank\[:batch\_size - (self.memory\_bank.shape\[0] - ptr)] = keys\[self.memory\_bank.shape\[0] - ptr:]

&nbsp;           ptr = batch\_size - (self.memory\_bank.shape\[0] - ptr)

&nbsp;       else:

&nbsp;           self.memory\_bank\[ptr:ptr + batch\_size] = keys

&nbsp;           ptr += batch\_size

&nbsp;

&nbsp;       self.memory\_ptr\[0] = ptr % self.memory\_bank.shape\[0]

&nbsp;

&nbsp;   def forward(self, x\_q, x\_k=None):

&nbsp;       """Forward pass for contrastive learning"""

&nbsp;       # Compute query features

&nbsp;       q = self.encoder(x\_q)

&nbsp;       q = self.projection\_head(q)

&nbsp;       q = F.normalize(q, dim=1)

&nbsp;

&nbsp;       # Compute key features if x\_k is provided

&nbsp;       if x\_k is not None:

&nbsp;           with torch.no\_grad():

&nbsp;               # Update momentum encoder

&nbsp;               self.\_update\_momentum\_encoder()

&nbsp;

&nbsp;               # Compute key features

&nbsp;               k = self.momentum\_encoder(x\_k)

&nbsp;               k = self.momentum\_projection\_head(k)

&nbsp;               k = F.normalize(k, dim=1)

&nbsp;

&nbsp;               # Update memory bank

&nbsp;               self.\_update\_memory\_bank(k)

&nbsp;

&nbsp;               # Get hard negatives from memory bank

&nbsp;               batch\_size = q.shape\[0]

&nbsp;               indices = torch.randint(0, self.memory\_bank.shape\[0], size=(batch\_size \* 10,))

&nbsp;               hard\_negatives = self.memory\_bank\[indices].view(batch\_size, 10, -1)

&nbsp;

&nbsp;               return q, k, hard\_negatives

&nbsp;

&nbsp;       return q

&nbsp;

&nbsp;   def compute\_contrastive\_loss(self, q, k, hard\_negatives):

&nbsp;       """Compute contrastive loss with hard negatives"""

&nbsp;       batch\_size = q.shape\[0]

&nbsp;

&nbsp;       # Positive logits: dot product between q and k

&nbsp;       l\_pos = torch.einsum('nc,nc->n', \[q, k]).unsqueeze(-1)

&nbsp;

&nbsp;       # Negative logits: dot product between q and hard negatives

&nbsp;       l\_neg = torch.einsum('nc,nck->nk', \[q, hard\_negatives])

&nbsp;

&nbsp;       # Concatenate positive and negative logits

&nbsp;       logits = torch.cat(\[l\_pos, l\_neg], dim=1)

&nbsp;

&nbsp;       # Apply temperature scaling

&nbsp;       logits /= self.temperature

&nbsp;

&nbsp;       # Create labels (first position is positive)

&nbsp;       labels = torch.zeros(batch\_size, dtype=torch.long).to(device)

&nbsp;

&nbsp;       # Compute cross-entropy loss

&nbsp;       loss = F.cross\_entropy(logits, labels)

&nbsp;

&nbsp;       return loss



class MammogramModule(nn.Module):

&nbsp;   """CNN + Attention module for mammogram feature extraction"""

&nbsp;   def \_\_init\_\_(self, pretrained: bool = True, feature\_dim: int = 512):

&nbsp;       super(MammogramModule, self).\_\_init\_\_()

&nbsp;

&nbsp;       # Use a pre-trained ResNet as the backbone

&nbsp;       self.backbone = models.resnet50(pretrained=pretrained)

&nbsp;       self.backbone.fc = nn.Identity()  # Remove final classification layer

&nbsp;

&nbsp;       # Attention mechanism

&nbsp;       self.attention = nn.Sequential(

&nbsp;           nn.Conv2d(2048, 512, kernel\_size=1),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Conv2d(512, 1, kernel\_size=1),

&nbsp;           nn.Sigmoid()

&nbsp;       )

&nbsp;

&nbsp;       # Feature projection

&nbsp;       self.projection = nn.Sequential(

&nbsp;           nn.Linear(2048, feature\_dim),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Dropout(0.3),

&nbsp;           nn.Linear(feature\_dim, feature\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Initialize weights

&nbsp;       self.\_initialize\_weights()

&nbsp;

&nbsp;   def \_initialize\_weights(self):

&nbsp;       """Initialize weights for attention mechanism"""

&nbsp;       for m in self.attention:

&nbsp;           if isinstance(m, nn.Conv2d):

&nbsp;               nn.init.kaiming\_normal\_(m.weight, mode='fan\_out', nonlinearity='relu')

&nbsp;               if m.bias is not None:

&nbsp;                   nn.init.constant\_(m.bias, 0)

&nbsp;

&nbsp;   def forward(self, x):

&nbsp;       # Extract features using backbone

&nbsp;       features = self.backbone.conv1(x)

&nbsp;       features = self.backbone.bn1(features)

&nbsp;       features = self.backbone.relu(features)

&nbsp;       features = self.backbone.maxpool(features)

&nbsp;

&nbsp;       features = self.backbone.layer1(features)

&nbsp;       features = self.backbone.layer2(features)

&nbsp;       features = self.backbone.layer3(features)

&nbsp;       features = self.backbone.layer4(features)

&nbsp;

&nbsp;       # Apply attention mechanism

&nbsp;       attention\_weights = self.attention(features)

&nbsp;       attended\_features = features \* attention\_weights

&nbsp;

&nbsp;       # Global average pooling

&nbsp;       pooled\_features = F.adaptive\_avg\_pool2d(attended\_features, (1, 1))

&nbsp;       pooled\_features = torch.flatten(pooled\_features, 1)

&nbsp;

&nbsp;       # Project features

&nbsp;       output = self.projection(pooled\_features)

&nbsp;

&nbsp;       return output, attention\_weights



class HistopathologyModule(nn.Module):

&nbsp;   """ViT + MIL module for histopathology feature extraction"""

&nbsp;   def \_\_init\_\_(self, img\_size: int = 224, patch\_size: int = 16,

&nbsp;                embed\_dim: int = 768, depth: int = 12, num\_heads: int = 12,

&nbsp;                feature\_dim: int = 512, pretrained: bool = True):

&nbsp;       super(HistopathologyModule, self).\_\_init\_\_()

&nbsp;

&nbsp;       # Use a pre-trained Vision Transformer as the backbone

&nbsp;       if pretrained:

&nbsp;           self.vit = AutoModel.from\_pretrained('google/vit-base-patch16-224')

&nbsp;       else:

&nbsp;           from transformers import ViTModel

&nbsp;           self.vit = ViTModel(

&nbsp;               image\_size=img\_size,

&nbsp;               patch\_size=patch\_size,

&nbsp;               embed\_dim=embed\_dim,

&nbsp;               depth=depth,

&nbsp;               num\_heads=num\_heads

&nbsp;           )

&nbsp;

&nbsp;       # MIL aggregation mechanism

&nbsp;       self.attention\_pooling = nn.Sequential(

&nbsp;           nn.Linear(embed\_dim, 128),

&nbsp;           nn.Tanh(),

&nbsp;           nn.Linear(128, 1)

&nbsp;       )

&nbsp;

&nbsp;       # Feature projection

&nbsp;       self.projection = nn.Sequential(

&nbsp;           nn.Linear(embed\_dim, feature\_dim),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Dropout(0.3),

&nbsp;           nn.Linear(feature\_dim, feature\_dim)

&nbsp;       )

&nbsp;

&nbsp;   def forward(self, patches):

&nbsp;       # Process patches through ViT

&nbsp;       batch\_size, num\_patches, C, H, W = patches.shape

&nbsp;       patches = patches.view(batch\_size \* num\_patches, C, H, W)

&nbsp;

&nbsp;       # Extract patch features using ViT

&nbsp;       outputs = self.vit(patches)

&nbsp;       patch\_features = outputs.last\_hidden\_state\[:, 0, :]  # CLS token

&nbsp;       patch\_features = patch\_features.view(batch\_size, num\_patches, -1)

&nbsp;

&nbsp;       # Apply attention-based MIL pooling

&nbsp;       attention\_scores = self.attention\_pooling(patch\_features)

&nbsp;       attention\_weights = F.softmax(attention\_scores, dim=1)

&nbsp;

&nbsp;       # Aggregate patch features

&nbsp;       aggregated\_features = torch.sum(patch\_features \* attention\_weights, dim=1)

&nbsp;

&nbsp;       # Project features

&nbsp;       output = self.projection(aggregated\_features)

&nbsp;

&nbsp;       return output, attention\_weights



class ClinicalTextModule(nn.Module):

&nbsp;   """BioBERT module for clinical text feature extraction"""

&nbsp;   def \_\_init\_\_(self, model\_name: str = "dmis-lab/biobert-v1.1", feature\_dim: int = 512):

&nbsp;       super(ClinicalTextModule, self).\_\_init\_\_()

&nbsp;

&nbsp;       # Load pre-trained BioBERT model

&nbsp;       self.tokenizer = AutoTokenizer.from\_pretrained(model\_name)

&nbsp;       self.bert = AutoModel.from\_pretrained(model\_name)

&nbsp;

&nbsp;       # Freeze some layers to reduce computation

&nbsp;       for param in self.bert.embeddings.parameters():

&nbsp;           param.requires\_grad = False

&nbsp;

&nbsp;       for i in range(8):  # Freeze first 8 layers

&nbsp;           for param in self.bert.encoder.layer\[i].parameters():

&nbsp;               param.requires\_grad = False

&nbsp;

&nbsp;       # Feature projection

&nbsp;       self.projection = nn.Sequential(

&nbsp;           nn.Linear(768, feature\_dim),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Dropout(0.3),

&nbsp;           nn.Linear(feature\_dim, feature\_dim)

&nbsp;       )

&nbsp;

&nbsp;   def forward(self, texts):

&nbsp;       # Tokenize texts

&nbsp;       encoded\_input = self.tokenizer(

&nbsp;           texts,

&nbsp;           padding=True,

&nbsp;           truncation=True,

&nbsp;           max\_length=512,

&nbsp;           return\_tensors="pt"

&nbsp;       ).to(device)

&nbsp;

&nbsp;       # Extract features using BioBERT

&nbsp;       outputs = self.bert(\*\*encoded\_input)

&nbsp;

&nbsp;       # Use CLS token as the sentence representation

&nbsp;       text\_features = outputs.last\_hidden\_state\[:, 0, :]

&nbsp;

&nbsp;       # Project features

&nbsp;       output = self.projection(text\_features)

&nbsp;

&nbsp;       return output



class MultiModalDataset(Dataset):

&nbsp;   """Dataset for multi-modal breast cancer data"""

&nbsp;   def \_\_init\_\_(self, data\_dir: str, split: str = "train", transform=None):

&nbsp;       self.data\_dir = data\_dir

&nbsp;       self.split = split

&nbsp;       self.transform = transform

&nbsp;

&nbsp;       # Load metadata

&nbsp;       metadata\_path = os.path.join(data\_dir, "harmonized\_data.json")

&nbsp;       with open(metadata\_path, 'r') as f:

&nbsp;           self.data = json.load(f)

&nbsp;

&nbsp;       # Filter by split if needed

&nbsp;       if split != "all":

&nbsp;           # In a real implementation, we would have a split file

&nbsp;           # For now, we'll use a simple random split

&nbsp;           random.shuffle(self.data)

&nbsp;           split\_idx = int(len(self.data) \* 0.8)

&nbsp;           if split == "train":

&nbsp;               self.data = self.data\[:split\_idx]

&nbsp;           else:

&nbsp;               self.data = self.data\[split\_idx:]

&nbsp;

&nbsp;   def \_\_len\_\_(self):

&nbsp;       return len(self.data)

&nbsp;

&nbsp;   def \_\_getitem\_\_(self, idx):

&nbsp;       item = self.data\[idx]

&nbsp;       modality = item\["modality"]

&nbsp;

&nbsp;       if modality == "mammogram":

&nbsp;           # Load mammogram image

&nbsp;           img\_path = item\["data\_path"]

&nbsp;           image = Image.open(img\_path).convert('RGB')

&nbsp;

&nbsp;           if self.transform:

&nbsp;               image = self.transform(image)

&nbsp;

&nbsp;           # Get label

&nbsp;           label = item.get("labels", {}).get("malignancy", 0)

&nbsp;

&nbsp;           return {

&nbsp;               "modality": "mammogram",

&nbsp;               "image": image,

&nbsp;               "label": label,

&nbsp;               "patient\_id": item\["patient\_id"],

&nbsp;               "study\_id": item\["study\_id"]

&nbsp;           }

&nbsp;

&nbsp;       elif modality == "histopathology":

&nbsp;           # Load histopathology image

&nbsp;           img\_path = item\["data\_path"]

&nbsp;           image = Image.open(img\_path).convert('RGB')

&nbsp;

&nbsp;           if self.transform:

&nbsp;               image = self.transform(image)

&nbsp;

&nbsp;           # Get label

&nbsp;           label = item.get("labels", {}).get("malignancy", 0)

&nbsp;

&nbsp;           return {

&nbsp;               "modality": "histopathology",

&nbsp;               "image": image,

&nbsp;               "label": label,

&nbsp;               "patient\_id": item\["patient\_id"],

&nbsp;               "study\_id": item\["study\_id"]

&nbsp;           }

&nbsp;

&nbsp;       elif modality == "clinical\_text":

&nbsp;           # Load clinical text

&nbsp;           text = item.get("content", "")

&nbsp;

&nbsp;           # Get label

&nbsp;           label = item.get("labels", {}).get("malignancy", 0)

&nbsp;

&nbsp;           return {

&nbsp;               "modality": "clinical\_text",

&nbsp;               "text": text,

&nbsp;               "label": label,

&nbsp;               "patient\_id": item\["patient\_id"],

&nbsp;               "study\_id": item\["study\_id"]

&nbsp;           }

&nbsp;

&nbsp;       else:

&nbsp;           raise ValueError(f"Unknown modality: {modality}")



class MultiModalProcessor:

&nbsp;   """Multi-modal input processing with self-supervised contrastive learning"""

&nbsp;   def \_\_init\_\_(self, data\_dir: str, config: Dict = None):

&nbsp;       self.data\_dir = data\_dir

&nbsp;       self.config = config or {}

&nbsp;

&nbsp;       # Initialize augmentations

&nbsp;       self.mammogram\_aug = MedicalImageAugmentation(modality="mammogram")

&nbsp;       self.histopathology\_aug = MedicalImageAugmentation(modality="histopathology")

&nbsp;

&nbsp;       # Initialize models

&nbsp;       self.\_initialize\_models()

&nbsp;

&nbsp;       # Initialize datasets and dataloaders

&nbsp;       self.\_initialize\_dataloaders()

&nbsp;

&nbsp;   def \_initialize\_models(self):

&nbsp;       """Initialize all models"""

&nbsp;       # Contrastive learning model for mammograms

&nbsp;       self.contrastive\_model = ContrastiveLearningModel(

&nbsp;           base\_encoder=self.config.get("base\_encoder", "resnet50"),

&nbsp;           projection\_dim=self.config.get("projection\_dim", 128),

&nbsp;           temperature=self.config.get("temperature", 0.07),

&nbsp;           momentum=self.config.get("momentum", 0.999)

&nbsp;       ).to(device)

&nbsp;

&nbsp;       # Mammogram module

&nbsp;       self.mammogram\_module = MammogramModule(

&nbsp;           pretrained=self.config.get("pretrained", True),

&nbsp;           feature\_dim=self.config.get("feature\_dim", 512)

&nbsp;       ).to(device)

&nbsp;

&nbsp;       # Histopathology module

&nbsp;       self.histopathology\_module = HistopathologyModule(

&nbsp;           img\_size=self.config.get("img\_size", 224),

&nbsp;           patch\_size=self.config.get("patch\_size", 16),

&nbsp;           embed\_dim=self.config.get("embed\_dim", 768),

&nbsp;           depth=self.config.get("depth", 12),

&nbsp;           num\_heads=self.config.get("num\_heads", 12),

&nbsp;           feature\_dim=self.config.get("feature\_dim", 512),

&nbsp;           pretrained=self.config.get("pretrained", True)

&nbsp;       ).to(device)

&nbsp;

&nbsp;       # Clinical text module

&nbsp;       self.clinical\_text\_module = ClinicalTextModule(

&nbsp;           model\_name=self.config.get("text\_model", "dmis-lab/biobert-v1.1"),

&nbsp;           feature\_dim=self.config.get("feature\_dim", 512)

&nbsp;       ).to(device)

&nbsp;

&nbsp;   def \_initialize\_dataloaders(self):

&nbsp;       """Initialize datasets and dataloaders"""

&nbsp;       # Train dataset

&nbsp;       train\_dataset = MultiModalDataset(

&nbsp;           data\_dir=self.data\_dir,

&nbsp;           split="train",

&nbsp;           transform=None  # We'll apply augmentations in the training loop

&nbsp;       )

&nbsp;

&nbsp;       # Validation dataset

&nbsp;       val\_dataset = MultiModalDataset(

&nbsp;           data\_dir=self.data\_dir,

&nbsp;           split="val",

&nbsp;           transform=transforms.Compose(\[

&nbsp;               transforms.ToTensor(),

&nbsp;               transforms.Normalize(mean=\[0.485, 0.456, 0.406], std=\[0.229, 0.224, 0.225])

&nbsp;           ])

&nbsp;       )

&nbsp;

&nbsp;       # Create dataloaders

&nbsp;       self.train\_loader = DataLoader(

&nbsp;           train\_dataset,

&nbsp;           batch\_size=self.config.get("batch\_size", 16),

&nbsp;           shuffle=True,

&nbsp;           num\_workers=self.config.get("num\_workers", 4),

&nbsp;           pin\_memory=True

&nbsp;       )

&nbsp;

&nbsp;       self.val\_loader = DataLoader(

&nbsp;           val\_dataset,

&nbsp;           batch\_size=self.config.get("batch\_size", 16),

&nbsp;           shuffle=False,

&nbsp;           num\_workers=self.config.get("num\_workers", 4),

&nbsp;           pin\_memory=True

&nbsp;       )

&nbsp;

&nbsp;   def pretrain\_contrastive(self, epochs: int = 10, lr: float = 1e-4):

&nbsp;       """Pre-train mammogram module using self-supervised contrastive learning"""

&nbsp;       print("Starting contrastive pre-training for mammograms...")

&nbsp;

&nbsp;       # Optimizer

&nbsp;       optimizer = torch.optim.Adam(

&nbsp;           self.contrastive\_model.parameters(),

&nbsp;           lr=lr,

&nbsp;           weight\_decay=1e-4

&nbsp;       )

&nbsp;

&nbsp;       # Learning rate scheduler

&nbsp;       scheduler = torch.optim.lr\_scheduler.CosineAnnealingLR(

&nbsp;           optimizer, T\_max=epochs, eta\_min=lr/100

&nbsp;       )

&nbsp;

&nbsp;       # Training loop

&nbsp;       for epoch in range(epochs):

&nbsp;           self.contrastive\_model.train()

&nbsp;           total\_loss = 0.0

&nbsp;

&nbsp;           for batch in tqdm(self.train\_loader, desc=f"Epoch {epoch+1}/{epochs}"):

&nbsp;               # Skip non-mammogram data

&nbsp;               if batch\["modality"]\[0] != "mammogram":

&nbsp;                   continue

&nbsp;

&nbsp;               # Get images

&nbsp;               images = batch\["image"].to(device)

&nbsp;

&nbsp;               # Apply different augmentations to create two views

&nbsp;               x\_q = torch.stack(\[self.mammogram\_aug(Image.fromarray((img.permute(1,2,0).cpu().numpy() \* 255).astype(np.uint8))) for img in images])

&nbsp;               x\_k = torch.stack(\[self.mammogram\_aug(Image.fromarray((img.permute(1,2,0).cpu().numpy() \* 255).astype(np.uint8))) for img in images])

&nbsp;

&nbsp;               # Forward pass

&nbsp;               q, k, hard\_negatives = self.contrastive\_model(x\_q, x\_k)

&nbsp;

&nbsp;               # Compute contrastive loss

&nbsp;               loss = self.contrastive\_model.compute\_contrastive\_loss(q, k, hard\_negatives)

&nbsp;

&nbsp;               # Backward pass

&nbsp;               optimizer.zero\_grad()

&nbsp;               loss.backward()

&nbsp;               optimizer.step()

&nbsp;

&nbsp;               total\_loss += loss.item()

&nbsp;

&nbsp;           # Update learning rate

&nbsp;           scheduler.step()

&nbsp;

&nbsp;           # Print epoch statistics

&nbsp;           avg\_loss = total\_loss / len(self.train\_loader)

&nbsp;           print(f"Epoch {epoch+1}/{epochs}, Loss: {avg\_loss:.4f}")

&nbsp;

&nbsp;       # Copy weights to mammogram module

&nbsp;       self.\_copy\_contrastive\_weights()

&nbsp;

&nbsp;       print("Contrastive pre-training completed!")

&nbsp;

&nbsp;   def \_copy\_contrastive\_weights(self):

&nbsp;       """Copy weights from contrastive model to mammogram module"""

&nbsp;       # Copy encoder weights

&nbsp;       for (name\_q, param\_q), (name\_k, param\_k) in zip(

&nbsp;           self.contrastive\_model.encoder.named\_parameters(),

&nbsp;           self.mammogram\_module.backbone.named\_parameters()

&nbsp;       ):

&nbsp;           if name\_q == name\_k:

&nbsp;               param\_k.data.copy\_(param\_q.data)

&nbsp;

&nbsp;   def train\_mammogram\_module(self, epochs: int = 10, lr: float = 1e-4):

&nbsp;       """Train mammogram module with supervised learning"""

&nbsp;       print("Training mammogram module...")

&nbsp;

&nbsp;       # Optimizer

&nbsp;       optimizer = torch.optim.Adam(

&nbsp;           self.mammogram\_module.parameters(),

&nbsp;           lr=lr,

&nbsp;           weight\_decay=1e-4

&nbsp;       )

&nbsp;

&nbsp;       # Loss function

&nbsp;       criterion = nn.CrossEntropyLoss().to(device) # Move criterion to device

&nbsp;

&nbsp;       # Training loop

&nbsp;       for epoch in range(epochs):

&nbsp;           self.mammogram\_module.train()

&nbsp;           total\_loss = 0.0

&nbsp;           correct = 0

&nbsp;           total = 0

&nbsp;

&nbsp;           for batch in tqdm(self.train\_loader, desc=f"Epoch {epoch+1}/{epochs}"):

&nbsp;               # Skip non-mammogram data

&nbsp;               if batch\["modality"]\[0] != "mammogram":

&nbsp;                   continue

&nbsp;

&nbsp;               # Get images and labels

&nbsp;               images = batch\["image"].to(device)

&nbsp;               labels = batch\["label"].to(device)

&nbsp;

&nbsp;               # Apply augmentation

&nbsp;               images = torch.stack(\[self.mammogram\_aug(Image.fromarray((img.permute(1,2,0).cpu().numpy() \* 255).astype(np.uint8))) for img in images])

&nbsp;

&nbsp;               # Forward pass

&nbsp;               features, \_ = self.mammogram\_module(images)

&nbsp;

&nbsp;               # Simple linear classifier for training

&nbsp;               logits = nn.Linear(512, 2).to(device)(features) # Move classifier to device

&nbsp;

&nbsp;               # Compute loss

&nbsp;               loss = criterion(logits, labels)

&nbsp;

&nbsp;               # Backward pass

&nbsp;               optimizer.zero\_grad()

&nbsp;               loss.backward()

&nbsp;               optimizer.step()

&nbsp;

&nbsp;               # Statistics

&nbsp;               total\_loss += loss.item()

&nbsp;               \_, predicted = torch.max(logits.data, 1)

&nbsp;               total += labels.size(0)

&nbsp;               correct += (predicted == labels).sum().item()

&nbsp;

&nbsp;           # Print epoch statistics

&nbsp;           avg\_loss = total\_loss / len(self.train\_loader)

&nbsp;           accuracy = 100 \* correct / total

&nbsp;           print(f"Epoch {epoch+1}/{epochs}, Loss: {avg\_loss:.4f}, Accuracy: {accuracy:.2f}%")

&nbsp;

&nbsp;       print("Mammogram module training completed!")

&nbsp;

&nbsp;   def train\_histopathology\_module(self, epochs: int = 10, lr: float = 1e-4):

&nbsp;       """Train histopathology module with supervised learning"""

&nbsp;       print("Training histopathology module...")

&nbsp;

&nbsp;       # Optimizer

&nbsp;       optimizer = torch.optim.Adam(

&nbsp;           self.histopathology\_module.parameters(),

&nbsp;           lr=lr,

&nbsp;           weight\_decay=1e-4

&nbsp;       )

&nbsp;

&nbsp;       # Loss function

&nbsp;       criterion = nn.CrossEntropyLoss().to(device) # Move criterion to device

&nbsp;

&nbsp;       # Training loop

&nbsp;       for epoch in range(epochs):

&nbsp;           self.histopathology\_module.train()

&nbsp;           total\_loss = 0.0

&nbsp;           correct = 0

&nbsp;           total = 0

&nbsp;

&nbsp;           for batch in tqdm(self.train\_loader, desc=f"Epoch {epoch+1}/{epochs}"):

&nbsp;               # Skip non-histopathology data

&nbsp;               if batch\["modality"]\[0] != "histopathology":

&nbsp;                   continue

&nbsp;

&nbsp;               # Get images and labels

&nbsp;               images = batch\["image"].to(device)

&nbsp;               labels = batch\["label"].to(device)

&nbsp;

&nbsp;               # Apply augmentation

&nbsp;               images = torch.stack(\[self.histopathology\_aug(Image.fromarray((img.permute(1,2,0).cpu().numpy() \* 255).astype(np.uint8))) for img in images])

&nbsp;

&nbsp;               # Create patches (simplified - in practice, you'd extract patches from WSI)

&nbsp;               batch\_size = images.shape\[0]

&nbsp;               num\_patches = 16  # Use 16 patches per image

&nbsp;               patches = \[]

&nbsp;

&nbsp;               for i in range(batch\_size):

&nbsp;                   img = images\[i]

&nbsp;                   # Extract random patches

&nbsp;                   for \_ in range(num\_patches):

&nbsp;                       h, w = img.shape\[1], img.shape\[2]

&nbsp;                       th, tw = 224, 224  # Patch size

&nbsp;

&nbsp;                       i1 = torch.randint(0, h - th + 1, size=(1,)).item()

&nbsp;                       j1 = torch.randint(0, w - tw + 1, size=(1,)).item()

&nbsp;

&nbsp;                       patch = img\[:, i1:i1+th, j1:j1+tw]

&nbsp;                       patches.append(patch)

&nbsp;

&nbsp;               patches = torch.stack(patches)

&nbsp;               patches = patches.view(batch\_size, num\_patches, 3, 224, 224)

&nbsp;

&nbsp;               # Forward pass

&nbsp;               features, \_ = self.histopathology\_module(patches)

&nbsp;

&nbsp;               # Simple linear classifier for training

&nbsp;               logits = nn.Linear(512, 2).to(device)(features) # Move classifier to device

&nbsp;

&nbsp;               # Compute loss

&nbsp;               loss = criterion(logits, labels)

&nbsp;

&nbsp;               # Backward pass

&nbsp;               optimizer.zero\_grad()

&nbsp;               loss.backward()

&nbsp;               optimizer.step()

&nbsp;

&nbsp;               # Statistics

&nbsp;               total\_loss += loss.item()

&nbsp;               \_, predicted = torch.max(logits.data, 1)

&nbsp;               total += labels.size(0)

&nbsp;               correct += (predicted == labels).sum().item()

&nbsp;

&nbsp;           # Print epoch statistics

&nbsp;           avg\_loss = total\_loss / len(self.train\_loader)

&nbsp;           accuracy = 100 \* correct / total

&nbsp;           print(f"Epoch {epoch+1}/{epochs}, Loss: {avg\_loss:.4f}, Accuracy: {accuracy:.2f}%")

&nbsp;

&nbsp;       print("Histopathology module training completed!")

&nbsp;

&nbsp;   def train\_clinical\_text\_module(self, epochs: int = 10, lr: float = 1e-5):

&nbsp;       """Train clinical text module with supervised learning"""

&nbsp;       print("Training clinical text module...")

&nbsp;

&nbsp;       # Optimizer

&nbsp;       optimizer = torch.optim.Adam(

&nbsp;           self.clinical\_text\_module.parameters(),

&nbsp;           lr=lr,

&nbsp;           weight\_decay=1e-4

&nbsp;       )

&nbsp;

&nbsp;       # Loss function

&nbsp;       criterion = nn.CrossEntropyLoss().to(device) # Move criterion to device

&nbsp;

&nbsp;       # Training loop

&nbsp;       for epoch in range(epochs):

&nbsp;           self.clinical\_text\_module.train()

&nbsp;           total\_loss = 0.0

&nbsp;           correct = 0

&nbsp;           total = 0

&nbsp;

&nbsp;           for batch in tqdm(self.train\_loader, desc=f"Epoch {epoch+1}/{epochs}"):

&nbsp;               # Skip non-text data

&nbsp;               if batch\["modality"]\[0] != "clinical\_text":

&nbsp;                   continue

&nbsp;

&nbsp;               # Get texts and labels

&nbsp;               texts = batch\["text"]

&nbsp;               labels = batch\["label"].to(device)

&nbsp;

&nbsp;               # Forward pass

&nbsp;               features = self.clinical\_text\_module(texts)

&nbsp;

&nbsp;               # Simple linear classifier for training

&nbsp;               logits = nn.Linear(512, 2).to(device)(features) # Move classifier to device

&nbsp;

&nbsp;               # Compute loss

&nbsp;               loss = criterion(logits, labels)

&nbsp;

&nbsp;               # Backward pass

&nbsp;               optimizer.zero\_grad()

&nbsp;               loss.backward()

&nbsp;               optimizer.step()

&nbsp;

&nbsp;               # Statistics

&nbsp;               total\_loss += loss.item()

&nbsp;               \_, predicted = torch.max(logits.data, 1)

&nbsp;               total += labels.size(0)

&nbsp;               correct += (predicted == labels).sum().item()

&nbsp;

&nbsp;           # Print epoch statistics

&nbsp;           avg\_loss = total\_loss / len(self.train\_loader)

&nbsp;           accuracy = 100 \* correct / total

&nbsp;           print(f"Epoch {epoch+1}/{epochs}, Loss: {avg\_loss:.4f}, Accuracy: {accuracy:.2f}%")

&nbsp;

&nbsp;       print("Clinical text module training completed!")

&nbsp;

&nbsp;   def extract\_features(self, save\_path: str = None):

&nbsp;       """Extract features from all modalities"""

&nbsp;       print("Extracting features from all modalities...")

&nbsp;

&nbsp;       # Set models to evaluation mode

&nbsp;       self.mammogram\_module.eval()

&nbsp;       self.histopathology\_module.eval()

&nbsp;       self.clinical\_text\_module.eval()

&nbsp;

&nbsp;       features = {

&nbsp;           "mammogram": \[],

&nbsp;           "histopathology": \[],

&nbsp;           "clinical\_text": \[]

&nbsp;       }

&nbsp;

&nbsp;       metadata = \[]

&nbsp;

&nbsp;       with torch.no\_grad():

&nbsp;           for batch in tqdm(self.val\_loader, desc="Extracting features"):

&nbsp;               modality = batch\["modality"]\[0]

&nbsp;

&nbsp;               if modality == "mammogram":

&nbsp;                   # Get images

&nbsp;                   images = batch\["image"].to(device)

&nbsp;

&nbsp;                   # Extract features

&nbsp;                   feat, \_ = self.mammogram\_module(images)

&nbsp;

&nbsp;                   # Save features

&nbsp;                   features\["mammogram"].append(feat.cpu().numpy())

&nbsp;

&nbsp;                   # Save metadata

&nbsp;                   for i in range(len(batch\["patient\_id"])):

&nbsp;                       metadata.append({

&nbsp;                           "modality": "mammogram",

&nbsp;                           "patient\_id": batch\["patient\_id"]\[i],

&nbsp;                           "study\_id": batch\["study\_id"]\[i],

&nbsp;                           "label": batch\["label"]\[i].item(),

&nbsp;                           "feature\_idx": len(features\["mammogram"]) - 1

&nbsp;                       })

&nbsp;

&nbsp;               elif modality == "histopathology":

&nbsp;                   # Get images

&nbsp;                   images = batch\["image"].to(device)

&nbsp;

&nbsp;                   # Create patches

&nbsp;                   batch\_size = images.shape\[0]

&nbsp;                   num\_patches = 16

&nbsp;                   patches = \[]

&nbsp;

&nbsp;                   for i in range(batch\_size):

&nbsp;                       img = images\[i]

&nbsp;                       # Extract random patches

&nbsp;                       for \_ in range(num\_patches):

&nbsp;                           h, w = img.shape\[1], img.shape\[2]

&nbsp;                           th, tw = 224, 224  # Patch size

&nbsp;

&nbsp;                           i1 = torch.randint(0, h - th + 1, size=(1,)).item()

&nbsp;                           j1 = torch.randint(0, w - tw + 1, size=(1,)).item()

&nbsp;

&nbsp;                           patch = img\[:, i1:i1+th, j1:j1+tw]

&nbsp;                           patches.append(patch)

&nbsp;

&nbsp;                   patches = torch.stack(patches)

&nbsp;                   patches = patches.view(batch\_size, num\_patches, 3, 224, 224)

&nbsp;

&nbsp;                   # Extract features

&nbsp;                   feat, \_ = self.histopathology\_module(patches)

&nbsp;

&nbsp;                   # Save features

&nbsp;                   features\["histopathology"].append(feat.cpu().numpy())

&nbsp;

&nbsp;                   # Save metadata

&nbsp;                   for i in range(len(batch\["patient\_id"])):

&nbsp;                       metadata.append({

&nbsp;                           "modality": "histopathology",

&nbsp;                           "patient\_id": batch\["patient\_id"]\[i],

&nbsp;                           "study\_id": batch\["study\_id"]\[i],

&nbsp;                           "label": batch\["label"]\[i].item(),

&nbsp;                           "feature\_idx": len(features\["histopathology"]) - 1

&nbsp;                       })

&nbsp;

&nbsp;               elif modality == "clinical\_text":

&nbsp;                   # Get texts

&nbsp;                   texts = batch\["text"]

&nbsp;

&nbsp;                   # Extract features

&nbsp;                   feat = self.clinical\_text\_module(texts)

&nbsp;

&nbsp;                   # Save features

&nbsp;                   features\["clinical\_text"].append(feat.cpu().numpy())

&nbsp;

&nbsp;                   # Save metadata

&nbsp;                   for i in range(len(batch\["patient\_id"])):

&nbsp;                       metadata.append({

&nbsp;                           "modality": "clinical\_text",

&nbsp;                           "patient\_id": batch\["patient\_id"]\[i],

&nbsp;                           "study\_id": batch\["study\_id"]\[i],

&nbsp;                           "label": batch\["label"]\[i].item(),

&nbsp;                           "feature\_idx": len(features\["clinical\_text"]) - 1

&nbsp;                       })

&nbsp;

&nbsp;       # Concatenate features

&nbsp;       for modality in features:

&nbsp;           if features\[modality]:

&nbsp;               features\[modality] = np.concatenate(features\[modality], axis=0)

&nbsp;

&nbsp;       # Save features if path is provided

&nbsp;       if save\_path:

&nbsp;           os.makedirs(save\_path, exist\_ok=True)

&nbsp;

&nbsp;           # Save features

&nbsp;           for modality, feat in features.items():

&nbsp;               if feat.size > 0:

&nbsp;                   np.save(os.path.join(save\_path, f"{modality}\_features.npy"), feat)

&nbsp;

&nbsp;           # Save metadata

&nbsp;           with open(os.path.join(save\_path, "metadata.json"), 'w') as f:

&nbsp;               json.dump(metadata, f, indent=2)

&nbsp;

&nbsp;           print(f"Features saved to {save\_path}")

&nbsp;

&nbsp;       return features, metadata



\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Configuration

&nbsp;   config = {

&nbsp;       "base\_encoder": "resnet50",

&nbsp;       "projection\_dim": 128,

&nbsp;       "temperature": 0.07,

&nbsp;       "momentum": 0.999,

&nbsp;       "pretrained": True,

&nbsp;       "feature\_dim": 512,

&nbsp;       "img\_size": 224,

&nbsp;       "patch\_size": 16,

&nbsp;       "embed\_dim": 768,

&nbsp;       "depth": 12,

&nbsp;       "num\_heads": 12,

&nbsp;       "text\_model": "dmis-lab/biobert-v1.1",

&nbsp;       "batch\_size": 8,

&nbsp;       "num\_workers": 4

&nbsp;   }

&nbsp;

&nbsp;   # Initialize multi-modal processor

&nbsp;   processor = MultiModalProcessor(

&nbsp;       data\_dir="output",

&nbsp;       config=config

&nbsp;   )

&nbsp;

&nbsp;   # Pre-train mammogram module with contrastive learning

&nbsp;   processor.pretrain\_contrastive(epochs=5, lr=1e-4)

&nbsp;

&nbsp;   # Train mammogram module

&nbsp;   processor.train\_mammogram\_module(epochs=5, lr=1e-4)

&nbsp;

&nbsp;   # Train histopathology module

&nbsp;   processor.train\_histopathology\_module(epochs=5, lr=1e-4)

&nbsp;

&nbsp;   # Train clinical text module

&nbsp;   processor.train\_clinical\_text\_module(epochs=5, lr=1e-5)

&nbsp;

&nbsp;   # Extract features

&nbsp;   features, metadata = processor.extract\_features(save\_path="output/features")

&nbsp;

&nbsp;   print("Multi-modal input processing completed!")





\# Set device

device = torch.device('cuda' if torch.cuda.is\_available() else 'cpu')



class KnowledgeGraphEncoder(nn.Module):

&nbsp;   """Encodes the knowledge graph into embeddings"""

&nbsp;   def \_\_init\_\_(self, kg: Graph, embedding\_dim: int = 128, hidden\_dim: int = 256):

&nbsp;       super(KnowledgeGraphEncoder, self).\_\_init\_\_()

&nbsp;       self.kg = kg

&nbsp;       self.embedding\_dim = embedding\_dim

&nbsp;       self.hidden\_dim = hidden\_dim

&nbsp;

&nbsp;       # Extract entities and relations from the knowledge graph

&nbsp;       self.entities, self.entity\_to\_idx = self.\_extract\_entities()

&nbsp;       self.relations, self.relation\_to\_idx = self.\_extract\_relations()

&nbsp;

&nbsp;       # Create edge index and edge type for PyG

&nbsp;       self.edge\_index, self.edge\_type = self.\_create\_pyg\_graph()

&nbsp;

&nbsp;       # Initialize entity embeddings

&nbsp;       self.entity\_embeddings = nn.Embedding(len(self.entities), embedding\_dim)

&nbsp;       nn.init.xavier\_uniform\_(self.entity\_embeddings.weight)

&nbsp;

&nbsp;       # R-GCN layers

&nbsp;       self.conv1 = RGCNConv(embedding\_dim, hidden\_dim, num\_relations=len(self.relations))

&nbsp;       self.conv2 = RGCNConv(hidden\_dim, embedding\_dim, num\_relations=len(self.relations))

&nbsp;

&nbsp;       # Relation embeddings

&nbsp;       self.relation\_embeddings = nn.Embedding(len(self.relations), embedding\_dim)

&nbsp;       nn.init.xavier\_uniform\_(self.relation\_embeddings.weight)

&nbsp;

&nbsp;   def \_extract\_entities(self) -> Tuple\[List\[str], Dict\[str, int]]:

&nbsp;       """Extract unique entities from the knowledge graph"""

&nbsp;       entities = set()

&nbsp;       for s, p, o in self.kg:

&nbsp;           entities.add(str(s))

&nbsp;           if isinstance(o, URIRef):

&nbsp;               entities.add(str(o))

&nbsp;

&nbsp;       entities = sorted(entities)

&nbsp;       entity\_to\_idx = {entity: idx for idx, entity in enumerate(entities)}

&nbsp;

&nbsp;       return entities, entity\_to\_idx

&nbsp;

&nbsp;   def \_extract\_relations(self) -> Tuple\[List\[str], Dict\[str, int]]:

&nbsp;       """Extract unique relations from the knowledge graph"""

&nbsp;       relations = set()

&nbsp;       for s, p, o in self.kg:

&nbsp;           relations.add(str(p))

&nbsp;

&nbsp;       relations = sorted(relations)

&nbsp;       relation\_to\_idx = {relation: idx for idx, relation in enumerate(relations)}

&nbsp;

&nbsp;       return relations, relation\_to\_idx

&nbsp;

&nbsp;   def \_create\_pyg\_graph(self) -> Tuple\[torch.Tensor, torch.Tensor]:

&nbsp;       """Create PyG graph from RDF graph"""

&nbsp;       edge\_index = \[]

&nbsp;       edge\_type = \[]

&nbsp;

&nbsp;       for s, p, o in self.kg:

&nbsp;           if isinstance(o, URIRef):

&nbsp;               s\_idx = self.entity\_to\_idx\[str(s)]

&nbsp;               o\_idx = self.entity\_to\_idx\[str(o)]

&nbsp;               p\_idx = self.relation\_to\_idx\[str(p)]

&nbsp;

&nbsp;               edge\_index.append(\[s\_idx, o\_idx])

&nbsp;               edge\_type.append(p\_idx)

&nbsp;

&nbsp;       edge\_index = torch.tensor(edge\_index, dtype=torch.long).t().contiguous()

&nbsp;       edge\_type = torch.tensor(edge\_type, dtype=torch.long)

&nbsp;

&nbsp;       return edge\_index, edge\_type

&nbsp;

&nbsp;   def forward(self, entities: List\[str]) -> torch.Tensor:

&nbsp;       """Forward pass to get entity embeddings"""

&nbsp;       # Get entity indices

&nbsp;       entity\_indices = \[self.entity\_to\_idx.get(entity, 0) for entity in entities]

&nbsp;       entity\_indices = torch.tensor(entity\_indices, dtype=torch.long).to(device)

&nbsp;

&nbsp;       # Get initial entity embeddings

&nbsp;       x = self.entity\_embeddings(entity\_indices)

&nbsp;

&nbsp;       # Apply R-GCN layers

&nbsp;       x = F.relu(self.conv1(x, self.edge\_index, self.edge\_type))

&nbsp;       x = self.conv2(x, self.edge\_index, self.edge\_type)

&nbsp;

&nbsp;       return x

&nbsp;

&nbsp;   def get\_relation\_embeddings(self, relations: List\[str]) -> torch.Tensor:

&nbsp;       """Get relation embeddings"""

&nbsp;       relation\_indices = \[self.relation\_to\_idx.get(relation, 0) for relation in relations]

&nbsp;       relation\_indices = torch.tensor(relation\_indices, dtype=torch.long).to(device)

&nbsp;

&nbsp;       return self.relation\_embeddings(relation\_indices)



class CrossModalAttention(nn.Module):

&nbsp;   """Cross-modal attention mechanism for fusing multi-modal features"""

&nbsp;   def \_\_init\_\_(self, feature\_dim: int, num\_heads: int = 8, dropout: float = 0.1):

&nbsp;       super(CrossModalAttention, self).\_\_init\_\_()

&nbsp;       self.feature\_dim = feature\_dim

&nbsp;       self.num\_heads = num\_heads

&nbsp;       self.head\_dim = feature\_dim // num\_heads

&nbsp;

&nbsp;       assert self.head\_dim \* num\_heads == feature\_dim, "feature\_dim must be divisible by num\_heads"

&nbsp;

&nbsp;       # Query, Key, Value projections for each modality

&nbsp;       self.mammo\_q = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.mammo\_k = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.mammo\_v = nn.Linear(feature\_dim, feature\_dim)

&nbsp;

&nbsp;       self.histo\_q = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.histo\_k = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.histo\_v = nn.Linear(feature\_dim, feature\_dim)

&nbsp;

&nbsp;       self.text\_q = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.text\_k = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.text\_v = nn.Linear(feature\_dim, feature\_dim)

&nbsp;

&nbsp;       # Output projection

&nbsp;       self.output\_proj = nn.Linear(feature\_dim \* 3, feature\_dim)

&nbsp;

&nbsp;       # Dropout

&nbsp;       self.dropout = nn.Dropout(dropout)

&nbsp;

&nbsp;       # Layer normalization

&nbsp;       self.layer\_norm = nn.LayerNorm(feature\_dim)

&nbsp;

&nbsp;   def scaled\_dot\_product\_attention(self, q, k, v, mask=None):

&nbsp;       """Compute scaled dot-product attention"""

&nbsp;       attn\_scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.head\_dim)

&nbsp;

&nbsp;       if mask is not None:

&nbsp;           attn\_scores = attn\_scores.masked\_fill(mask == 0, -1e9)

&nbsp;

&nbsp;       attn\_probs = F.softmax(attn\_scores, dim=-1)

&nbsp;       attn\_probs = self.dropout(attn\_probs)

&nbsp;

&nbsp;       output = torch.matmul(attn\_probs, v)

&nbsp;       return output, attn\_probs

&nbsp;

&nbsp;   def split\_heads(self, x):

&nbsp;       """Split the last dimension into (num\_heads, head\_dim)"""

&nbsp;       batch\_size = x.size(0)

&nbsp;       return x.view(batch\_size, -1, self.num\_heads, self.head\_dim).transpose(1, 2)

&nbsp;

&nbsp;   def combine\_heads(self, x):

&nbsp;       """Combine the attention heads back to original shape"""

&nbsp;       batch\_size = x.size(0)

&nbsp;       return x.transpose(1, 2).contiguous().view(batch\_size, -1, self.feature\_dim)

&nbsp;

&nbsp;   def forward(self, mammo\_features, histo\_features, text\_features, kg\_embeddings=None):

&nbsp;       """Forward pass for cross-modal attention"""

&nbsp;       batch\_size = mammo\_features.size(0)

&nbsp;

&nbsp;       # Project to query, key, value

&nbsp;       mammo\_q = self.mammo\_q(mammo\_features)

&nbsp;       mammo\_k = self.mammo\_k(mammo\_features)

&nbsp;       mammo\_v = self.mammo\_v(mammo\_features)

&nbsp;

&nbsp;       histo\_q = self.histo\_q(histo\_features)

&nbsp;       histo\_k = self.histo\_k(histo\_features)

&nbsp;       histo\_v = self.histo\_v(histo\_features)

&nbsp;

&nbsp;       text\_q = self.text\_q(text\_features)

&nbsp;       text\_k = self.text\_k(text\_features)

&nbsp;       text\_v = self.text\_v(text\_features)

&nbsp;

&nbsp;       # Add knowledge graph embeddings if available

&nbsp;       if kg\_embeddings is not None:

&nbsp;           # Reshape kg\_embeddings to match batch size

&nbsp;           kg\_expanded = kg\_embeddings.unsqueeze(0).expand(batch\_size, -1, -1)

&nbsp;

&nbsp;           # Concatenate with original features

&nbsp;           mammo\_q = torch.cat(\[mammo\_q, kg\_expanded], dim=1)

&nbsp;           mammo\_k = torch.cat(\[mammo\_k, kg\_expanded], dim=1)

&nbsp;           mammo\_v = torch.cat(\[mammo\_v, kg\_expanded], dim=1)

&nbsp;

&nbsp;           histo\_q = torch.cat(\[histo\_q, kg\_expanded], dim=1)

&nbsp;           histo\_k = torch.cat(\[histo\_k, kg\_expanded], dim=1)

&nbsp;           histo\_v = torch.cat(\[histo\_v, kg\_expanded], dim=1)

&nbsp;

&nbsp;           text\_q = torch.cat(\[text\_q, kg\_expanded], dim=1)

&nbsp;           text\_k = torch.cat(\[text\_k, kg\_expanded], dim=1)

&nbsp;           text\_v = torch.cat(\[text\_v, kg\_expanded], dim=1)

&nbsp;

&nbsp;       # Split into heads

&nbsp;       mammo\_q = self.split\_heads(mammo\_q)

&nbsp;       mammo\_k = self.split\_heads(mammo\_k)

&nbsp;       mammo\_v = self.split\_heads(mammo\_v)

&nbsp;

&nbsp;       histo\_q = self.split\_heads(histo\_q)

&nbsp;       histo\_k = self.split\_heads(histo\_k)

&nbsp;       histo\_v = self.split\_heads(histo\_v)

&nbsp;

&nbsp;       text\_q = self.split\_heads(text\_q)

&nbsp;       text\_k = self.split\_heads(text\_k)

&nbsp;       text\_v = self.split\_heads(text\_v)

&nbsp;

&nbsp;       # Cross-modal attention: Mammogram attends to Histopathology and Text

&nbsp;       mammo\_histo\_attn, \_ = self.scaled\_dot\_product\_attention(mammo\_q, histo\_k, histo\_v)

&nbsp;       mammo\_text\_attn, \_ = self.scaled\_dot\_product\_attention(mammo\_q, text\_k, text\_v)

&nbsp;

&nbsp;       # Cross-modal attention: Histopathology attends to Mammogram and Text

&nbsp;       histo\_mammo\_attn, \_ = self.scaled\_dot\_product\_attention(histo\_q, mammo\_k, mammo\_v)

&nbsp;       histo\_text\_attn, \_ = self.scaled\_dot\_product\_attention(histo\_q, text\_k, text\_v)

&nbsp;

&nbsp;       # Cross-modal attention: Text attends to Mammogram and Histopathology

&nbsp;       text\_mammo\_attn, \_ = self.scaled\_dot\_product\_attention(text\_q, mammo\_k, mammo\_v)

&nbsp;       text\_histo\_attn, \_ = self.scaled\_dot\_product\_attention(text\_q, histo\_k, histo\_v)

&nbsp;

&nbsp;       # Combine attention outputs

&nbsp;       mammo\_combined = torch.cat(\[mammo\_histo\_attn, mammo\_text\_attn], dim=-1)

&nbsp;       histo\_combined = torch.cat(\[histo\_mammo\_attn, histo\_text\_attn], dim=-1)

&nbsp;       text\_combined = torch.cat(\[text\_mammo\_attn, text\_histo\_attn], dim=-1)

&nbsp;

&nbsp;       # Combine heads

&nbsp;       mammo\_combined = self.combine\_heads(mammo\_combined)

&nbsp;       histo\_combined = self.combine\_heads(histo\_combined)

&nbsp;       text\_combined = self.combine\_heads(text\_combined)

&nbsp;

&nbsp;       # Concatenate all modalities

&nbsp;       fused\_features = torch.cat(\[mammo\_combined, histo\_combined, text\_combined], dim=-1)

&nbsp;

&nbsp;       # Project to output dimension

&nbsp;       output = self.output\_proj(fused\_features)

&nbsp;       output = self.dropout(output)

&nbsp;       output = self.layer\_norm(output)

&nbsp;

&nbsp;       return output



class EvidentialHead(nn.Module):

&nbsp;   """Evidential head for uncertainty quantification"""

&nbsp;   def \_\_init\_\_(self, feature\_dim: int, num\_classes: int = 2):

&nbsp;       super(EvidentialHead, self).\_\_init\_\_()

&nbsp;       self.feature\_dim = feature\_dim

&nbsp;       self.num\_classes = num\_classes

&nbsp;

&nbsp;       # Evidential output layers

&nbsp;       self.alpha = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, feature\_dim // 2),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Linear(feature\_dim // 2, num\_classes),

&nbsp;           nn.Softplus()

&nbsp;       )

&nbsp;

&nbsp;       self.beta = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, feature\_dim // 2),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Linear(feature\_dim // 2, num\_classes),

&nbsp;           nn.Softplus()

&nbsp;       )

&nbsp;

&nbsp;       self.gamma = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, feature\_dim // 2),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Linear(feature\_dim // 2, num\_classes)

&nbsp;       )

&nbsp;

&nbsp;       self.nu = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, feature\_dim // 2),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Linear(feature\_dim // 2, num\_classes),

&nbsp;           nn.Softplus()

&nbsp;       )

&nbsp;

&nbsp;   def forward(self, x):

&nbsp;       """Forward pass to get evidential parameters"""

&nbsp;       alpha = self.alpha(x) + 1.0  # Ensure alpha > 1

&nbsp;       beta = self.beta(x) + 1.0    # Ensure beta > 1

&nbsp;       gamma = self.gamma(x)

&nbsp;       nu = self.nu(x) + 1.0        # Ensure nu > 1

&nbsp;

&nbsp;       return alpha, beta, gamma, nu

&nbsp;

&nbsp;   def get\_uncertainty(self, alpha, beta, gamma, nu):

&nbsp;       """Calculate uncertainty measures from evidential parameters"""

&nbsp;       # Expected probability

&nbsp;       expected\_prob = alpha / (alpha + beta)

&nbsp;

&nbsp;       # Total uncertainty (epistemic)

&nbsp;       total\_uncertainty = self.num\_classes / torch.sum(alpha + beta, dim=1, keepdim=True)

&nbsp;

&nbsp;       # Aleatoric uncertainty

&nbsp;       aleatoric\_uncertainty = torch.sum(

&nbsp;           (alpha \* beta) / ((alpha + beta)\*\*2 \* (alpha + beta + 1)),

&nbsp;           dim=1, keepdim=True

&nbsp;       )

&nbsp;

&nbsp;       # Expected value

&nbsp;       expected\_value = gamma

&nbsp;

&nbsp;       return expected\_prob, total\_uncertainty, aleatoric\_uncertainty, expected\_value



class NeuroSymbolicFusion(nn.Module):

&nbsp;   """Neuro-Symbolic Fusion Component"""

&nbsp;   def \_\_init\_\_(self, kg: Graph, feature\_dim: int = 512, num\_classes: int = 2,

&nbsp;                num\_heads: int = 8, dropout: float = 0.1):

&nbsp;       super(NeuroSymbolicFusion, self).\_\_init\_\_()

&nbsp;       self.feature\_dim = feature\_dim

&nbsp;       self.num\_classes = num\_classes

&nbsp;

&nbsp;       # Knowledge graph encoder

&nbsp;       self.kg\_encoder = KnowledgeGraphEncoder(kg, embedding\_dim=feature\_dim)

&nbsp;

&nbsp;       # Cross-modal attention

&nbsp;       self.cross\_modal\_attention = CrossModalAttention(

&nbsp;           feature\_dim=feature\_dim,

&nbsp;           num\_heads=num\_heads,

&nbsp;           dropout=dropout

&nbsp;       )

&nbsp;

&nbsp;       # Evidential head

&nbsp;       self.evidential\_head = EvidentialHead(feature\_dim, num\_classes)

&nbsp;

&nbsp;       # Modality-specific projections

&nbsp;       self.mammo\_proj = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.histo\_proj = nn.Linear(feature\_dim, feature\_dim)

&nbsp;       self.text\_proj = nn.Linear(feature\_dim, feature\_dim)

&nbsp;

&nbsp;       # Layer normalization

&nbsp;       self.layer\_norm = nn.LayerNorm(feature\_dim)

&nbsp;

&nbsp;       # Dropout

&nbsp;       self.dropout = nn.Dropout(dropout)

&nbsp;

&nbsp;   def forward(self, mammo\_features, histo\_features, text\_features, entities=None):

&nbsp;       """Forward pass for neuro-symbolic fusion"""

&nbsp;       batch\_size = mammo\_features.size(0)

&nbsp;

&nbsp;       # Project features

&nbsp;       mammo\_proj = self.mammo\_proj(mammo\_features)

&nbsp;       histo\_proj = self.histo\_proj(histo\_features)

&nbsp;       text\_proj = self.text\_proj(text\_features)

&nbsp;

&nbsp;       # Get knowledge graph embeddings

&nbsp;       if entities is None:

&nbsp;           # Default entities related to breast cancer

&nbsp;           entities = \["breast\_cancer", "malignancy", "benign", "mammogram", "histopathology"]

&nbsp;

&nbsp;       kg\_embeddings = self.kg\_encoder(entities)

&nbsp;

&nbsp;       # Apply cross-modal attention

&nbsp;       fused\_features = self.cross\_modal\_attention(

&nbsp;           mammo\_proj, histo\_proj, text\_proj, kg\_embeddings

&nbsp;       )

&nbsp;

&nbsp;       # Apply layer norm and dropout

&nbsp;       fused\_features = self.layer\_norm(fused\_features)

&nbsp;       fused\_features = self.dropout(fused\_features)

&nbsp;

&nbsp;       # Get evidential parameters

&nbsp;       alpha, beta, gamma, nu = self.evidential\_head(fused\_features)

&nbsp;

&nbsp;       # Calculate uncertainty measures

&nbsp;       expected\_prob, total\_uncertainty, aleatoric\_uncertainty, expected\_value = self.evidential\_head.get\_uncertainty(

&nbsp;           alpha, beta, gamma, nu

&nbsp;       )

&nbsp;

&nbsp;       return {

&nbsp;           "expected\_prob": expected\_prob,

&nbsp;           "total\_uncertainty": total\_uncertainty,

&nbsp;           "aleatoric\_uncertainty": aleatoric\_uncertainty,

&nbsp;           "expected\_value": expected\_value,

&nbsp;           "alpha": alpha,

&nbsp;           "beta": beta,

&nbsp;           "gamma": gamma,

&nbsp;           "nu": nu,

&nbsp;           "fused\_features": fused\_features

&nbsp;       }

&nbsp;

&nbsp;   def update\_knowledge\_graph(self, new\_triples: List\[Tuple]):

&nbsp;       """Update the knowledge graph with new triples"""

&nbsp;       # Add new triples to the knowledge graph

&nbsp;       for s, p, o in new\_triples:

&nbsp;           if isinstance(o, str):

&nbsp;               o = Literal(o)

&nbsp;           self.kg\_encoder.kg.add((s, p, o))

&nbsp;

&nbsp;       # Update entity and relation mappings

&nbsp;       self.kg\_encoder.entities, self.kg\_encoder.entity\_to\_idx = self.kg\_encoder.\_extract\_entities()

&nbsp;       self.kg\_encoder.relations, self.kg\_encoder.relation\_to\_idx = self.kg\_encoder.\_extract\_relations()

&nbsp;

&nbsp;       # Update PyG graph

&nbsp;       self.kg\_encoder.edge\_index, self.kg\_encoder.edge\_type = self.kg\_encoder.\_create\_pyg\_graph()

&nbsp;

&nbsp;       # Reinitialize embeddings for new entities and relations

&nbsp;       self.kg\_encoder.entity\_embeddings = nn.Embedding(

&nbsp;           len(self.kg\_encoder.entities),

&nbsp;           self.kg\_encoder.embedding\_dim

&nbsp;       )

&nbsp;       nn.init.xavier\_uniform\_(self.kg\_encoder.entity\_embeddings.weight)

&nbsp;

&nbsp;       self.kg\_encoder.relation\_embeddings = nn.Embedding(

&nbsp;           len(self.kg\_encoder.relations),

&nbsp;           self.kg\_encoder.embedding\_dim

&nbsp;       )

&nbsp;       nn.init.xavier\_uniform\_(self.kg\_encoder.relation\_embeddings.weight)

&nbsp;

&nbsp;       # Reinitialize R-GCN layers

&nbsp;       self.kg\_encoder.conv1 = RGCNConv(

&nbsp;           self.kg\_encoder.embedding\_dim,

&nbsp;           self.kg\_encoder.hidden\_dim,

&nbsp;           num\_relations=len(self.kg\_encoder.relations)

&nbsp;       )

&nbsp;       self.kg\_encoder.conv2 = RGCNConv(

&nbsp;           self.kg\_encoder.hidden\_dim,

&nbsp;           self.kg\_encoder.embedding\_dim,

&nbsp;           num\_relations=len(self.kg\_encoder.relations)

&nbsp;       )



class NeuroSymbolicFusionTrainer:

&nbsp;   """Trainer for the Neuro-Symbolic Fusion Component"""

&nbsp;   def \_\_init\_\_(self, model: NeuroSymbolicFusion, lr: float = 1e-4):

&nbsp;       self.model = model

&nbsp;       self.lr = lr

&nbsp;       self.optimizer = torch.optim.Adam(model.parameters(), lr=lr)

&nbsp;       self.device = next(model.parameters()).device

&nbsp;

&nbsp;   def evidential\_loss(self, alpha, beta, gamma, nu, target, epsilon=1e-8):

&nbsp;       """Compute evidential loss using negative log likelihood"""

&nbsp;       # Ensure parameters are within valid range

&nbsp;       alpha = torch.clamp(alpha, min=1.0 + epsilon)

&nbsp;       beta = torch.clamp(beta, min=1.0 + epsilon)

&nbsp;       nu = torch.clamp(nu, min=1.0 + epsilon)

&nbsp;

&nbsp;       # Convert target to one-hot encoding

&nbsp;       target\_one\_hot = F.one\_hot(target, num\_classes=self.model.num\_classes).float()

&nbsp;

&nbsp;       # Compute negative log likelihood

&nbsp;       S = alpha + beta

&nbsp;       A = target\_one\_hot \* (torch.digamma(alpha + nu) - torch.digamma(alpha)) + \\

&nbsp;           (1 - target\_one\_hot) \* (torch.digamma(beta + nu) - torch.digamma(beta))

&nbsp;

&nbsp;       nll = -torch.sum(A, dim=1)

&nbsp;

&nbsp;       # Compute KL divergence between Dirichlet and uniform

&nbsp;       uniform\_alpha = torch.ones\_like(alpha)

&nbsp;       kl = torch.lgamma(alpha.sum(dim=1)) - torch.lgamma(uniform\_alpha.sum(dim=1)) - \\

&nbsp;            torch.sum(torch.lgamma(alpha) - torch.lgamma(uniform\_alpha), dim=1) + \\

&nbsp;            torch.sum((alpha - uniform\_alpha) \* (torch.digamma(alpha) - torch.digamma(uniform\_alpha.sum(dim=1, keepdim=True))), dim=1)

&nbsp;

&nbsp;       # Total loss

&nbsp;       loss = nll + kl

&nbsp;

&nbsp;       return loss.mean()

&nbsp;

&nbsp;   def train\_epoch(self, dataloader):

&nbsp;       """Train for one epoch"""

&nbsp;       self.model.train()

&nbsp;       total\_loss = 0.0

&nbsp;       correct = 0

&nbsp;       total = 0

&nbsp;

&nbsp;       for batch in dataloader:

&nbsp;           # Move data to device

&nbsp;           mammo\_features = batch\["mammo\_features"].to(self.device)

&nbsp;           histo\_features = batch\["histo\_features"].to(self.device)

&nbsp;           text\_features = batch\["text\_features"].to(self.device)

&nbsp;           labels = batch\["label"].to(self.device)

&nbsp;

&nbsp;           # Forward pass

&nbsp;           outputs = self.model(mammo\_features, histo\_features, text\_features)

&nbsp;

&nbsp;           # Compute loss

&nbsp;           loss = self.evidential\_loss(

&nbsp;               outputs\["alpha"], outputs\["beta"], outputs\["gamma"], outputs\["nu"], labels

&nbsp;           )

&nbsp;

&nbsp;           # Backward pass

&nbsp;           self.optimizer.zero\_grad()

&nbsp;           loss.backward()

&nbsp;           self.optimizer.step()

&nbsp;

&nbsp;           # Statistics

&nbsp;           total\_loss += loss.item()

&nbsp;           \_, predicted = torch.max(outputs\["expected\_prob"], 1)

&nbsp;           total += labels.size(0)

&nbsp;           correct += (predicted == labels).sum().item()

&nbsp;

&nbsp;       avg\_loss = total\_loss / len(dataloader)

&nbsp;       accuracy = 100 \* correct / total

&nbsp;

&nbsp;       return avg\_loss, accuracy

&nbsp;

&nbsp;   def validate(self, dataloader):

&nbsp;       """Validate the model"""

&nbsp;       self.model.eval()

&nbsp;       total\_loss = 0.0

&nbsp;       correct = 0

&nbsp;       total = 0

&nbsp;       total\_uncertainty = 0.0

&nbsp;       aleatoric\_uncertainty = 0.0

&nbsp;

&nbsp;       with torch.no\_grad():

&nbsp;           for batch in dataloader:

&nbsp;               # Move data to device

&nbsp;               mammo\_features = batch\["mammo\_features"].to(self.device)

&nbsp;               histo\_features = batch\["histo\_features"].to(self.device)

&nbsp;               text\_features = batch\["text\_features"].to(self.device)

&nbsp;               labels = batch\["label"].to(self.device)

&nbsp;

&nbsp;               # Forward pass

&nbsp;               outputs = self.model(mammo\_features, histo\_features, text\_features)

&nbsp;

&nbsp;               # Compute loss

&nbsp;               loss = self.evidential\_loss(

&nbsp;                   outputs\["alpha"], outputs\["beta"], outputs\["gamma"], outputs\["nu"], labels

&nbsp;               )

&nbsp;

&nbsp;               # Statistics

&nbsp;               total\_loss += loss.item()

&nbsp;               \_, predicted = torch.max(outputs\["expected\_prob"], 1)

&nbsp;               total += labels.size(0)

&nbsp;               correct += (predicted == labels).sum().item()

&nbsp;

&nbsp;               # Accumulate uncertainty measures

&nbsp;               total\_uncertainty += outputs\["total\_uncertainty"].sum().item()

&nbsp;               aleatoric\_uncertainty += outputs\["aleatoric\_uncertainty"].sum().item()

&nbsp;

&nbsp;       avg\_loss = total\_loss / len(dataloader)

&nbsp;       accuracy = 100 \* correct / total

&nbsp;       avg\_total\_uncertainty = total\_uncertainty / total

&nbsp;       avg\_aleatoric\_uncertainty = aleatoric\_uncertainty / total

&nbsp;

&nbsp;       return avg\_loss, accuracy, avg\_total\_uncertainty, avg\_aleatoric\_uncertainty

&nbsp;

&nbsp;   def train(self, train\_loader, val\_loader, epochs: int = 10, save\_path: str = None):

&nbsp;       """Train the model"""

&nbsp;       print("Starting training...")

&nbsp;

&nbsp;       best\_val\_accuracy = 0.0

&nbsp;

&nbsp;       for epoch in range(epochs):

&nbsp;           # Train

&nbsp;           train\_loss, train\_accuracy = self.train\_epoch(train\_loader)

&nbsp;

&nbsp;           # Validate

&nbsp;           val\_loss, val\_accuracy, val\_total\_uncertainty, val\_aleatoric\_uncertainty = self.validate(val\_loader)

&nbsp;

&nbsp;           # Print statistics

&nbsp;           print(f"Epoch {epoch+1}/{epochs}")

&nbsp;           print(f"  Train Loss: {train\_loss:.4f}, Train Accuracy: {train\_accuracy:.2f}%")

&nbsp;           print(f"  Val Loss: {val\_loss:.4f}, Val Accuracy: {val\_accuracy:.2f}%")

&nbsp;           print(f"  Val Total Uncertainty: {val\_total\_uncertainty:.4f}, Val Aleatoric Uncertainty: {val\_aleatoric\_uncertainty:.4f}")

&nbsp;

&nbsp;           # Save best model

&nbsp;           if val\_accuracy > best\_val\_accuracy and save\_path:

&nbsp;               best\_val\_accuracy = val\_accuracy

&nbsp;               torch.save(self.model.state\_dict(), save\_path)

&nbsp;               print(f"  Model saved to {save\_path}")

&nbsp;

&nbsp;       print("Training completed!")



\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Load knowledge graph

&nbsp;   kg\_path = "output/knowledge\_graph.ttl"

&nbsp;   kg = Graph()

&nbsp;   kg.parse(kg\_path, format="turtle")

&nbsp;

&nbsp;   # Create model

&nbsp;   model = NeuroSymbolicFusion(kg, feature\_dim=512, num\_classes=2).to(device)

&nbsp;

&nbsp;   # Create trainer

&nbsp;   trainer = NeuroSymbolicFusionTrainer(model, lr=1e-4)

&nbsp;

&nbsp;   # Create dummy data for demonstration

&nbsp;   batch\_size = 8

&nbsp;   feature\_dim = 512

&nbsp;

&nbsp;   # Create dummy features

&nbsp;   mammo\_features = torch.randn(batch\_size, feature\_dim).to(device)

&nbsp;   histo\_features = torch.randn(batch\_size, feature\_dim).to(device)

&nbsp;   text\_features = torch.randn(batch\_size, feature\_dim).to(device)

&nbsp;   labels = torch.randint(0, 2, (batch\_size,)).to(device)

&nbsp;

&nbsp;   # Create dummy dataloader

&nbsp;   dummy\_data = \[{

&nbsp;       "mammo\_features": mammo\_features\[i:i+1],

&nbsp;       "histo\_features": histo\_features\[i:i+1],

&nbsp;       "text\_features": text\_features\[i:i+1],

&nbsp;       "label": labels\[i:i+1]

&nbsp;   } for i in range(batch\_size)]

&nbsp;

&nbsp;   dummy\_loader = torch.utils.data.DataLoader(dummy\_data, batch\_size=batch\_size)

&nbsp;

&nbsp;   # Train the model

&nbsp;   trainer.train(dummy\_loader, dummy\_loader, epochs=5, save\_path="output/neuro\_symbolic\_fusion.pt")

&nbsp;

&nbsp;   # Test with new data

&nbsp;   model.eval()

&nbsp;   with torch.no\_grad():

&nbsp;       outputs = model(mammo\_features, histo\_features, text\_features)

&nbsp;

&nbsp;       print("\\nPrediction Results:")

&nbsp;       print(f"Expected Probability: {outputs\['expected\_prob']}")

&nbsp;       print(f"Total Uncertainty: {outputs\['total\_uncertainty']}")

&nbsp;       print(f"Aleatoric Uncertainty: {outputs\['aleatoric\_uncertainty']}")

&nbsp;

&nbsp;       # Update knowledge graph with new triples

&nbsp;       new\_triples = \[

&nbsp;           (URIRef("http://example.org/medical/patient\_123"),

&nbsp;            URIRef("http://example.org/breast/hasFinding"),

&nbsp;            URIRef("http://example.org/medical/Mass")),

&nbsp;           (URIRef("http://example.org/medical/Mass"),

&nbsp;            URIRef("http://example.org/breast/hasLocation"),

&nbsp;            Literal("upper outer quadrant"))

&nbsp;       ]

&nbsp;

&nbsp;       model.update\_knowledge\_graph(new\_triples)

&nbsp;       print("Knowledge graph updated with new triples!")


\# Set device

device = torch.device('cuda' if torch.cuda.is\_available() else 'cpu')



class SymbolicConstraintLayer(nn.Module):

&nbsp;   """Layer that enforces symbolic constraints on generated data"""

&nbsp;   def \_\_init\_\_(self, kg: Graph, constraint\_rules: List\[Dict]):

&nbsp;       super(SymbolicConstraintLayer, self).\_\_init\_\_()

&nbsp;       self.kg = kg

&nbsp;       self.constraint\_rules = constraint\_rules

&nbsp;

&nbsp;       # Extract entities and relations from knowledge graph

&nbsp;       self.entities, self.entity\_to\_idx = self.\_extract\_entities()

&nbsp;       self.relations, self.relation\_to\_idx = self.\_extract\_relations()

&nbsp;

&nbsp;       # Create constraint matrix

&nbsp;       self.constraint\_matrix = self.\_create\_constraint\_matrix()

&nbsp;

&nbsp;   def \_extract\_entities(self) -> Tuple\[List\[str], Dict\[str, int]]:

&nbsp;       """Extract unique entities from the knowledge graph"""

&nbsp;       entities = set()

&nbsp;       for s, p, o in self.kg:

&nbsp;           entities.add(str(s))

&nbsp;           if isinstance(o, URIRef):

&nbsp;               entities.add(str(o))

&nbsp;

&nbsp;       entities = sorted(entities)

&nbsp;       entity\_to\_idx = {entity: idx for idx, entity in enumerate(entities)}

&nbsp;

&nbsp;       return entities, entity\_to\_idx

&nbsp;

&nbsp;   def \_extract\_relations(self) -> Tuple\[List\[str], Dict\[str, int]]:

&nbsp;       """Extract unique relations from the knowledge graph"""

&nbsp;       relations = set()

&nbsp;       for s, p, o in self.kg:

&nbsp;           relations.add(str(p))

&nbsp;

&nbsp;       relations = sorted(relations)

&nbsp;       relation\_to\_idx = {relation: idx for idx, relation in enumerate(relations)}

&nbsp;

&nbsp;       return relations, relation\_to\_idx

&nbsp;

&nbsp;   def \_create\_constraint\_matrix(self) -> torch.Tensor:

&nbsp;       """Create a matrix representing symbolic constraints"""

&nbsp;       num\_entities = len(self.entities)

&nbsp;       num\_relations = len(self.relations)

&nbsp;

&nbsp;       # Initialize constraint matrix

&nbsp;       constraint\_matrix = torch.zeros(num\_entities, num\_relations, num\_entities)

&nbsp;

&nbsp;       # Fill constraint matrix based on knowledge graph

&nbsp;       for s, p, o in self.kg:

&nbsp;           if isinstance(o, URIRef):

&nbsp;               s\_idx = self.entity\_to\_idx.get(str(s), 0)

&nbsp;               p\_idx = self.relation\_to\_idx.get(str(p), 0)

&nbsp;               o\_idx = self.entity\_to\_idx.get(str(o), 0)

&nbsp;

&nbsp;               if 0 <= s\_idx < num\_entities and 0 <= p\_idx < num\_relations and 0 <= o\_idx < num\_entities:

&nbsp;                   constraint\_matrix\[s\_idx, p\_idx, o\_idx] = 1.0

&nbsp;

&nbsp;       # Add constraint rules

&nbsp;       for rule in self.constraint\_rules:

&nbsp;           if rule\["type"] == "implication":

&nbsp;               subject = rule\["subject"]

&nbsp;               predicate = rule\["predicate"]

&nbsp;               object = rule\["object"]

&nbsp;

&nbsp;               s\_idx = self.entity\_to\_idx.get(subject, -1)

&nbsp;               p\_idx = self.relation\_to\_idx.get(predicate, -1)

&nbsp;               o\_idx = self.entity\_to\_idx.get(object, -1)

&nbsp;

&nbsp;               if s\_idx >= 0 and p\_idx >= 0 and o\_idx >= 0:

&nbsp;                   constraint\_matrix\[s\_idx, p\_idx, o\_idx] = 1.0

&nbsp;

&nbsp;       return constraint\_matrix.to(device)

&nbsp;

&nbsp;   def forward(self, entity\_embeddings: torch.Tensor) -> torch.Tensor:

&nbsp;       """Apply symbolic constraints to entity embeddings"""

&nbsp;       # Get entity indices

&nbsp;       batch\_size = entity\_embeddings.size(0)

&nbsp;

&nbsp;       # Apply constraint matrix

&nbsp;       # We'll use a differentiable approach to enforce constraints

&nbsp;       # This is a simplified version - in practice, you might use more sophisticated methods

&nbsp;

&nbsp;       # Compute constraint violation

&nbsp;       constraint\_violation = torch.zeros(batch\_size, device=device)

&nbsp;

&nbsp;       for i in range(batch\_size):

&nbsp;           # For each entity in the batch, check if it satisfies constraints

&nbsp;           # This is a simplified approach - in practice, you would use more sophisticated methods

&nbsp;

&nbsp;           # Get the entity embedding

&nbsp;           emb = entity\_embeddings\[i]

&nbsp;

&nbsp;           # Compute constraint score (simplified)

&nbsp;           # In practice, you would use the constraint matrix to compute a more sophisticated score

&nbsp;           constraint\_score = torch.mean(emb)

&nbsp;

&nbsp;           # Add to violation if constraint is not satisfied

&nbsp;           if constraint\_score < 0.5:  # Arbitrary threshold for demonstration

&nbsp;               constraint\_violation\[i] = 1.0

&nbsp;

&nbsp;       return constraint\_violation



class Generator(nn.Module):

&nbsp;   """Generator that creates fused representations with uncertainty"""

&nbsp;   def \_\_init\_\_(self, noise\_dim: int, feature\_dim: int, output\_dim: int,

&nbsp;                num\_classes: int = 2, hidden\_dim: int = 512):

&nbsp;       super(Generator, self).\_\_init\_\_()

&nbsp;       self.noise\_dim = noise\_dim

&nbsp;       self.feature\_dim = feature\_dim

&nbsp;       self.output\_dim = output\_dim

&nbsp;       self.num\_classes = num\_classes

&nbsp;

&nbsp;       # Noise processing layers

&nbsp;       self.noise\_fc = nn.Sequential(

&nbsp;           nn.Linear(noise\_dim, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.BatchNorm1d(hidden\_dim),

&nbsp;           nn.Linear(hidden\_dim, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.BatchNorm1d(hidden\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Feature processing layers

&nbsp;       self.feature\_fc = nn.Sequential(

&nbsp;           nn.Linear(feature\_dim, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.BatchNorm1d(hidden\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Fusion layers

&nbsp;       self.fusion = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim \* 2, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.BatchNorm1d(hidden\_dim),

&nbsp;           nn.Linear(hidden\_dim, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.BatchNorm1d(hidden\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Output layers for fused representation

&nbsp;       self.output\_fc = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim, output\_dim),

&nbsp;           nn.Tanh()

&nbsp;       )

&nbsp;

&nbsp;       # Uncertainty estimation layers

&nbsp;       self.uncertainty\_fc = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim, hidden\_dim // 2),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.Linear(hidden\_dim // 2, 2),  # Total and aleatoric uncertainty

&nbsp;           nn.Softplus()

&nbsp;       )

&nbsp;

&nbsp;       # Classification layers

&nbsp;       self.classification\_fc = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim, hidden\_dim // 2),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.Linear(hidden\_dim // 2, num\_classes),

&nbsp;           nn.Softmax(dim=1)

&nbsp;       )

&nbsp;

&nbsp;   def forward(self, noise: torch.Tensor, features: torch.Tensor) -> Dict\[str, torch.Tensor]:

&nbsp;       """Forward pass for the generator"""

&nbsp;       # Process noise

&nbsp;       noise\_output = self.noise\_fc(noise)

&nbsp;

&nbsp;       # Process features

&nbsp;       feature\_output = self.feature\_fc(features)

&nbsp;

&nbsp;       # Fuse noise and features

&nbsp;       fused = torch.cat(\[noise\_output, feature\_output], dim=1)

&nbsp;       fused\_output = self.fusion(fused)

&nbsp;

&nbsp;       # Generate fused representation

&nbsp;       fused\_representation = self.output\_fc(fused\_output)

&nbsp;

&nbsp;       # Estimate uncertainty

&nbsp;       uncertainty = self.uncertainty\_fc(fused\_output)

&nbsp;       total\_uncertainty = uncertainty\[:, 0]

&nbsp;       aleatoric\_uncertainty = uncertainty\[:, 1]

&nbsp;

&nbsp;       # Generate classification

&nbsp;       classification = self.classification\_fc(fused\_output)

&nbsp;

&nbsp;       return {

&nbsp;           "fused\_representation": fused\_representation,

&nbsp;           "total\_uncertainty": total\_uncertainty,

&nbsp;           "aleatoric\_uncertainty": aleatoric\_uncertainty,

&nbsp;           "classification": classification

&nbsp;       }



class Discriminator(nn.Module):

&nbsp;   """Discriminator that evaluates realism with symbolic constraints"""

&nbsp;   def \_\_init\_\_(self, input\_dim: int, hidden\_dim: int = 512, symbolic\_layer: Optional\[SymbolicConstraintLayer] = None):

&nbsp;       super(Discriminator, self).\_\_init\_\_()

&nbsp;       self.input\_dim = input\_dim

&nbsp;       self.symbolic\_layer = symbolic\_layer

&nbsp;

&nbsp;       # Input processing layers

&nbsp;       self.input\_fc = nn.Sequential(

&nbsp;           nn.Linear(input\_dim, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.Dropout(0.3)

&nbsp;       )

&nbsp;

&nbsp;       # Hidden layers

&nbsp;       self.hidden\_layers = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim, hidden\_dim),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.Dropout(0.3),

&nbsp;           nn.Linear(hidden\_dim, hidden\_dim // 2),

&nbsp;           nn.LeakyReLU(0.2),

&nbsp;           nn.Dropout(0.3)

&nbsp;       )

&nbsp;

&nbsp;       # Realism evaluation layers

&nbsp;       self.realism\_fc = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim // 2, 1),

&nbsp;           nn.Sigmoid()

&nbsp;       )

&nbsp;

&nbsp;       # Constraint satisfaction layers

&nbsp;       self.constraint\_fc = nn.Sequential(

&nbsp;           nn.Linear(hidden\_dim // 2, 1),

&nbsp;           nn.Sigmoid()

&nbsp;       )

&nbsp;

&nbsp;       # Combined evaluation layers

&nbsp;       self.combined\_fc = nn.Sequential(

&nbsp;           nn.Linear(2, 1),

&nbsp;           nn.Sigmoid()

&nbsp;       )

&nbsp;

&nbsp;   def forward(self, x: torch.Tensor) -> Dict\[str, torch.Tensor]:

&nbsp;       """Forward pass for the discriminator"""

&nbsp;       # Process input

&nbsp;       input\_output = self.input\_fc(x)

&nbsp;

&nbsp;       # Hidden layers

&nbsp;       hidden\_output = self.hidden\_layers(input\_output)

&nbsp;

&nbsp;       # Evaluate realism

&nbsp;       realism\_score = self.realism\_fc(hidden\_output)

&nbsp;

&nbsp;       # Evaluate constraint satisfaction

&nbsp;       if self.symbolic\_layer is not None:

&nbsp;           constraint\_violation = self.symbolic\_layer(x)

&nbsp;           constraint\_score = 1.0 - constraint\_violation.unsqueeze(1)

&nbsp;       else:

&nbsp;           constraint\_score = torch.ones(x.size(0), 1, device=device)

&nbsp;

&nbsp;       # Combine realism and constraint scores

&nbsp;       combined\_input = torch.cat(\[realism\_score, constraint\_score], dim=1)

&nbsp;       combined\_score = self.combined\_fc(combined\_input)

&nbsp;

&nbsp;       return {

&nbsp;           "realism\_score": realism\_score,

&nbsp;           "constraint\_score": constraint\_score,

&nbsp;           "combined\_score": combined\_score

&nbsp;       }



class GradientPenalty(Function):

&nbsp;   """Gradient penalty for Wasserstein GAN with gradient penalty"""

&nbsp;   @staticmethod

&nbsp;   def forward(ctx, inputs, outputs):

&nbsp;       ctx.save\_for\_backward(inputs, outputs)

&nbsp;       return outputs

&nbsp;

&nbsp;   @staticmethod

&nbsp;   def backward(ctx, grad\_output):

&nbsp;       inputs, outputs = ctx.saved\_tensors

&nbsp;       grad\_inputs = None

&nbsp;

&nbsp;       if ctx.needs\_input\_grad\[0]:

&nbsp;           grad\_inputs = grad\_output.clone()

&nbsp;           # Compute gradient penalty

&nbsp;           gradient = torch.autograd.grad(

&nbsp;               outputs=outputs,

&nbsp;               inputs=inputs,

&nbsp;               grad\_outputs=torch.ones\_like(outputs),

&nbsp;               create\_graph=True,

&nbsp;               retain\_graph=True,

&nbsp;               only\_inputs=True

&nbsp;           )\[0]

&nbsp;

&nbsp;           # Compute gradient penalty

&nbsp;           gradient\_penalty = ((gradient.norm(2, dim=1) - 1) \*\* 2).mean()

&nbsp;           grad\_inputs += gradient\_penalty

&nbsp;

&nbsp;       return grad\_inputs



class SymbolicGuidedGAN(nn.Module):

&nbsp;   """GAN Architecture with Symbolic Guidance"""

&nbsp;   def \_\_init\_\_(self, kg: Graph, noise\_dim: int = 100, feature\_dim: int = 512,

&nbsp;                output\_dim: int = 512, num\_classes: int = 2, hidden\_dim: int = 512,

&nbsp;                lambda\_gp: float = 10.0, lambda\_constraint: float = 1.0):

&nbsp;       super(SymbolicGuidedGAN, self).\_\_init\_\_()

&nbsp;       self.noise\_dim = noise\_dim

&nbsp;       self.feature\_dim = feature\_dim

&nbsp;       self.output\_dim = output\_dim

&nbsp;       self.num\_classes = num\_classes

&nbsp;       self.lambda\_gp = lambda\_gp

&nbsp;       self.lambda\_constraint = lambda\_constraint

&nbsp;

&nbsp;       # Define constraint rules

&nbsp;       self.constraint\_rules = \[

&nbsp;           {

&nbsp;               "type": "implication",

&nbsp;               "description": "Malignant tumors should have spiculated margins",

&nbsp;               "subject": "MalignantTumor",

&nbsp;               "predicate": "hasMargin",

&nbsp;               "object": "Spiculated"

&nbsp;           },

&nbsp;           {

&nbsp;               "type": "implication",

&nbsp;               "description": "High breast density increases cancer risk",

&nbsp;               "subject": "BreastDensityD",

&nbsp;               "predicate": "hasRiskLevel",

&nbsp;               "object": "High"

&nbsp;           },

&nbsp;           {

&nbsp;               "type": "implication",

&nbsp;               "description": "Microcalcifications can indicate malignancy",

&nbsp;               "subject": "Microcalcification",

&nbsp;               "predicate": "canIndicate",

&nbsp;               "object": "Malignancy"

&nbsp;           }

&nbsp;       ]

&nbsp;

&nbsp;       # Create symbolic constraint layer

&nbsp;       self.symbolic\_layer = SymbolicConstraintLayer(kg, self.constraint\_rules)

&nbsp;

&nbsp;       # Create generator and discriminator

&nbsp;       self.generator = Generator(

&nbsp;           noise\_dim=noise\_dim,

&nbsp;           feature\_dim=feature\_dim,

&nbsp;           output\_dim=output\_dim,

&nbsp;           num\_classes=num\_classes,

&nbsp;           hidden\_dim=hidden\_dim

&nbsp;       )

&nbsp;

&nbsp;       self.discriminator = Discriminator(

&nbsp;           input\_dim=output\_dim,

&nbsp;           hidden\_dim=hidden\_dim,

&nbsp;           symbolic\_layer=self.symbolic\_layer

&nbsp;       )

&nbsp;

&nbsp;       # Move models to device

&nbsp;       self.generator.to(device)

&nbsp;       self.discriminator.to(device)

&nbsp;

&nbsp;   def generate\_noise(self, batch\_size: int) -> torch.Tensor:

&nbsp;       """Generate noise for the generator"""

&nbsp;       return torch.randn(batch\_size, self.noise\_dim, device=device)

&nbsp;

&nbsp;   def train\_discriminator(self, real\_features: torch.Tensor, real\_labels: torch.Tensor,

&nbsp;                         optimizer\_D: torch.optim.Optimizer) -> Dict\[str, float]:

&nbsp;       """Train the discriminator"""

&nbsp;       batch\_size = real\_features.size(0)

&nbsp;

&nbsp;       # Reset gradients

&nbsp;       optimizer\_D.zero\_grad()

&nbsp;

&nbsp;       # Train with real samples

&nbsp;       real\_outputs = self.discriminator(real\_features)

&nbsp;       real\_loss = F.binary\_cross\_entropy(

&nbsp;           real\_outputs\["combined\_score"],

&nbsp;           torch.ones(batch\_size, 1, device=device)

&nbsp;       )

&nbsp;

&nbsp;       # Generate fake samples

&nbsp;       noise = self.generate\_noise(batch\_size)

&nbsp;       fake\_outputs = self.generator(noise, real\_features)

&nbsp;       fake\_features = fake\_outputs\["fused\_representation"].detach()

&nbsp;

&nbsp;       # Train with fake samples

&nbsp;       fake\_outputs = self.discriminator(fake\_features)

&nbsp;       fake\_loss = F.binary\_cross\_entropy(

&nbsp;           fake\_outputs\["combined\_score"],

&nbsp;           torch.zeros(batch\_size, 1, device=device)

&nbsp;       )

&nbsp;

&nbsp;       # Compute gradient penalty

&nbsp;       alpha = torch.rand(batch\_size, 1, device=device)

&nbsp;       interpolates = alpha \* real\_features + (1 - alpha) \* fake\_features

&nbsp;       interpolates.requires\_grad\_(True)

&nbsp;

&nbsp;       d\_interpolates = self.discriminator(interpolates)\["combined\_score"]

&nbsp;

&nbsp;       # Apply gradient penalty

&nbsp;       gradients = torch.autograd.grad(

&nbsp;           outputs=d\_interpolates,

&nbsp;           inputs=interpolates,

&nbsp;           grad\_outputs=torch.ones\_like(d\_interpolates),

&nbsp;           create\_graph=True,

&nbsp;           retain\_graph=True,

&nbsp;           only\_inputs=True

&nbsp;       )\[0]

&nbsp;

&nbsp;       gradients = gradients.view(batch\_size, -1)

&nbsp;       gradient\_penalty = ((gradients.norm(2, dim=1) - 1) \*\* 2).mean()

&nbsp;

&nbsp;       # Compute constraint penalty

&nbsp;       constraint\_penalty = torch.mean(1.0 - fake\_outputs\["constraint\_score"])

&nbsp;

&nbsp;       # Total discriminator loss

&nbsp;       d\_loss = real\_loss + fake\_loss + self.lambda\_gp \* gradient\_penalty + self.lambda\_constraint \* constraint\_penalty

&nbsp;

&nbsp;       # Backward pass

&nbsp;       d\_loss.backward()

&nbsp;       optimizer\_D.step()

&nbsp;

&nbsp;       return {

&nbsp;           "d\_loss": d\_loss.item(),

&nbsp;           "real\_loss": real\_loss.item(),

&nbsp;           "fake\_loss": fake\_loss.item(),

&nbsp;           "gradient\_penalty": gradient\_penalty.item(),

&nbsp;           "constraint\_penalty": constraint\_penalty.item()

&nbsp;       }

&nbsp;

&nbsp;   def train\_generator(self, real\_features: torch.Tensor, real\_labels: torch.Tensor,

&nbsp;                     optimizer\_G: torch.optim.Optimizer) -> Dict\[str, float]:

&nbsp;       """Train the generator"""

&nbsp;       batch\_size = real\_features.size(0)

&nbsp;

&nbsp;       # Reset gradients

&nbsp;       optimizer\_G.zero\_grad()

&nbsp;

&nbsp;       # Generate fake samples

&nbsp;       noise = self.generate\_noise(batch\_size)

&nbsp;       fake\_outputs = self.generator(noise, real\_features)

&nbsp;       fake\_features = fake\_outputs\["fused\_representation"]

&nbsp;

&nbsp;       # Train generator to fool discriminator

&nbsp;       d\_outputs = self.discriminator(fake\_features)

&nbsp;       g\_loss = F.binary\_cross\_entropy(

&nbsp;           d\_outputs\["combined\_score"],

&nbsp;           torch.ones(batch\_size, 1, device=device)

&nbsp;       )

&nbsp;

&nbsp;       # Add classification loss

&nbsp;       classification\_loss = F.cross\_entropy(

&nbsp;           fake\_outputs\["classification"],

&nbsp;           real\_labels

&nbsp;       )

&nbsp;

&nbsp;       # Add uncertainty regularization (encourage reasonable uncertainty)

&nbsp;       uncertainty\_reg = torch.mean(fake\_outputs\["total\_uncertainty"])

&nbsp;

&nbsp;       # Total generator loss

&nbsp;       g\_total = g\_loss + 0.1 \* classification\_loss + 0.01 \* uncertainty\_reg

&nbsp;

&nbsp;       # Backward pass

&nbsp;       g\_total.backward()

&nbsp;       optimizer\_G.step()

&nbsp;

&nbsp;       return {

&nbsp;           "g\_loss": g\_loss.item(),

&nbsp;           "classification\_loss": classification\_loss.item(),

&nbsp;           "uncertainty\_reg": uncertainty\_reg.item(),

&nbsp;           "g\_total": g\_total.item()

&nbsp;       }

&nbsp;

&nbsp;   def generate\_samples(self, features: torch.Tensor, num\_samples: int = 1) -> Dict\[str, torch.Tensor]:

&nbsp;       """Generate samples using the trained generator"""

&nbsp;       self.generator.eval()

&nbsp;

&nbsp;       with torch.no\_grad():

&nbsp;           batch\_size = features.size(0)

&nbsp;           samples = \[]

&nbsp;

&nbsp;           for \_ in range(num\_samples):

&nbsp;               noise = self.generate\_noise(batch\_size)

&nbsp;               outputs = self.generator(noise, features)

&nbsp;               samples.append(outputs)

&nbsp;

&nbsp;           # Aggregate samples

&nbsp;           aggregated = {

&nbsp;               "fused\_representation": torch.mean(\[s\["fused\_representation"] for s in samples], dim=0),

&nbsp;               "total\_uncertainty": torch.mean(\[s\["total\_uncertainty"] for s in samples], dim=0),

&nbsp;               "aleatoric\_uncertainty": torch.mean(\[s\["aleatoric\_uncertainty"] for s in samples], dim=0),

&nbsp;               "classification": torch.mean(\[s\["classification"] for s in samples], dim=0)

&nbsp;           }

&nbsp;

&nbsp;       self.generator.train()

&nbsp;       return aggregated



class SymbolicGuidedGANTrainer:

&nbsp;   """Trainer for the Symbolic Guided GAN"""

&nbsp;   def \_\_init\_\_(self, model: SymbolicGuidedGAN, lr\_G: float = 1e-4, lr\_D: float = 1e-4):

&nbsp;       self.model = model

&nbsp;       self.lr\_G = lr\_G

&nbsp;       self.lr\_D = lr\_D

&nbsp;

&nbsp;       # Create optimizers

&nbsp;       self.optimizer\_G = torch.optim.Adam(model.generator.parameters(), lr=lr\_G, betas=(0.5, 0.999))

&nbsp;       self.optimizer\_D = torch.optim.Adam(model.discriminator.parameters(), lr=lr\_D, betas=(0.5, 0.999))

&nbsp;

&nbsp;       # Create schedulers

&nbsp;       self.scheduler\_G = torch.optim.lr\_scheduler.StepLR(self.optimizer\_G, step\_size=10, gamma=0.9)

&nbsp;       self.scheduler\_D = torch.optim.lr\_scheduler.StepLR(self.optimizer\_D, step\_size=10, gamma=0.9)

&nbsp;

&nbsp;   def train\_epoch(self, dataloader):

&nbsp;       """Train for one epoch"""

&nbsp;       d\_losses = \[]

&nbsp;       g\_losses = \[]

&nbsp;

&nbsp;       for batch in dataloader:

&nbsp;           features = batch\["features"].to(device)

&nbsp;           labels = batch\["label"].to(device)

&nbsp;

&nbsp;           # Train discriminator

&nbsp;           d\_metrics = self.model.train\_discriminator(features, labels, self.optimizer\_D)

&nbsp;           d\_losses.append(d\_metrics)

&nbsp;

&nbsp;           # Train generator

&nbsp;           g\_metrics = self.model.train\_generator(features, labels, self.optimizer\_G)

&nbsp;           g\_losses.append(g\_metrics)

&nbsp;

&nbsp;       # Compute average metrics

&nbsp;       avg\_d\_metrics = {

&nbsp;           key: sum(\[d\[key] for d in d\_losses]) / len(d\_losses)

&nbsp;           for key in d\_losses\[0].keys()

&nbsp;       }

&nbsp;

&nbsp;       avg\_g\_metrics = {

&nbsp;           key: sum(\[g\[key] for g in g\_losses]) / len(g\_losses)

&nbsp;           for key in g\_losses\[0].keys()

&nbsp;       }

&nbsp;

&nbsp;       # Update learning rates

&nbsp;       self.scheduler\_G.step()

&nbsp;       self.scheduler\_D.step()

&nbsp;

&nbsp;       return avg\_d\_metrics, avg\_g\_metrics

&nbsp;

&nbsp;   def train(self, dataloader, epochs: int = 100, save\_path: str = None, save\_interval: int = 10):

&nbsp;       """Train the model"""

&nbsp;       print("Starting Symbolic Guided GAN training...")

&nbsp;

&nbsp;       best\_g\_loss = float('inf')

&nbsp;

&nbsp;       for epoch in range(epochs):

&nbsp;           # Train for one epoch

&nbsp;           d\_metrics, g\_metrics = self.train\_epoch(dataloader)

&nbsp;

&nbsp;           # Print metrics

&nbsp;           print(f"Epoch {epoch+1}/{epochs}")

&nbsp;           print(f"  Discriminator - Loss: {d\_metrics\['d\_loss']:.4f}, Real: {d\_metrics\['real\_loss']:.4f}, Fake: {d\_metrics\['fake\_loss']:.4f}")

&nbsp;           print(f"  Generator - Loss: {g\_metrics\['g\_total']:.4f}, Classification: {g\_metrics\['classification\_loss']:.4f}")

&nbsp;           print(f"  Gradient Penalty: {d\_metrics\['gradient\_penalty']:.4f}, Constraint Penalty: {d\_metrics\['constraint\_penalty']:.4f}")

&nbsp;

&nbsp;           # Save model if improved

&nbsp;           if g\_metrics\['g\_total'] < best\_g\_loss and save\_path and (epoch + 1) % save\_interval == 0:

&nbsp;               best\_g\_loss = g\_metrics\['g\_total']

&nbsp;               torch.save({

&nbsp;                   'epoch': epoch,

&nbsp;                   'generator\_state\_dict': self.model.generator.state\_dict(),

&nbsp;                   'discriminator\_state\_dict': self.model.discriminator.state\_dict(),

&nbsp;                   'optimizer\_G\_state\_dict': self.optimizer\_G.state\_dict(),

&nbsp;                   'optimizer\_D\_state\_dict': self.optimizer\_D.state\_dict(),

&nbsp;                   'g\_loss': best\_g\_loss,

&nbsp;               }, save\_path)

&nbsp;               print(f"  Model saved to {save\_path}")

&nbsp;

&nbsp;       print("Training completed!")



\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Load knowledge graph

&nbsp;   kg\_path = "output/knowledge\_graph.ttl"

&nbsp;   kg = Graph()

&nbsp;   kg.parse(kg\_path, format="turtle")

&nbsp;

&nbsp;   # Create model

&nbsp;   model = SymbolicGuidedGAN(

&nbsp;       kg=kg,

&nbsp;       noise\_dim=100,

&nbsp;       feature\_dim=512,

&nbsp;       output\_dim=512,

&nbsp;       num\_classes=2,

&nbsp;       hidden\_dim=512,

&nbsp;       lambda\_gp=10.0,

&nbsp;       lambda\_constraint=1.0

&nbsp;   )

&nbsp;

&nbsp;   # Create trainer

&nbsp;   trainer = SymbolicGuidedGANTrainer(model, lr\_G=1e-4, lr\_D=1e-4)

&nbsp;

&nbsp;   # Create dummy data for demonstration

&nbsp;   batch\_size = 16

&nbsp;   feature\_dim = 512

&nbsp;

&nbsp;   # Create dummy features

&nbsp;   features = torch.randn(batch\_size, feature\_dim).to(device)

&nbsp;   labels = torch.randint(0, 2, (batch\_size,)).to(device)

&nbsp;

&nbsp;   # Create dummy dataloader

&nbsp;   dummy\_data = \[{

&nbsp;       "features": features\[i:i+1],

&nbsp;       "label": labels\[i:i+1]

&nbsp;   } for i in range(batch\_size)]

&nbsp;

&nbsp;   dummy\_loader = torch.utils.data.DataLoader(dummy\_data, batch\_size=batch\_size)

&nbsp;

&nbsp;   # Train the model

&nbsp;   trainer.train(dummy\_loader, epochs=5, save\_path="output/symbolic\_guided\_gan.pt")

&nbsp;

&nbsp;   # Generate samples

&nbsp;   model.eval()

&nbsp;   with torch.no\_grad():

&nbsp;       # Generate samples

&nbsp;       samples = model.generate\_samples(features\[:5], num\_samples=3)

&nbsp;

&nbsp;       print("\\nGenerated Samples:")

&nbsp;       print(f"Fused Representation Shape: {samples\['fused\_representation'].shape}")

&nbsp;       print(f"Total Uncertainty: {samples\['total\_uncertainty']}")

&nbsp;       print(f"Classification: {samples\['classification']}")

\# Set device

device = torch.device('cuda' if torch.cuda.is\_available() else 'cpu')



class AttentionVisualizer:

&nbsp;   """Visualizes attention weights from cross-modal attention"""

&nbsp;   def \_\_init\_\_(self, model: nn.Module):

&nbsp;       self.model = model

&nbsp;       self.model.eval()

&nbsp;       self.attention\_weights = {}

&nbsp;       self.hooks = \[]

&nbsp;

&nbsp;       # Register hooks to capture attention weights

&nbsp;       self.\_register\_hooks()

&nbsp;

&nbsp;   def \_register\_hooks(self):

&nbsp;       """Register forward hooks to capture attention weights"""

&nbsp;       # Hook for cross-modal attention

&nbsp;       def hook\_fn(module, input, output):

&nbsp;           if hasattr(module, 'attention\_weights'):

&nbsp;               self.attention\_weights\[module.\_\_class\_\_.\_\_name\_\_] = module.attention\_weights

&nbsp;

&nbsp;       # Find cross-modal attention modules and register hooks

&nbsp;       for name, module in self.model.named\_modules():

&nbsp;           if 'CrossModalAttention' in name:

&nbsp;               module.register\_forward\_hook(hook\_fn)

&nbsp;

&nbsp;   def visualize\_mammogram\_attention(self, image: torch.Tensor, attention: torch.Tensor,

&nbsp;                                   save\_path: str = None) -> np.ndarray:

&nbsp;       """Visualize attention on mammogram image"""

&nbsp;       # Convert tensor to numpy

&nbsp;       if isinstance(image, torch.Tensor):

&nbsp;           image = image.cpu().numpy()

&nbsp;

&nbsp;       # Normalize image

&nbsp;       image = (image - image.min()) / (image.max() - image.min())

&nbsp;

&nbsp;       # Resize attention to match image size

&nbsp;       if len(attention.shape) == 3:  # (batch, heads, seq\_len)

&nbsp;           attention = attention.mean(dim=1)  # Average over heads

&nbsp;

&nbsp;       # For mammogram, we assume attention is over spatial regions

&nbsp;       # Reshape attention to spatial dimensions

&nbsp;       h, w = image.shape\[1], image.shape\[2]

&nbsp;       attention\_map = attention.reshape(1, 1, int(math.sqrt(attention.shape\[-1])), int(math.sqrt(attention.shape\[-1])))

&nbsp;       attention\_map = F.interpolate(attention\_map, size=(h, w), mode='bilinear', align\_corners=False)

&nbsp;       attention\_map = attention\_map.squeeze().cpu().numpy()

&nbsp;

&nbsp;       # Create heatmap

&nbsp;       heatmap = plt.cm.jet(attention\_map)\[:, :, :3]  # Use RGB channels

&nbsp;       heatmap = (heatmap \* 255).astype(np.uint8)

&nbsp;

&nbsp;       # Overlay heatmap on original image

&nbsp;       original\_image = (image.transpose(1, 2, 0) \* 255).astype(np.uint8)

&nbsp;       overlay = cv2.addWeighted(original\_image, 0.6, heatmap, 0.4, 0)

&nbsp;

&nbsp;       # Save or return

&nbsp;       if save\_path:

&nbsp;           plt.figure(figsize=(10, 10))

&nbsp;           plt.imshow(overlay)

&nbsp;           plt.axis('off')

&nbsp;           plt.savefig(save\_path, bbox\_inches='tight', pad\_inches=0)

&nbsp;           plt.close()

&nbsp;

&nbsp;       return overlay

&nbsp;

&nbsp;   def visualize\_histopathology\_attention(self, image: torch.Tensor, attention: torch.Tensor,

&nbsp;                                        patch\_size: int = 224, save\_path: str = None) -> np.ndarray:

&nbsp;       """Visualize attention on histopathology image"""

&nbsp;       # Convert tensor to numpy

&nbsp;       if isinstance(image, torch.Tensor):

&nbsp;           image = image.cpu().numpy()

&nbsp;

&nbsp;       # Normalize image

&nbsp;       image = (image - image.min()) / (image.max() - image.min())

&nbsp;

&nbsp;       # For histopathology, we assume attention is over patches

&nbsp;       # Create a grid of patches

&nbsp;       h, w = image.shape\[1], image.shape\[2]

&nbsp;       grid\_h = h // patch\_size

&nbsp;       grid\_w = w // patch\_size

&nbsp;

&nbsp;       # Reshape attention to grid

&nbsp;       if len(attention.shape) == 3:  # (batch, heads, seq\_len)

&nbsp;           attention = attention.mean(dim=1)  # Average over heads

&nbsp;

&nbsp;       attention\_grid = attention.reshape(grid\_h, grid\_w)

&nbsp;

&nbsp;       # Upsample attention to original image size

&nbsp;       attention\_map = F.interpolate(

&nbsp;           torch.tensor(attention\_grid).unsqueeze(0).unsqueeze(0),

&nbsp;           size=(h, w),

&nbsp;           mode='bilinear',

&nbsp;           align\_corners=False

&nbsp;       ).squeeze().numpy()

&nbsp;

&nbsp;       # Create heatmap

&nbsp;       heatmap = plt.cm.jet(attention\_map)\[:, :, :3]  # Use RGB channels

&nbsp;       heatmap = (heatmap \* 255).astype(np.uint8)

&nbsp;

&nbsp;       # Overlay heatmap on original image

&nbsp;       original\_image = (image.transpose(1, 2, 0) \* 255).astype(np.uint8)

&nbsp;       overlay = cv2.addWeighted(original\_image, 0.6, heatmap, 0.4, 0)

&nbsp;

&nbsp;       # Save or return

&nbsp;       if save\_path:

&nbsp;           plt.figure(figsize=(10, 10))

&nbsp;           plt.imshow(overlay)

&nbsp;           plt.axis('off')

&nbsp;           plt.savefig(save\_path, bbox\_inches='tight', pad\_inches=0)

&nbsp;           plt.close()

&nbsp;

&nbsp;       return overlay

&nbsp;

&nbsp;   def visualize\_text\_attention(self, text: str, attention: torch.Tensor, save\_path: str = None) -> str:

&nbsp;       """Visualize attention on text"""

&nbsp;       # Tokenize text

&nbsp;       tokens = text.split()

&nbsp;

&nbsp;       # Get attention weights

&nbsp;       if len(attention.shape) == 3:  # (batch, heads, seq\_len)

&nbsp;           attention = attention.mean(dim=1)  # Average over heads

&nbsp;

&nbsp;       attention = attention.squeeze().cpu().numpy()

&nbsp;

&nbsp;       # Normalize attention weights

&nbsp;       attention = (attention - attention.min()) / (attention.max() - attention.min() + 1e-8)

&nbsp;

&nbsp;       # Create HTML with highlighted text

&nbsp;       html = ""

&nbsp;       for token, attn in zip(tokens, attention):

&nbsp;           # Scale attention for color intensity

&nbsp;           intensity = int(attn \* 255)

&nbsp;           color = f"rgb(255, {255-intensity}, {255-intensity})"

&nbsp;           html += f'<span style="background-color: {color}; padding: 0.2em; margin: 0.1em; border-radius: 0.2em;">{token}</span> '

&nbsp;

&nbsp;       # Save or return

&nbsp;       if save\_path:

&nbsp;           with open(save\_path, 'w') as f:

&nbsp;               f.write(f"<html><body style='font-size: 16px; line-height: 1.5;'>{html}</body></html>")

&nbsp;

&nbsp;       return html

&nbsp;

&nbsp;   def visualize\_cross\_modal\_attention(self, save\_path: str = None):

&nbsp;       """Visualize cross-modal attention as a heatmap"""

&nbsp;       # Get attention weights from the model

&nbsp;       if not self.attention\_weights:

&nbsp;           print("No attention weights captured. Run model inference first.")

&nbsp;           return

&nbsp;

&nbsp;       # Create a heatmap for each cross-modal attention

&nbsp;       for module\_name, weights in self.attention\_weights.items():

&nbsp;           if weights is None:

&nbsp;               continue

&nbsp;

&nbsp;           # Average over batch and heads

&nbsp;           if len(weights.shape) == 4:  # (batch, heads, seq\_len, seq\_len)

&nbsp;               weights = weights.mean(dim=(0, 1))  # Average over batch and heads

&nbsp;

&nbsp;           # Create heatmap

&nbsp;           plt.figure(figsize=(10, 8))

&nbsp;           sns.heatmap(weights.cpu().numpy(), cmap='viridis')

&nbsp;           plt.title(f"{module\_name} Attention Heatmap")

&nbsp;           plt.xlabel("Key Position")

&nbsp;           plt.ylabel("Query Position")

&nbsp;

&nbsp;           if save\_path:

&nbsp;               plt.savefig(f"{save\_path}\_{module\_name}.png", bbox\_inches='tight')

&nbsp;               plt.close()

&nbsp;           else:

&nbsp;               plt.show()



class CounterfactualGenerator:

&nbsp;   """Generates counterfactual examples by perturbing input features"""

&nbsp;   def \_\_init\_\_(self, model: nn.Module, feature\_names: List\[str] = None):

&nbsp;       self.model = model

&nbsp;       self.model.eval()

&nbsp;       self.feature\_names = feature\_names or \[]

&nbsp;

&nbsp;   def generate\_counterfactual(self, input\_data: Dict\[str, torch.Tensor],

&nbsp;                              target\_class: int, max\_iterations: int = 100,

&nbsp;                              lr: float = 0.01, lambda\_param: float = 0.1) -> Dict\[str, torch.Tensor]:

&nbsp;       """Generate counterfactual example using gradient-based optimization"""

&nbsp;       # Create a copy of the input data

&nbsp;       cf\_data = {k: v.clone().detach().requires\_grad\_(True) for k, v in input\_data.items()}

&nbsp;

&nbsp;       # Define optimizer

&nbsp;       optimizer = torch.optim.Adam(\[cf\_data\[k] for k in cf\_data if cf\_data\[k].requires\_grad], lr=lr)

&nbsp;

&nbsp;       # Get original prediction

&nbsp;       with torch.no\_grad():

&nbsp;           original\_output = self.model(\*\*input\_data)

&nbsp;           original\_pred = original\_output.get("expected\_prob", original\_output.get("classification", None))

&nbsp;           if original\_pred is None:

&nbsp;               raise ValueError("Model output does not contain expected\_prob or classification")

&nbsp;           original\_class = original\_pred.argmax(dim=1).item()

&nbsp;

&nbsp;       # If already target class, return original

&nbsp;       if original\_class == target\_class:

&nbsp;           return input\_data

&nbsp;

&nbsp;       # Optimization loop

&nbsp;       for i in range(max\_iterations):

&nbsp;           optimizer.zero\_grad()

&nbsp;

&nbsp;           # Forward pass

&nbsp;           output = self.model(\*\*cf\_data)

&nbsp;           pred = output.get("expected\_prob", output.get("classification", None))

&nbsp;           if pred is None:

&nbsp;               raise ValueError("Model output does not contain expected\_prob or classification")

&nbsp;

&nbsp;           # Compute loss

&nbsp;           # 1. Loss for target class

&nbsp;           target\_loss = F.cross\_entropy(pred, torch.tensor(\[target\_class], device=device))

&nbsp;

&nbsp;           # 2. Loss for proximity to original input

&nbsp;           proximity\_loss = 0

&nbsp;           for k in cf\_data:

&nbsp;               if cf\_data\[k].requires\_grad:

&nbsp;                   proximity\_loss += F.mse\_loss(cf\_data\[k], input\_data\[k])

&nbsp;

&nbsp;           # 3. Loss for sparsity (minimal changes)

&nbsp;           sparsity\_loss = 0

&nbsp;           for k in cf\_data:

&nbsp;               if cf\_data\[k].requires\_grad:

&nbsp;                   sparsity\_loss += torch.mean(torch.abs(cf\_data\[k] - input\_data\[k]))

&nbsp;

&nbsp;           # Total loss

&nbsp;           total\_loss = target\_loss + lambda\_param \* proximity\_loss + 0.01 \* sparsity\_loss

&nbsp;

&nbsp;           # Backward pass

&nbsp;           total\_loss.backward()

&nbsp;           optimizer.step()

&nbsp;

&nbsp;           # Check if we've reached the target class

&nbsp;           current\_class = pred.argmax(dim=1).item()

&nbsp;           if current\_class == target\_class:

&nbsp;               break

&nbsp;

&nbsp;       # Return counterfactual data

&nbsp;       return cf\_data

&nbsp;

&nbsp;   def generate\_counterfactual\_explanations(self, input\_data: Dict\[str, torch.Tensor],

&nbsp;                                          save\_path: str = None) -> Dict\[str, Any]:

&nbsp;       """Generate counterfactual explanations for each class"""

&nbsp;       explanations = {}

&nbsp;

&nbsp;       # Get original prediction

&nbsp;       with torch.no\_grad():

&nbsp;           original\_output = self.model(\*\*input\_data)

&nbsp;           original\_pred = original\_output.get("expected\_prob", original\_output.get("classification", None))

&nbsp;           if original\_pred is None:

&nbsp;               raise ValueError("Model output does not contain expected\_prob or classification")

&nbsp;           original\_class = original\_pred.argmax(dim=1).item()

&nbsp;           original\_prob = original\_pred\[0, original\_class].item()

&nbsp;

&nbsp;       explanations\["original"] = {

&nbsp;           "class": original\_class,

&nbsp;           "probability": original\_prob

&nbsp;       }

&nbsp;

&nbsp;       # Generate counterfactual for each class

&nbsp;       num\_classes = original\_pred.shape\[1]

&nbsp;       for target\_class in range(num\_classes):

&nbsp;           if target\_class == original\_class:

&nbsp;               continue

&nbsp;

&nbsp;           # Generate counterfactual

&nbsp;           cf\_data = self.generate\_counterfactual(input\_data, target\_class)

&nbsp;

&nbsp;           # Get counterfactual prediction

&nbsp;           with torch.no\_grad():

&nbsp;               cf\_output = self.model(\*\*cf\_data)

&nbsp;               cf\_pred = cf\_output.get("expected\_prob", cf\_output.get("classification", None))

&nbsp;               if cf\_pred is None:

&nbsp;                   raise ValueError("Model output does not contain expected\_prob or classification")

&nbsp;               cf\_prob = cf\_pred\[0, target\_class].item()

&nbsp;

&nbsp;           # Calculate feature importance

&nbsp;           feature\_importance = {}

&nbsp;           for k in cf\_data:

&nbsp;               if cf\_data\[k].requires\_grad:

&nbsp;                   # Calculate relative change

&nbsp;                   change = torch.abs(cf\_data\[k] - input\_data\[k]).mean().item()

&nbsp;                   feature\_importance\[k] = change

&nbsp;

&nbsp;           explanations\[f"counterfactual\_{target\_class}"] = {

&nbsp;               "class": target\_class,

&nbsp;               "probability": cf\_prob,

&nbsp;               "feature\_importance": feature\_importance,

&nbsp;               "cf\_data": {k: v.detach().cpu().numpy() for k, v in cf\_data.items()}

&nbsp;           }

&nbsp;

&nbsp;       # Save explanations

&nbsp;       if save\_path:

&nbsp;           with open(save\_path, 'w') as f:

&nbsp;               # Convert numpy arrays to lists for JSON serialization

&nbsp;               json\_explanations = copy.deepcopy(explanations)

&nbsp;               for cf\_key in \[k for k in json\_explanations if k.startswith("counterfactual")]:

&nbsp;                   if "cf\_data" in json\_explanations\[cf\_key]:

&nbsp;                       for data\_key, data\_value in json\_explanations\[cf\_key]\["cf\_data"].items():

&nbsp;                           if isinstance(data\_value, np.ndarray):

&nbsp;                               json\_explanations\[cf\_key]\["cf\_data"]\[data\_key] = data\_value.tolist()

&nbsp;

&nbsp;               json.dump(json\_explanations, f, indent=2)

&nbsp;

&nbsp;       return explanations

&nbsp;

&nbsp;   def visualize\_counterfactual(self, original\_data: Dict\[str, torch.Tensor],

&nbsp;                              counterfactual\_data: Dict\[str, torch.Tensor],

&nbsp;                              modality: str, save\_path: str = None):

&nbsp;       """Visualize counterfactual examples"""

&nbsp;       if modality == "mammogram" or modality == "histopathology":

&nbsp;           # Get original and counterfactual images

&nbsp;           original\_img = original\_data.get("image", original\_data.get(f"{modality}\_features", None))

&nbsp;           counterfactual\_img = counterfactual\_data.get("image", counterfactual\_data.get(f"{modality}\_features", None))

&nbsp;

&nbsp;           if original\_img is None or counterfactual\_img is None:

&nbsp;               print(f"No {modality} data found for visualization")

&nbsp;               return

&nbsp;

&nbsp;           # Convert to numpy

&nbsp;           if isinstance(original\_img, torch.Tensor):

&nbsp;               original\_img = original\_img.cpu().numpy()

&nbsp;           if isinstance(counterfactual\_img, torch.Tensor):

&nbsp;               counterfactual\_img = counterfactual\_img.cpu().numpy()

&nbsp;

&nbsp;           # Normalize images

&nbsp;           original\_img = (original\_img - original\_img.min()) / (original\_img.max() - original\_img.min() + 1e-8)

&nbsp;           counterfactual\_img = (counterfactual\_img - counterfactual\_img.min()) / (counterfactual\_img.max() - counterfactual\_img.min() + 1e-8)

&nbsp;

&nbsp;           # Calculate difference

&nbsp;           diff = np.abs(original\_img - counterfactual\_img)

&nbsp;

&nbsp;           # Create visualization

&nbsp;           fig, axes = plt.subplots(1, 3, figsize=(15, 5))

&nbsp;

&nbsp;           # Original image

&nbsp;           axes\[0].imshow(original\_img.transpose(1, 2, 0))

&nbsp;           axes\[0].set\_title("Original")

&nbsp;           axes\[0].axis('off')

&nbsp;

&nbsp;           # Counterfactual image

&nbsp;           axes\[1].imshow(counterfactual\_img.transpose(1, 2, 0))

&nbsp;           axes\[1].set\_title("Counterfactual")

&nbsp;           axes\[1].axis('off')

&nbsp;

&nbsp;           # Difference

&nbsp;           axes\[2].imshow(diff.transpose(1, 2, 0), cmap='hot')

&nbsp;           axes\[2].set\_title("Difference")

&nbsp;           axes\[2].axis('off')

&nbsp;

&nbsp;           if save\_path:

&nbsp;               plt.savefig(save\_path, bbox\_inches='tight')

&nbsp;               plt.close()

&nbsp;           else:

&nbsp;               plt.show()

&nbsp;

&nbsp;       elif modality == "clinical\_text":

&nbsp;           # Get original and counterfactual text

&nbsp;           original\_text = original\_data.get("text", "")

&nbsp;           counterfactual\_text = counterfactual\_data.get("text", "")

&nbsp;

&nbsp;           if not original\_text or not counterfactual\_text:

&nbsp;               print("No text data found for visualization")

&nbsp;               return

&nbsp;

&nbsp;           # Create visualization

&nbsp;           fig, axes = plt.subplots(1, 2, figsize=(15, 5))

&nbsp;

&nbsp;           # Original text

&nbsp;           axes\[0].text(0.5, 0.5, original\_text, ha='center', va='center', wrap=True)

&nbsp;           axes\[0].set\_title("Original Text")

&nbsp;           axes\[0].axis('off')

&nbsp;

&nbsp;           # Counterfactual text

&nbsp;           axes\[1].text(0.5, 0.5, counterfactual\_text, ha='center', va='center', wrap=True)

&nbsp;           axes\[1].set\_title("Counterfactual Text")

&nbsp;           axes\[1].axis('off')

&nbsp;

&nbsp;           if save\_path:

&nbsp;               plt.savefig(save\_path, bbox\_inches='tight')

&nbsp;               plt.close()

&nbsp;           else:

&nbsp;               plt.show()



class RuleExtractor:

&nbsp;   """Extracts human-readable rules from the model's decision process"""

&nbsp;   def \_\_init\_\_(self, model: nn.Module, feature\_names: List\[str] = None):

&nbsp;       self.model = model

&nbsp;       self.model.eval()

&nbsp;       self.feature\_names = feature\_names or \[]

&nbsp;       self.rules = \[]

&nbsp;

&nbsp;   def extract\_decision\_rules(self, dataloader, max\_depth: int = 3) -> str:

&nbsp;       """Extract decision rules using a decision tree surrogate"""

&nbsp;       # Collect model predictions and features

&nbsp;       all\_features = \[]

&nbsp;       all\_predictions = \[]

&nbsp;

&nbsp;       with torch.no\_grad():

&nbsp;           for batch in dataloader:

&nbsp;               # Get features

&nbsp;               features = \[]

&nbsp;               for key in \["mammo\_features", "histo\_features", "text\_features"]:

&nbsp;                   if key in batch:

&nbsp;                       features.append(batch\[key].cpu().numpy())

&nbsp;

&nbsp;               if features:

&nbsp;                   features = np.concatenate(features, axis=1)

&nbsp;                   all\_features.append(features)

&nbsp;

&nbsp;               # Get predictions

&nbsp;               outputs = self.model(\*\*{k: v.to(device) for k, v in batch.items() if k.endswith("\_features")})

&nbsp;               pred = outputs.get("expected\_prob", outputs.get("classification", None))

&nbsp;               if pred is not None:

&nbsp;                   pred\_class = pred.argmax(dim=1).cpu().numpy()

&nbsp;                   all\_predictions.append(pred\_class)

&nbsp;

&nbsp;       # Concatenate all batches

&nbsp;       X = np.concatenate(all\_features, axis=0)

&nbsp;       y = np.concatenate(all\_predictions, axis=0)

&nbsp;

&nbsp;       # Train decision tree

&nbsp;       tree = DecisionTreeClassifier(max\_depth=max\_depth, random\_state=42)

&nbsp;       tree.fit(X, y)

&nbsp;

&nbsp;       # Extract rules

&nbsp;       feature\_names = self.feature\_names or \[f"feature\_{i}" for i in range(X.shape\[1])]

&nbsp;       rules = export\_text(tree, feature\_names=feature\_names)

&nbsp;

&nbsp;       return rules

&nbsp;

&nbsp;   def extract\_linear\_rules(self, dataloader) -> str:

&nbsp;       """Extract linear rules using a logistic regression surrogate"""

&nbsp;       # Collect model predictions and features

&nbsp;       all\_features = \[]

&nbsp;       all\_predictions = \[]

&nbsp;

&nbsp;       with torch.no\_grad():

&nbsp;           for batch in dataloader:

&nbsp;               # Get features

&nbsp;               features = \[]

&nbsp;               for key in \["mammo\_features", "histo\_features", "text\_features"]:

&nbsp;                   if key in batch:

&nbsp;                       features.append(batch\[key].cpu().numpy())

&nbsp;

&nbsp;               if features:

&nbsp;                   features = np.concatenate(features, axis=1)

&nbsp;                   all\_features.append(features)

&nbsp;

&nbsp;               # Get predictions

&nbsp;               outputs = self.model(\*\*{k: v.to(device) for k, v in batch.items() if k.endswith("\_features")})

&nbsp;               pred = outputs.get("expected\_prob", outputs.get("classification", None))

&nbsp;               if pred is not None:

&nbsp;                   pred\_class = pred.argmax(dim=1).cpu().numpy()

&nbsp;                   all\_predictions.append(pred\_class)

&nbsp;

&nbsp;       # Concatenate all batches

&nbsp;       X = np.concatenate(all\_features, axis=0)

&nbsp;       y = np.concatenate(all\_predictions, axis=0)

&nbsp;

&nbsp;       # Train logistic regression

&nbsp;       lr = LogisticRegression(max\_iter=1000, random\_state=42)

&nbsp;       lr.fit(X, y)

&nbsp;

&nbsp;       # Extract rules

&nbsp;       feature\_names = self.feature\_names or \[f"feature\_{i}" for i in range(X.shape\[1])]

&nbsp;       rules = "Linear Rules:\\n"

&nbsp;

&nbsp;       for i, class\_idx in enumerate(lr.classes\_):

&nbsp;           rules += f"\\nClass {class\_idx}:\\n"

&nbsp;           coef = lr.coef\_\[i]

&nbsp;

&nbsp;           # Sort features by absolute coefficient value

&nbsp;           sorted\_idx = np.argsort(np.abs(coef))\[::-1]

&nbsp;

&nbsp;           for idx in sorted\_idx\[:10]:  # Top 10 features

&nbsp;               if coef\[idx] > 0:

&nbsp;                   rules += f"  + {coef\[idx]:.4f} \* {feature\_names\[idx]}\\n"

&nbsp;               else:

&nbsp;                   rules += f"  - {abs(coef\[idx]):.4f} \* {feature\_names\[idx]}\\n"

&nbsp;

&nbsp;           rules += f"  Intercept: {lr.intercept\_\[i]:.4f}\\n"

&nbsp;

&nbsp;       return rules

&nbsp;

&nbsp;   def extract\_attention\_rules(self, attention\_weights: Dict\[str, torch.Tensor],

&nbsp;                             threshold: float = 0.5) -> List\[str]:

&nbsp;       """Extract rules based on attention weights"""

&nbsp;       rules = \[]

&nbsp;

&nbsp;       for module\_name, weights in attention\_weights.items():

&nbsp;           if weights is None:

&nbsp;               continue

&nbsp;

&nbsp;           # Average over batch and heads

&nbsp;           if len(weights.shape) == 4:  # (batch, heads, seq\_len, seq\_len)

&nbsp;               weights = weights.mean(dim=(0, 1))  # Average over batch and heads

&nbsp;

&nbsp;           # Get top attention weights

&nbsp;           weights\_np = weights.cpu().numpy()

&nbsp;

&nbsp;           # Find important connections

&nbsp;           important\_connections = np.where(weights\_np > threshold)

&nbsp;

&nbsp;           for query\_idx, key\_idx in zip(\*important\_connections):

&nbsp;               if self.feature\_names:

&nbsp;                   if query\_idx < len(self.feature\_names) and key\_idx < len(self.feature\_names):

&nbsp;                       rule = f"High attention from {self.feature\_names\[query\_idx]} to {self.feature\_names\[key\_idx]}"

&nbsp;                       rules.append(rule)

&nbsp;               else:

&nbsp;                   rule = f"High attention from position {query\_idx} to position {key\_idx}"

&nbsp;                   rules.append(rule)

&nbsp;

&nbsp;       return rules

&nbsp;

&nbsp;   def extract\_knowledge\_graph\_rules(self, kg: Graph, min\_confidence: float = 0.8) -> List\[str]:

&nbsp;       """Extract rules from the knowledge graph"""

&nbsp;       rules = \[]

&nbsp;

&nbsp;       # Extract entities and relations

&nbsp;       entities = set()

&nbsp;       relations = set()

&nbsp;

&nbsp;       for s, p, o in kg:

&nbsp;           entities.add(str(s))

&nbsp;           if isinstance(o, URIRef):

&nbsp;               entities.add(str(o))

&nbsp;           relations.add(str(p))

&nbsp;

&nbsp;       # Create entity and relation indices

&nbsp;       entity\_to\_idx = {entity: idx for idx, entity in enumerate(sorted(entities))}

&nbsp;       relation\_to\_idx = {relation: idx for idx, relation in enumerate(sorted(relations))}

&nbsp;

&nbsp;       # Create adjacency matrix

&nbsp;       num\_entities = len(entities)

&nbsp;       num\_relations = len(relations)

&nbsp;       adj\_matrix = np.zeros((num\_entities, num\_relations, num\_entities))

&nbsp;

&nbsp;       # Fill adjacency matrix

&nbsp;       for s, p, o in kg:

&nbsp;           if isinstance(o, URIRef):

&nbsp;               s\_idx = entity\_to\_idx.get(str(s), -1)

&nbsp;               p\_idx = relation\_to\_idx.get(str(p), -1)

&nbsp;               o\_idx = entity\_to\_idx.get(str(o), -1)

&nbsp;

&nbsp;               if s\_idx >= 0 and p\_idx >= 0 and o\_idx >= 0:

&nbsp;                   adj\_matrix\[s\_idx, p\_idx, o\_idx] = 1

&nbsp;

&nbsp;       # Extract rules using association rule mining (simplified)

&nbsp;       # In practice, you would use more sophisticated algorithms

&nbsp;

&nbsp;       # Find frequent patterns

&nbsp;       for s\_idx in range(num\_entities):

&nbsp;           for p\_idx in range(num\_relations):

&nbsp;               for o\_idx in range(num\_entities):

&nbsp;                   if adj\_matrix\[s\_idx, p\_idx, o\_idx] > 0:

&nbsp;                       # Check for implications

&nbsp;                       for p2\_idx in range(num\_relations):

&nbsp;                           for o2\_idx in range(num\_entities):

&nbsp;                               if adj\_matrix\[s\_idx, p2\_idx, o2\_idx] > 0:

&nbsp;                                   # Found a potential rule: (s, p, o) => (s, p2, o2)

&nbsp;                                   confidence = adj\_matrix\[s\_idx, p2\_idx, o2\_idx]  # Simplified confidence

&nbsp;

&nbsp;                                   if confidence >= min\_confidence:

&nbsp;                                       s\_entity = sorted(entities)\[s\_idx]

&nbsp;                                       p\_relation = sorted(relations)\[p\_idx]

&nbsp;                                       o\_entity = sorted(entities)\[o\_idx]

&nbsp;                                       p2\_relation = sorted(relations)\[p2\_idx]

&nbsp;                                       o2\_entity = sorted(entities)\[o2\_idx]

&nbsp;

&nbsp;                                       rule = f"If {s\_entity} {p\_relation} {o\_entity}, then {s\_entity} {p2\_relation} {o2\_entity} (confidence: {confidence:.2f})"

&nbsp;                                       rules.append(rule)

&nbsp;

&nbsp;       return rules

&nbsp;

&nbsp;   def extract\_all\_rules(self, dataloader, kg: Graph, attention\_weights: Dict\[str, torch.Tensor] = None,

&nbsp;                        save\_path: str = None) -> Dict\[str, Union\[str, List\[str]]]:

&nbsp;       """Extract all types of rules"""

&nbsp;       all\_rules = {}

&nbsp;

&nbsp;       # Extract decision rules

&nbsp;       print("Extracting decision rules...")

&nbsp;       decision\_rules = self.extract\_decision\_rules(dataloader)

&nbsp;       all\_rules\["decision\_rules"] = decision\_rules

&nbsp;

&nbsp;       # Extract linear rules

&nbsp;       print("Extracting linear rules...")

&nbsp;       linear\_rules = self.extract\_linear\_rules(dataloader)

&nbsp;       all\_rules\["linear\_rules"] = linear\_rules

&nbsp;

&nbsp;       # Extract attention rules

&nbsp;       if attention\_weights:

&nbsp;           print("Extracting attention rules...")

&nbsp;           attention\_rules = self.extract\_attention\_rules(attention\_weights)

&nbsp;           all\_rules\["attention\_rules"] = attention\_rules

&nbsp;

&nbsp;       # Extract knowledge graph rules

&nbsp;       print("Extracting knowledge graph rules...")

&nbsp;       kg\_rules = self.extract\_knowledge\_graph\_rules(kg)

&nbsp;       all\_rules\["kg\_rules"] = kg\_rules

&nbsp;

&nbsp;       # Save rules

&nbsp;       if save\_path:

&nbsp;           os.makedirs(os.path.dirname(save\_path), exist\_ok=True)

&nbsp;           with open(save\_path, 'w') as f:

&nbsp;               # Convert to JSON-serializable format

&nbsp;               json\_rules = {

&nbsp;                   "decision\_rules": all\_rules\["decision\_rules"],

&nbsp;                   "linear\_rules": all\_rules\["linear\_rules"],

&nbsp;                   "attention\_rules": all\_rules.get("attention\_rules", \[]),

&nbsp;                   "kg\_rules": all\_rules\["kg\_rules"]

&nbsp;               }

&nbsp;               json.dump(json\_rules, f, indent=2)

&nbsp;

&nbsp;       return all\_rules



class ExplainabilityModule:

&nbsp;   """Main explainability module that combines all explainability components"""

&nbsp;   def \_\_init\_\_(self, model: nn.Module, kg: Graph, feature\_names: List\[str] = None):

&nbsp;       self.model = model

&nbsp;       self.kg = kg

&nbsp;       self.feature\_names = feature\_names or \[]

&nbsp;

&nbsp;       # Initialize components

&nbsp;       self.attention\_visualizer = AttentionVisualizer(model)

&nbsp;       self.counterfactual\_generator = CounterfactualGenerator(model, feature\_names)

&nbsp;       self.rule\_extractor = RuleExtractor(model, feature\_names)

&nbsp;

&nbsp;       # Storage for explanations

&nbsp;       self.explanations = {}

&nbsp;

&nbsp;   def explain\_instance(self, input\_data: Dict\[str, torch.Tensor],

&nbsp;                       output\_dir: str = None) -> Dict\[str, Any]:

&nbsp;       """Generate a comprehensive explanation for a single instance"""

&nbsp;       explanation = {}

&nbsp;

&nbsp;       # Create output directory if needed

&nbsp;       if output\_dir:

&nbsp;           os.makedirs(output\_dir, exist\_ok=True)

&nbsp;

&nbsp;       # 1. Attention visualization

&nbsp;       print("Generating attention visualizations...")

&nbsp;       attention\_dir = os.path.join(output\_dir, "attention") if output\_dir else None

&nbsp;

&nbsp;       # Get model output to capture attention weights

&nbsp;       with torch.no\_grad():

&nbsp;           output = self.model(\*\*input\_data)

&nbsp;

&nbsp;       # Visualize mammogram attention

&nbsp;       if "mammo\_features" in input\_data:

&nbsp;           mammogram\_attention = output.get("mammo\_attention", torch.ones(1, 1, 16\*16))

&nbsp;           mammogram\_vis = self.attention\_visualizer.visualize\_mammogram\_attention(

&nbsp;               input\_data\["mammo\_features"],

&nbsp;               mammogram\_attention,

&nbsp;               save\_path=os.path.join(attention\_dir, "mammogram\_attention.png") if attention\_dir else None

&nbsp;           )

&nbsp;           explanation\["mammogram\_attention"] = mammogram\_vis

&nbsp;

&nbsp;       # Visualize histopathology attention

&nbsp;       if "histo\_features" in input\_data:

&nbsp;           histopathology\_attention = output.get("histo\_attention", torch.ones(1, 1, 16))

&nbsp;           histopathology\_vis = self.attention\_visualizer.visualize\_histopathology\_attention(

&nbsp;               input\_data\["histo\_features"],

&nbsp;               histopathology\_attention,

&nbsp;               save\_path=os.path.join(attention\_dir, "histopathology\_attention.png") if attention\_dir else None

&nbsp;           )

&nbsp;           explanation\["histopathology\_attention"] = histopathology\_vis

&nbsp;

&nbsp;       # Visualize text attention

&nbsp;       if "text\_features" in input\_data:

&nbsp;           text\_attention = output.get("text\_attention", torch.ones(1, 1, 50))

&nbsp;           text = input\_data.get("text", "No text provided")

&nbsp;           text\_vis = self.attention\_visualizer.visualize\_text\_attention(

&nbsp;               text,

&nbsp;               text\_attention,

&nbsp;               save\_path=os.path.join(attention\_dir, "text\_attention.html") if attention\_dir else None

&nbsp;           )

&nbsp;           explanation\["text\_attention"] = text\_vis

&nbsp;

&nbsp;       # 2. Counterfactual explanations

&nbsp;       print("Generating counterfactual explanations...")

&nbsp;       cf\_dir = os.path.join(output\_dir, "counterfactuals") if output\_dir else None

&nbsp;

&nbsp;       cf\_explanations = self.counterfactual\_generator.generate\_counterfactual\_explanations(

&nbsp;           input\_data,

&nbsp;           save\_path=os.path.join(cf\_dir, "cf\_explanations.json") if cf\_dir else None

&nbsp;       )

&nbsp;       explanation\["counterfactuals"] = cf\_explanations

&nbsp;

&nbsp;       # Visualize counterfactuals

&nbsp;       for cf\_key, cf\_data in cf\_explanations.items():

&nbsp;           if cf\_key.startswith("counterfactual\_") and "cf\_data" in cf\_data:

&nbsp;               cf\_data\_tensor = {k: torch.tensor(v) if isinstance(v, np.ndarray) else v

&nbsp;                                for k, v in cf\_data\["cf\_data"].items()}

&nbsp;

&nbsp;               # Visualize mammogram counterfactual

&nbsp;               if "mammo\_features" in cf\_data\_tensor:

&nbsp;                   self.counterfactual\_generator.visualize\_counterfactual(

&nbsp;                       input\_data,

&nbsp;                       cf\_data\_tensor,

&nbsp;                       "mammogram",

&nbsp;                       save\_path=os.path.join(cf\_dir, f"{cf\_key}\_mammogram.png") if cf\_dir else None

&nbsp;                   )

&nbsp;

&nbsp;               # Visualize histopathology counterfactual

&nbsp;               if "histo\_features" in cf\_data\_tensor:

&nbsp;                   self.counterfactual\_generator.visualize\_counterfactual(

&nbsp;                       input\_data,

&nbsp;                       cf\_data\_tensor,

&nbsp;                       "histopathology",

&nbsp;                       save\_path=os.path.join(cf\_dir, f"{cf\_key}\_histopathology.png") if cf\_dir else None

&nbsp;                   )

&nbsp;

&nbsp;               # Visualize text counterfactual

&nbsp;               if "text\_features" in cf\_data\_tensor:

&nbsp;                   self.counterfactual\_generator.visualize\_counterfactual(

&nbsp;                       input\_data,

&nbsp;                       cf\_data\_tensor,

&nbsp;                       "clinical\_text",

&nbsp;                       save\_path=os.path.join(cf\_dir, f"{cf\_key}\_text.png") if cf\_dir else None

&nbsp;                   )

&nbsp;

&nbsp;       # 3. Rule extraction (for the model in general, not specific to this instance)

&nbsp;       print("Extracting rules...")

&nbsp;       rules\_dir = os.path.join(output\_dir, "rules") if output\_dir else None

&nbsp;

&nbsp;       # Get attention weights from the model

&nbsp;       attention\_weights = self.attention\_visualizer.attention\_weights

&nbsp;

&nbsp;       # Extract all rules

&nbsp;       # Note: This requires a dataloader, which we don't have for a single instance

&nbsp;       # In practice, you would pass a dataloader from your validation set

&nbsp;       # For demonstration, we'll skip this part

&nbsp;

&nbsp;       # Save explanation

&nbsp;       if output\_dir:

&nbsp;           with open(os.path.join(output\_dir, "explanation.json"), 'w') as f:

&nbsp;               # Convert numpy arrays to lists for JSON serialization

&nbsp;               json\_explanation = copy.deepcopy(explanation)

&nbsp;               json.dump(json\_explanation, f, indent=2)

&nbsp;

&nbsp;       return explanation

&nbsp;

&nbsp;   def explain\_model(self, dataloader, output\_dir: str = None) -> Dict\[str, Any]:

&nbsp;       """Generate model-level explanations"""

&nbsp;       model\_explanation = {}

&nbsp;

&nbsp;       # Create output directory if needed

&nbsp;       if output\_dir:

&nbsp;           os.makedirs(output\_dir, exist\_ok=True)

&nbsp;

&nbsp;       # 1. Visualize cross-modal attention

&nbsp;       print("Visualizing cross-modal attention...")

&nbsp;       attention\_dir = os.path.join(output\_dir, "attention") if output\_dir else None

&nbsp;

&nbsp;       # Run model on a batch to capture attention weights

&nbsp;       batch = next(iter(dataloader))

&nbsp;       with torch.no\_grad():

&nbsp;           self.model(\*\*{k: v.to(device) for k, v in batch.items() if k.endswith("\_features")})

&nbsp;

&nbsp;       # Visualize cross-modal attention

&nbsp;       self.attention\_visualizer.visualize\_cross\_modal\_attention(

&nbsp;           save\_path=os.path.join(attention\_dir, "cross\_modal\_attention.png") if attention\_dir else None

&nbsp;       )

&nbsp;

&nbsp;       # 2. Extract rules

&nbsp;       print("Extracting rules...")

&nbsp;       rules\_dir = os.path.join(output\_dir, "rules") if output\_dir else None

&nbsp;

&nbsp;       # Get attention weights from the model

&nbsp;       attention\_weights = self.attention\_visualizer.attention\_weights

&nbsp;

&nbsp;       # Extract all rules

&nbsp;       all\_rules = self.rule\_extractor.extract\_all\_rules(

&nbsp;           dataloader,

&nbsp;           self.kg,

&nbsp;           attention\_weights,

&nbsp;           save\_path=os.path.join(rules\_dir, "all\_rules.json") if rules\_dir else None

&nbsp;       )

&nbsp;       model\_explanation\["rules"] = all\_rules

&nbsp;

&nbsp;       # Save model explanation

&nbsp;       if output\_dir:

&nbsp;           with open(os.path.join(output\_dir, "model\_explanation.json"), 'w') as f:

&nbsp;               json.dump(model\_explanation, f, indent=2)

&nbsp;

&nbsp;       return model\_explanation



\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Load knowledge graph

&nbsp;   kg\_path = "output/knowledge\_graph.ttl"

&nbsp;   kg = Graph()

&nbsp;   kg.parse(kg\_path, format="turtle")

&nbsp;

&nbsp;   # Create a dummy model for demonstration

&nbsp;   class DummyModel(nn.Module):

&nbsp;       def \_\_init\_\_(self):

&nbsp;           super(DummyModel, self).\_\_init\_\_()

&nbsp;           self.fc = nn.Linear(512, 2)

&nbsp;           self.attention\_weights = None

&nbsp;

&nbsp;       def forward(self, \*\*kwargs):

&nbsp;           # Dummy forward pass

&nbsp;           features = \[]

&nbsp;           for key in \["mammo\_features", "histo\_features", "text\_features"]:

&nbsp;               if key in kwargs:

&nbsp;                   features.append(kwargs\[key])

&nbsp;

&nbsp;           if features:

&nbsp;               x = torch.cat(features, dim=1)

&nbsp;               logits = self.fc(x)

&nbsp;               probs = F.softmax(logits, dim=1)

&nbsp;

&nbsp;               # Generate dummy attention weights

&nbsp;               batch\_size = x.size(0)

&nbsp;               self.attention\_weights = {

&nbsp;                   "CrossModalAttention": torch.randn(batch\_size, 8, 64, 64)  # (batch, heads, seq\_len, seq\_len)

&nbsp;               }

&nbsp;

&nbsp;               # Modality-specific attention

&nbsp;               if "mammo\_features" in kwargs:

&nbsp;                   self.attention\_weights\["mammo\_attention"] = torch.randn(batch\_size, 1, 256)

&nbsp;               if "histo\_features" in kwargs:

&nbsp;                   self.attention\_weights\["histo\_attention"] = torch.randn(batch\_size, 1, 16)

&nbsp;               if "text\_features" in kwargs:

&nbsp;                   self.attention\_weights\["text\_attention"] = torch.randn(batch\_size, 1, 50)

&nbsp;

&nbsp;               return {

&nbsp;                   "expected\_prob": probs,

&nbsp;                   "mammo\_attention": self.attention\_weights.get("mammo\_attention"),

&nbsp;                   "histo\_attention": self.attention\_weights.get("histo\_attention"),

&nbsp;                   "text\_attention": self.attention\_weights.get("text\_attention")

&nbsp;               }

&nbsp;

&nbsp;           return {"expected\_prob": torch.ones(1, 2) \* 0.5}

&nbsp;

&nbsp;   # Create model

&nbsp;   model = DummyModel().to(device)

&nbsp;

&nbsp;   # Create explainability module

&nbsp;   feature\_names = \[f"feature\_{i}" for i in range(512 \* 3)]  # Dummy feature names

&nbsp;   explainability = ExplainabilityModule(model, kg, feature\_names)

&nbsp;

&nbsp;   # Create dummy data for demonstration

&nbsp;   batch\_size = 1

&nbsp;   feature\_dim = 512

&nbsp;

&nbsp;   # Create dummy features

&nbsp;   input\_data = {

&nbsp;       "mammo\_features": torch.randn(batch\_size, feature\_dim).to(device),

&nbsp;       "histo\_features": torch.randn(batch\_size, feature\_dim).to(device),

&nbsp;       "text\_features": torch.randn(batch\_size, feature\_dim).to(device),

&nbsp;       "text": "Patient presents with a suspicious mass in the upper outer quadrant of the right breast."

&nbsp;   }

&nbsp;

&nbsp;   # Create dummy dataloader for model explanation

&nbsp;   dummy\_data = \[input\_data for \_ in range(10)]

&nbsp;   dummy\_loader = torch.utils.data.DataLoader(dummy\_data, batch\_size=batch\_size)

&nbsp;

&nbsp;   # Generate instance-level explanation

&nbsp;   instance\_explanation = explainability.explain\_instance(

&nbsp;       input\_data,

&nbsp;       output\_dir="output/explanation/instance"

&nbsp;   )

&nbsp;

&nbsp;   # Generate model-level explanation

&nbsp;   model\_explanation = explainability.explain\_model(

&nbsp;       dummy\_loader,

&nbsp;       output\_dir="output/explanation/model"

&nbsp;   )

&nbsp;

&nbsp;   print("Explainability analysis completed!")
\# Set device

device = torch.device('cuda' if torch.cuda.is\_available() else 'cpu')



\# Load medical NLP model

try:

&nbsp;   nlp = spacy.load("en\_core\_sci\_sm")

except:

&nbsp;   nlp = spacy.load("en\_core\_web\_sm")



class FusedRepresentationEncoder(nn.Module):

&nbsp;   """Encodes fused representations for conditioning the LLM"""

&nbsp;   def \_\_init\_\_(self, fused\_dim: int, hidden\_dim: int, output\_dim: int):

&nbsp;       super(FusedRepresentationEncoder, self).\_\_init\_\_()

&nbsp;       self.fused\_dim = fused\_dim

&nbsp;       self.hidden\_dim = hidden\_dim

&nbsp;       self.output\_dim = output\_dim

&nbsp;

&nbsp;       # Projection layers

&nbsp;       self.projection = nn.Sequential(

&nbsp;           nn.Linear(fused\_dim, hidden\_dim),

&nbsp;           nn.ReLU(),

&nbsp;           nn.Dropout(0.1),

&nbsp;           nn.Linear(hidden\_dim, output\_dim)

&nbsp;       )

&nbsp;

&nbsp;       # Attention mechanism

&nbsp;       self.attention = nn.MultiheadAttention(output\_dim, num\_heads=8, dropout=0.1)

&nbsp;

&nbsp;       # Layer normalization

&nbsp;       self.layer\_norm = nn.LayerNorm(output\_dim)

&nbsp;

&nbsp;   def forward(self, fused\_rep: torch.Tensor, uncertainty: torch.Tensor = None):

&nbsp;       """Forward pass to encode fused representation"""

&nbsp;       # Project fused representation

&nbsp;       encoded = self.projection(fused\_rep)

&nbsp;

&nbsp;       # Apply layer normalization

&nbsp;       encoded = self.layer\_norm(encoded)

&nbsp;

&nbsp;       # If uncertainty is provided, incorporate it

&nbsp;       if uncertainty is not None:

&nbsp;           # Expand uncertainty to match encoded dimensions

&nbsp;           uncertainty\_expanded = uncertainty.unsqueeze(-1).expand(-1, self.output\_dim)

&nbsp;           # Modulate encoded representation with uncertainty

&nbsp;           encoded = encoded \* (1.0 - uncertainty\_expanded)

&nbsp;

&nbsp;       return encoded



class LLMReportGenerator(nn.Module):

&nbsp;   """LLM-based report generator conditioned on fused representations"""

&nbsp;   def \_\_init\_\_(self, model\_name: str = "gpt2", fused\_dim: int = 512,

&nbsp;                output\_dim: int = 768, max\_length: int = 512):

&nbsp;       super(LLMReportGenerator, self).\_\_init\_\_()

&nbsp;       self.model\_name = model\_name

&nbsp;       self.fused\_dim = fused\_dim

&nbsp;       self.output\_dim = output\_dim

&nbsp;       self.max\_length = max\_length

&nbsp;

&nbsp;       # Load tokenizer and model

&nbsp;       if "t5" in model\_name.lower():

&nbsp;           self.tokenizer = T5Tokenizer.from\_pretrained(model\_name)

&nbsp;           self.model = T5ForConditionalGeneration.from\_pretrained(model\_name)

&nbsp;           self.model\_type = "t5"

&nbsp;       else:

&nbsp;           self.tokenizer = GPT2Tokenizer.from\_pretrained(model\_name)

&nbsp;           self.model = GPT2LMHeadModel.from\_pretrained(model\_name)

&nbsp;           self.model\_type = "gpt2"

&nbsp;

&nbsp;           # Add special tokens for conditioning

&nbsp;           if self.tokenizer.pad\_token is None:

&nbsp;               self.tokenizer.pad\_token = self.tokenizer.eos\_token

&nbsp;

&nbsp;           # Add special tokens for fused representation

&nbsp;           special\_tokens = {

&nbsp;               "fused\_start": "<|FUSED\_START|>",

&nbsp;               "fused\_end": "<|FUSED\_END|>",

&nbsp;               "uncertainty\_start": "<|UNCERTAINTY\_START|>",

&nbsp;               "uncertainty\_end": "<|UNCERTAINTY\_END|>"

&nbsp;           }

&nbsp;

&nbsp;           num\_added\_toks = self.tokenizer.add\_special\_tokens(special\_tokens)

&nbsp;           self.model.resize\_token\_embeddings(len(self.tokenizer))

&nbsp;

&nbsp;           # Store token IDs

&nbsp;           self.fused\_start\_token = self.tokenizer.convert\_tokens\_to\_ids(\["<|FUSED\_START|>"])\[0]

&nbsp;           self.fused\_end\_token = self.tokenizer.convert\_tokens\_to\_ids(\["<|FUSED\_END|>"])\[0]

&nbsp;           self.uncertainty\_start\_token = self.tokenizer.convert\_tokens\_to\_ids(\["<|UNCERTAINTY\_START|>"])\[0]

&nbsp;           self.uncertainty\_end\_token = self.tokenizer.convert\_tokens\_to\_ids(\["<|UNCERTAINTY\_END|>"])\[0]

&nbsp;

&nbsp;       # Fused representation encoder

&nbsp;       self.fused\_encoder = FusedRepresentationEncoder(fused\_dim, 1024, output\_dim)

&nbsp;

&nbsp;       # Projection for uncertainty

&nbsp;       self.uncertainty\_proj = nn.Linear(1, output\_dim)

&nbsp;

&nbsp;       # Move model to device

&nbsp;       self.model.to(device)

&nbsp;       self.fused\_encoder.to(device)

&nbsp;       self.uncertainty\_proj.to(device)

&nbsp;

&nbsp;   def \_prepare\_input\_with\_conditioning(self, fused\_rep: torch.Tensor,

&nbsp;                                      uncertainty: torch.Tensor = None,

&nbsp;                                      prompt: str = None):

&nbsp;       """Prepare input with fused representation conditioning"""

&nbsp;       batch\_size = fused\_rep.size(0)

&nbsp;

&nbsp;       # Encode fused representation

&nbsp;       fused\_encoded = self.fused\_encoder(fused\_rep, uncertainty)

&nbsp;

&nbsp;       # Convert to token embeddings

&nbsp;       fused\_tokens = fused\_encoded.unsqueeze(1)  # (batch\_size, 1, output\_dim)

&nbsp;

&nbsp;       # Encode uncertainty if provided

&nbsp;       uncertainty\_tokens = None

&nbsp;       if uncertainty is not None:

&nbsp;           uncertainty\_encoded = self.uncertainty\_proj(unsqueeze(-1))

&nbsp;           uncertainty\_tokens = uncertainty\_encoded.unsqueeze(1)  # (batch\_size, 1, output\_dim)

&nbsp;

&nbsp;       # Prepare prompt tokens

&nbsp;       prompt\_tokens = None

&nbsp;       if prompt:

&nbsp;           prompt\_tokens = self.tokenizer(prompt, return\_tensors="pt", padding=True, truncation=True)

&nbsp;           prompt\_tokens = prompt\_tokens.input\_ids.to(device)

&nbsp;           prompt\_embeddings = self.model.get\_input\_embeddings()(prompt\_tokens)

&nbsp;

&nbsp;       # Combine tokens

&nbsp;       if self.model\_type == "gpt2":

&nbsp;           # For GPT-2, we need to create a sequence of tokens

&nbsp;           # Start with special tokens

&nbsp;           input\_ids = torch.tensor(\[\[self.fused\_start\_token]] \* batch\_size).to(device)

&nbsp;

&nbsp;           # Add fused representation tokens

&nbsp;           # We'll use the fused\_encoded as embeddings for a special token

&nbsp;           # We'll create a sequence of tokens for the fused representation

&nbsp;           fused\_seq\_len = 10  # Number of tokens to represent fused representation

&nbsp;           fused\_seq = fused\_encoded.unsqueeze(1).expand(-1, fused\_seq\_len, -1)

&nbsp;

&nbsp;           # Add uncertainty tokens if provided

&nbsp;           uncertainty\_seq = None

&nbsp;           if uncertainty\_tokens is not None:

&nbsp;               uncertainty\_seq\_len = 5  # Number of tokens to represent uncertainty

&nbsp;               uncertainty\_seq = uncertainty\_tokens.unsqueeze(1).expand(-1, uncertainty\_seq\_len, -1)

&nbsp;

&nbsp;           # Combine all tokens

&nbsp;           if prompt\_tokens is not None:

&nbsp;               # Combine prompt, fused, and uncertainty tokens

&nbsp;               combined\_embeddings = torch.cat(\[

&nbsp;                   self.model.get\_input\_embeddings()(input\_ids),

&nbsp;                   prompt\_embeddings,

&nbsp;                   fused\_seq,

&nbsp;                   uncertainty\_seq if uncertainty\_seq is not None else torch.tensor(\[]).to(device)

&nbsp;               ], dim=1)

&nbsp;           else:

&nbsp;               combined\_embeddings = torch.cat(\[

&nbsp;                   self.model.get\_input\_embeddings()(input\_ids),

&nbsp;                   fused\_seq,

&nbsp;                   uncertainty\_seq if uncertainty\_seq is not None else torch.tensor(\[]).to(device)

&nbsp;               ], dim=1)

&nbsp;

&nbsp;           return combined\_embeddings

&nbsp;       else:

&nbsp;           # For T5, we'll use the fused representation as part of the input

&nbsp;           # We'll prepend it to the prompt

&nbsp;           if prompt:

&nbsp;               input\_text = f"Generate medical report based on: {prompt}"

&nbsp;           else:

&nbsp;               input\_text = "Generate medical report:"

&nbsp;

&nbsp;           # Tokenize input text

&nbsp;           input\_ids = self.tokenizer(input\_text, return\_tensors="pt", padding=True, truncation=True)

&nbsp;           input\_ids = input\_ids.input\_ids.to(device)

&nbsp;

&nbsp;           return input\_ids

&nbsp;

&nbsp;   def generate\_report(self, fused\_rep: torch.Tensor, uncertainty: torch.Tensor = None,

&nbsp;                      prompt: str = None, max\_length: int = None,

&nbsp;                      temperature: float = 1.0, top\_p: float = 0.9,

&nbsp;                      do\_sample: bool = True, num\_beams: int = 5) -> List\[str]:

&nbsp;       """Generate medical report conditioned on fused representation"""

&nbsp;       batch\_size = fused\_rep.size(0)

&nbsp;       max\_length = max\_length or self.max\_length

&nbsp;

&nbsp;       # Prepare input with conditioning

&nbsp;       if self.model\_type == "gpt2":

&nbsp;           # Get embeddings with conditioning

&nbsp;           inputs\_embeds = self.\_prepare\_input\_with\_conditioning(fused\_rep, uncertainty, prompt)

&nbsp;

&nbsp;           # Generate with conditioning

&nbsp;           with torch.no\_grad():

&nbsp;               outputs = self.model.generate(

&nbsp;                   inputs\_embeds=inputs\_embeds,

&nbsp;                   max\_length=max\_length,

&nbsp;                   temperature=temperature,

&nbsp;                   top\_p=top\_p,

&nbsp;                   do\_sample=do\_sample,

&nbsp;                   num\_beams=num\_beams,

&nbsp;                   pad\_token\_id=self.tokenizer.eos\_token\_id

&nbsp;               )

&nbsp;

&nbsp;           # Decode generated tokens

&nbsp;           generated\_texts = \[]

&nbsp;           for i in range(batch\_size):

&nbsp;               # Skip the conditioning tokens

&nbsp;               generated\_tokens = outputs\[i]\[inputs\_embeds.size(1):]

&nbsp;               generated\_text = self.tokenizer.decode(generated\_tokens, skip\_special\_tokens=True)

&nbsp;               generated\_texts.append(generated\_text)

&nbsp;       else:

&nbsp;           # For T5, prepare input text

&nbsp;           if prompt:

&nbsp;               input\_text = f"Generate medical report based on: {prompt}"

&nbsp;           else:

&nbsp;               input\_text = "Generate medical report:"

&nbsp;

&nbsp;           # Tokenize input text

&nbsp;           input\_ids = self.tokenizer(input\_text, return\_tensors="pt", padding=True, truncation=True)

&nbsp;           input\_ids = input\_ids.input\_ids.to(device)

&nbsp;

&nbsp;           # Generate report

&nbsp;           with torch.no\_grad():

&nbsp;               outputs = self.model.generate(

&nbsp;                   input\_ids=input\_ids,

&nbsp;                   max\_length=max\_length,

&nbsp;                   temperature=temperature,

&nbsp;                   top\_p=top\_p,

&nbsp;                   do\_sample=do\_sample,

&nbsp;                   num\_beams=num\_beams,

&nbsp;                   early\_stopping=True

&nbsp;               )

&nbsp;

&nbsp;           # Decode generated tokens

&nbsp;           generated\_texts = \[]

&nbsp;           for i in range(batch\_size):

&nbsp;               generated\_text = self.tokenizer.decode(outputs\[i], skip\_special\_tokens=True)

&nbsp;               generated\_texts.append(generated\_text)

&nbsp;

&nbsp;       return generated\_texts

&nbsp;

&nbsp;   def generate\_report\_with\_uncertainty(self, fused\_rep: torch.Tensor, uncertainty: torch.Tensor,

&nbsp;                                      prompt: str = None, max\_length: int = None) -> List\[Dict\[str, Any]]:

&nbsp;       """Generate medical report with uncertainty highlighting"""

&nbsp;       batch\_size = fused\_rep.size(0)

&nbsp;       max\_length = max\_length or self.max\_length

&nbsp;

&nbsp;       # Generate reports

&nbsp;       reports = self.generate\_report(fused\_rep, uncertainty, prompt, max\_length)

&nbsp;

&nbsp;       # Create report objects with uncertainty information

&nbsp;       report\_objects = \[]

&nbsp;       for i in range(batch\_size):

&nbsp;           report\_obj = {

&nbsp;               "report": reports\[i],

&nbsp;               "uncertainty": uncertainty\[i].item() if uncertainty is not None else None,

&nbsp;               "high\_uncertainty": uncertainty\[i].item() > 0.5 if uncertainty is not None else False

&nbsp;           }

&nbsp;           report\_objects.append(report\_obj)

&nbsp;

&nbsp;       return report\_objects



class ClinicalVerifier:

&nbsp;   """Verifies clinical consistency of generated reports"""

&nbsp;   def \_\_init\_\_(self, kg: Graph, rules: List\[Dict] = None):

&nbsp;       self.kg = kg

&nbsp;       self.rules = rules or self.\_default\_medical\_rules()

&nbsp;

&nbsp;       # Extract entities and relations from knowledge graph

&nbsp;       self.entities, self.entity\_to\_idx = self.\_extract\_entities()

&nbsp;       self.relations, self.relation\_to\_idx = self.\_extract\_relations()

&nbsp;

&nbsp;   def \_default\_medical\_rules(self) -> List\[Dict]:

&nbsp;       """Define default medical rules for verification"""

&nbsp;       return \[

&nbsp;           {

&nbsp;               "name": "malignancy\_consistency",

&nbsp;               "description": "Malignancy should be consistent across modalities",

&nbsp;               "pattern": r"(malignant|cancer|tumor)",

&nbsp;               "consistency\_check": self.\_check\_malignancy\_consistency

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "location\_consistency",

&nbsp;               "description": "Lesion location should be consistent",

&nbsp;               "pattern": r"(upper|lower|outer|inner|quadrant)",

&nbsp;               "consistency\_check": self.\_check\_location\_consistency

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "finding\_size",

&nbsp;               "description": "Size should be reasonable for the finding",

&nbsp;               "pattern": r"(\\d+\\.?\\d\*)\\s\*(cm|mm)",

&nbsp;               "consistency\_check": self.\_check\_size\_reasonableness

&nbsp;           },

&nbsp;           {

&nbsp;               "name": "biopsy\_recommendation",

&nbsp;               "description": "Biopsy should be recommended for suspicious findings",

&nbsp;               "pattern": r"(suspicious|concerning|worrisome)",

&nbsp;               "consistency\_check": self.\_check\_biopsy\_recommendation

&nbsp;           }

&nbsp;       ]

&nbsp;

&nbsp;   def \_extract\_entities(self) -> Tuple\[List\[str], Dict\[str, int]]:

&nbsp;       """Extract unique entities from the knowledge graph"""

&nbsp;       entities = set()

&nbsp;       for s, p, o in self.kg:

&nbsp;           entities.add(str(s))

&nbsp;           if isinstance(o, URIRef):

&nbsp;               entities.add(str(o))

&nbsp;

&nbsp;       entities = sorted(entities)

&nbsp;       entity\_to\_idx = {entity: idx for idx, entity in enumerate(entities)}

&nbsp;

&nbsp;       return entities, entity\_to\_idx

&nbsp;

&nbsp;   def \_extract\_relations(self) -> Tuple\[List\[str], Dict\[str, int]]:

&nbsp;       """Extract unique relations from the knowledge graph"""

&nbsp;       relations = set()

&nbsp;       for s, p, o in self.kg:

&nbsp;           relations.add(str(p))

&nbsp;

&nbsp;       relations = sorted(relations)

&nbsp;       relation\_to\_idx = {relation: idx for idx, relation in enumerate(relations)}

&nbsp;

&nbsp;       return relations, relation\_to\_idx

&nbsp;

&nbsp;   def \_check\_malignancy\_consistency(self, report: str, entities: List\[str]) -> Dict\[str, Any]:

&nbsp;       """Check malignancy consistency across modalities"""

&nbsp;       # Extract malignancy mentions

&nbsp;       malignancy\_mentions = \[]

&nbsp;       doc = nlp(report)

&nbsp;

&nbsp;       for ent in doc.ents:

&nbsp;           if ent.label\_ in \["DISEASE", "DISORDER"] and any(

&nbsp;               term in ent.text.lower() for term in \["malignant", "cancer", "tumor"]

&nbsp;           ):

&nbsp;               malignancy\_mentions.append(ent.text)

&nbsp;

&nbsp;       # Check for consistency

&nbsp;       consistent = True

&nbsp;       inconsistencies = \[]

&nbsp;

&nbsp;       if len(malignancy\_mentions) > 1:

&nbsp;           # Check if all mentions are consistent

&nbsp;           for i in range(len(malignancy\_mentions) - 1):

&nbsp;               for j in range(i + 1, len(malignancy\_mentions)):

&nbsp;                   # Simple check for consistency

&nbsp;                   if "malignant" in malignancy\_mentions\[i].lower() and "benign" in malignancy\_mentions\[j].lower():

&nbsp;                       consistent = False

&nbsp;                       inconsistencies.append(f"Contradictory terms: {malignancy\_mentions\[i]} and {malignancy\_mentions\[j]}")

&nbsp;

&nbsp;       return {

&nbsp;           "consistent": consistent,

&nbsp;           "inconsistencies": inconsistencies,

&nbsp;           "mentions": malignancy\_mentions

&nbsp;       }

&nbsp;

&nbsp;   def \_check\_location\_consistency(self, report: str, entities: List\[str]) -> Dict\[str, Any]:

&nbsp;       """Check location consistency in the report"""

&nbsp;       # Extract location mentions

&nbsp;       location\_mentions = \[]

&nbsp;       doc = nlp(report)

&nbsp;

&nbsp;       for ent in doc.ents:

&nbsp;           if ent.label\_ == "ANATOMY" or any(

&nbsp;               term in ent.text.lower() for term in \["upper", "lower", "outer", "inner", "quadrant"]

&nbsp;           ):

&nbsp;               location\_mentions.append(ent.text)

&nbsp;

&nbsp;       # Check for consistency

&nbsp;       consistent = True

&nbsp;       inconsistencies = \[]

&nbsp;

&nbsp;       if len(location\_mentions) > 1:

&nbsp;           # Check if all locations are compatible

&nbsp;           incompatible\_pairs = \[

&nbsp;               ("upper", "lower"),

&nbsp;               ("outer", "inner"),

&nbsp;               ("left", "right")

&nbsp;           ]

&nbsp;

&nbsp;           for i in range(len(location\_mentions) - 1):

&nbsp;               for j in range(i + 1, len(location\_mentions)):

&nbsp;                   for pair in incompatible\_pairs:

&nbsp;                       if pair\[0] in location\_mentions\[i].lower() and pair\[1] in location\_mentions\[j].lower():

&nbsp;                           consistent = False

&nbsp;                           inconsistencies.append(f"Potentially incompatible locations: {location\_mentions\[i]} and {location\_mentions\[j]}")

&nbsp;

&nbsp;       return {

&nbsp;           "consistent": consistent,

&nbsp;           "inconsistencies": inconsistencies,

&nbsp;           "mentions": location\_mentions

&nbsp;       }

&nbsp;

&nbsp;   def \_check\_size\_reasonableness(self, report: str, entities: List\[str]) -> Dict\[str, Any]:

&nbsp;       """Check if reported sizes are reasonable"""

&nbsp;       # Extract size mentions

&nbsp;       size\_mentions = \[]

&nbsp;       doc = nlp(report)

&nbsp;

&nbsp;       # Use regex to find size patterns

&nbsp;       size\_pattern = r"(\\d+\\.?\\d\*)\\s\*(cm|mm)"

&nbsp;       matches = re.finditer(size\_pattern, report)

&nbsp;

&nbsp;       for match in matches:

&nbsp;           value = float(match.group(1))

&nbsp;           unit = match.group(2)

&nbsp;

&nbsp;           # Convert to mm for consistency

&nbsp;           if unit == "cm":

&nbsp;               value\_mm = value \* 10

&nbsp;           else:

&nbsp;               value\_mm = value

&nbsp;

&nbsp;           size\_mentions.append({

&nbsp;               "text": match.group(0),

&nbsp;               "value": value,

&nbsp;               "unit": unit,

&nbsp;               "value\_mm": value\_mm

&nbsp;           })

&nbsp;

&nbsp;       # Check for reasonableness

&nbsp;       consistent = True

&nbsp;       inconsistencies = \[]

&nbsp;

&nbsp;       for size in size\_mentions:

&nbsp;           # Check if size is reasonable for a breast lesion

&nbsp;           if size\["value\_mm"] > 50:  # Greater than 5cm

&nbsp;               consistent = False

&nbsp;               inconsistencies.append(f"Unusually large size: {size\['text']}")

&nbsp;           elif size\["value\_mm"] < 2:  # Less than 2mm

&nbsp;               consistent = False

&nbsp;               inconsistencies.append(f"Unusually small size: {size\['text']}")

&nbsp;

&nbsp;       return {

&nbsp;           "consistent": consistent,

&nbsp;           "inconsistencies": inconsistencies,

&nbsp;           "mentions": size\_mentions

&nbsp;       }

&nbsp;

&nbsp;   def \_check\_biopsy\_recommendation(self, report: str, entities: List\[str]) -> Dict\[str, Any]:

&nbsp;       """Check if biopsy is recommended for suspicious findings"""

&nbsp;       # Extract suspicious findings

&nbsp;       suspicious\_mentions = \[]

&nbsp;       doc = nlp(report)

&nbsp;

&nbsp;       for ent in doc.ents:

&nbsp;           if any(term in ent.text.lower() for term in \["suspicious", "concerning", "worrisome"]):

&nbsp;               suspicious\_mentions.append(ent.text)

&nbsp;

&nbsp;       # Check for biopsy recommendation

&nbsp;       biopsy\_mentioned = any(term in report.lower() for term in \["biopsy", "fna", "core needle"])

&nbsp;

&nbsp;       # Check consistency

&nbsp;       consistent = True

&nbsp;       inconsistencies = \[]

&nbsp;

&nbsp;       if suspicious\_mentions and not biopsy\_mentioned:

&nbsp;           consistent = False

&nbsp;           inconsistencies.append("Suspicious findings mentioned but no biopsy recommendation")

&nbsp;

&nbsp;       return {

&nbsp;           "consistent": consistent,

&nbsp;           "inconsistencies": inconsistencies,

&nbsp;           "suspicious\_mentions": suspicious\_mentions,

&nbsp;           "biopsy\_mentioned": biopsy\_mentioned

&nbsp;       }

&nbsp;

&nbsp;   def verify\_report(self, report: str, entities: List\[str] = None) -> Dict\[str, Any]:

&nbsp;       """Verify clinical consistency of a report"""

&nbsp;       verification\_results = {

&nbsp;           "overall\_consistency": True,

&nbsp;           "rule\_results": {}

&nbsp;       }

&nbsp;

&nbsp;       # Apply each rule

&nbsp;       for rule in self.rules:

&nbsp;           rule\_name = rule\["name"]

&nbsp;           pattern = rule\["pattern"]

&nbsp;           check\_func = rule\["consistency\_check"]

&nbsp;

&nbsp;           # Check if pattern is present in the report

&nbsp;           if re.search(pattern, report, re.IGNORECASE):

&nbsp;               # Apply consistency check

&nbsp;               result = check\_func(report, entities or \[])

&nbsp;               verification\_results\["rule\_results"]\[rule\_name] = result

&nbsp;

&nbsp;               # Update overall consistency

&nbsp;               if not result\["consistent"]:

&nbsp;                   verification\_results\["overall\_consistency"] = False

&nbsp;           else:

&nbsp;               # Pattern not found, rule not applicable

&nbsp;               verification\_results\["rule\_results"]\[rule\_name] = {

&nbsp;                   "applicable": False,

&nbsp;                   "consistent": True

&nbsp;               }

&nbsp;

&nbsp;       return verification\_results

&nbsp;

&nbsp;   def highlight\_uncertainty\_in\_report(self, report: str, uncertainty: float) -> str:

&nbsp;       """Highlight uncertain parts of the report"""

&nbsp;       # Split report into sentences

&nbsp;       doc = nlp(report)

&nbsp;       sentences = \[sent.text for sent in doc.sents]

&nbsp;

&nbsp;       # Determine uncertainty level

&nbsp;       if uncertainty < 0.3:

&nbsp;           uncertainty\_level = "low"

&nbsp;           highlight\_start = "\[CONFIDENT: "

&nbsp;           highlight\_end = "]"

&nbsp;       elif uncertainty < 0.7:

&nbsp;           uncertainty\_level = "medium"

&nbsp;           highlight\_start = "\[MODERATE: "

&nbsp;           highlight\_end = "]"

&nbsp;       else:

&nbsp;           uncertainty\_level = "high"

&nbsp;           highlight\_start = "\[UNCERTAIN: "

&nbsp;           highlight\_end = "]"

&nbsp;

&nbsp;       # Highlight sentences based on uncertainty

&nbsp;       highlighted\_sentences = \[]

&nbsp;       for sentence in sentences:

&nbsp;           # Simple heuristic: sentences with hedging language are more uncertain

&nbsp;           hedging\_terms = \["may", "might", "possibly", "potentially", "could", "suggests", "consistent with"]

&nbsp;           has\_hedging = any(term in sentence.lower() for term in hedging\_terms)

&nbsp;

&nbsp;           if has\_hedging:

&nbsp;               highlighted\_sentences.append(f"{highlight\_start}{sentence.strip()}{highlight\_end}")

&nbsp;           else:

&nbsp;               highlighted\_sentences.append(sentence.strip())

&nbsp;

&nbsp;       # Add uncertainty note at the beginning

&nbsp;       highlighted\_report = f"Overall uncertainty level: {uncertainty\_level.upper()} ({uncertainty:.2f})\\n\\n"

&nbsp;       highlighted\_report += " ".join(highlighted\_sentences)

&nbsp;

&nbsp;       return highlighted\_report



class MedicalReportDataset(Dataset):

&nbsp;   """Dataset for fine-tuning the report generator"""

&nbsp;   def \_\_init\_\_(self, data: List\[Dict], tokenizer, max\_length: int = 512):

&nbsp;       self.data = data

&nbsp;       self.tokenizer = tokenizer

&nbsp;       self.max\_length = max\_length

&nbsp;

&nbsp;   def \_\_len\_\_(self):

&nbsp;       return len(self.data)

&nbsp;

&nbsp;   def \_\_getitem\_\_(self, idx):

&nbsp;       item = self.data\[idx]

&nbsp;

&nbsp;       # Get input and target texts

&nbsp;       input\_text = item.get("input", "")

&nbsp;       target\_text = item.get("target", "")

&nbsp;

&nbsp;       # Tokenize inputs

&nbsp;       input\_ids = self.tokenizer(input\_text, return\_tensors="pt", padding="max\_length",

&nbsp;                                 truncation=True, max\_length=self.max\_length)

&nbsp;       target\_ids = self.tokenizer(target\_text, return\_tensors="pt", padding="max\_length",

&nbsp;                                  truncation=True, max\_length=self.max\_length)

&nbsp;

&nbsp;       return {

&nbsp;           "input\_ids": input\_ids.input\_ids.squeeze(),

&nbsp;           "attention\_mask": input\_ids.attention\_mask.squeeze(),

&nbsp;           "labels": target\_ids.input\_ids.squeeze(),

&nbsp;           "decoder\_attention\_mask": target\_ids.attention\_mask.squeeze()

&nbsp;       }



class LLMReportGenerationPipeline:

&nbsp;   """Complete pipeline for LLM-based report generation"""

&nbsp;   def \_\_init\_\_(self, kg: Graph, model\_name: str = "gpt2", fused\_dim: int = 512,

&nbsp;                output\_dim: int = 768, max\_length: int = 512):

&nbsp;       self.kg = kg

&nbsp;       self.model\_name = model\_name

&nbsp;       self.fused\_dim = fused\_dim

&nbsp;       self.output\_dim = output\_dim

&nbsp;       self.max\_length = max\_length

&nbsp;

&nbsp;       # Initialize report generator

&nbsp;       self.report\_generator = LLMReportGenerator(

&nbsp;           model\_name=model\_name,

&nbsp;           fused\_dim=fused\_dim,

&nbsp;           output\_dim=output\_dim,

&nbsp;           max\_length=max\_length

&nbsp;       )

&nbsp;

&nbsp;       # Initialize clinical verifier

&nbsp;       self.verifier = ClinicalVerifier(kg)

&nbsp;

&nbsp;   def generate\_reports(self, fused\_rep: torch.Tensor, uncertainty: torch.Tensor = None,

&nbsp;                       prompts: List\[str] = None, max\_length: int = None,

&nbsp;                       verify: bool = True) -> List\[Dict\[str, Any]]:

&nbsp;       """Generate reports with optional verification"""

&nbsp;       batch\_size = fused\_rep.size(0)

&nbsp;       max\_length = max\_length or self.max\_length

&nbsp;

&nbsp;       # Generate reports with uncertainty

&nbsp;       if uncertainty is not None:

&nbsp;           reports = self.report\_generator.generate\_report\_with\_uncertainty(

&nbsp;               fused\_rep, uncertainty, prompts, max\_length

&nbsp;           )

&nbsp;       else:

&nbsp;           # Generate reports without uncertainty

&nbsp;           generated\_texts = self.report\_generator.generate\_report(

&nbsp;               fused\_rep, uncertainty, prompts, max\_length

&nbsp;           )

&nbsp;

&nbsp;           reports = \[]

&nbsp;           for i in range(batch\_size):

&nbsp;               report\_obj = {

&nbsp;                   "report": generated\_texts\[i],

&nbsp;                   "uncertainty": None,

&nbsp;                   "high\_uncertainty": False

&nbsp;               }

&nbsp;               reports.append(report\_obj)

&nbsp;

&nbsp;       # Verify reports if requested

&nbsp;       if verify:

&nbsp;           for i in range(batch\_size):

&nbsp;               report = reports\[i]\["report"]

&nbsp;               verification = self.verifier.verify\_report(report)

&nbsp;               reports\[i]\["verification"] = verification

&nbsp;

&nbsp;               # Highlight uncertainty in report

&nbsp;               if reports\[i]\["uncertainty"] is not None:

&nbsp;                   highlighted\_report = self.verifier.highlight\_uncertainty\_in\_report(

&nbsp;                       report, reports\[i]\["uncertainty"]

&nbsp;                   )

&nbsp;                   reports\[i]\["highlighted\_report"] = highlighted\_report

&nbsp;

&nbsp;       return reports

&nbsp;

&nbsp;   def fine\_tune(self, train\_data: List\[Dict], val\_data: List\[Dict] = None,

&nbsp;                output\_dir: str = None, epochs: int = 3, batch\_size: int = 4,

&nbsp;                learning\_rate: float = 5e-5):

&nbsp;       """Fine-tune the report generator on medical reports"""

&nbsp;       # Create datasets

&nbsp;       train\_dataset = MedicalReportDataset(train\_data, self.report\_generator.tokenizer, self.max\_length)

&nbsp;

&nbsp;       val\_dataset = None

&nbsp;       if val\_data:

&nbsp;           val\_dataset = MedicalReportDataset(val\_data, self.report\_generator.tokenizer, self.max\_length)

&nbsp;

&nbsp;       # Create data collator

&nbsp;       data\_collator = DataCollatorForSeq2Seq(

&nbsp;           tokenizer=self.report\_generator.tokenizer,

&nbsp;           model=self.report\_generator.model

&nbsp;       )

&nbsp;

&nbsp;       # Set up training arguments

&nbsp;       training\_args = TrainingArguments(

&nbsp;           output\_dir=output\_dir or "./results",

&nbsp;           num\_train\_epochs=epochs,

&nbsp;           per\_device\_train\_batch\_size=batch\_size,

&nbsp;           per\_device\_eval\_batch\_size=batch\_size,

&nbsp;           warmup\_steps=500,

&nbsp;           weight\_decay=0.01,

&nbsp;           logging\_dir="./logs",

&nbsp;           logging\_steps=10,

&nbsp;           save\_steps=500,

&nbsp;           evaluation\_strategy="epoch" if val\_dataset else "no",

&nbsp;           save\_strategy="epoch",

&nbsp;           load\_best\_model\_at\_end=True if val\_dataset else False,

&nbsp;           metric\_for\_best\_model="loss",

&nbsp;           greater\_is\_better=False,

&nbsp;           learning\_rate=learning\_rate

&nbsp;       )

&nbsp;

&nbsp;       # Create trainer

&nbsp;       trainer = Trainer(

&nbsp;           model=self.report\_generator.model,

&nbsp;           args=training\_args,

&nbsp;           data\_collator=data\_collator,

&nbsp;           train\_dataset=train\_dataset,

&nbsp;           eval\_dataset=val\_dataset

&nbsp;       )

&nbsp;

&nbsp;       # Fine-tune model

&nbsp;       trainer.train()

&nbsp;

&nbsp;       # Save model

&nbsp;       if output\_dir:

&nbsp;           trainer.save\_model(output\_dir)

&nbsp;           self.report\_generator.tokenizer.save\_pretrained(output\_dir)

&nbsp;

&nbsp;       return trainer



\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Load knowledge graph

&nbsp;   kg\_path = "output/knowledge\_graph.ttl"

&nbsp;   kg = Graph()

&nbsp;   kg.parse(kg\_path, format="turtle")

&nbsp;

&nbsp;   # Create pipeline

&nbsp;   pipeline = LLMReportGenerationPipeline(

&nbsp;       kg=kg,

&nbsp;       model\_name="gpt2",  # Use "t5-small" for T5 model

&nbsp;       fused\_dim=512,

&nbsp;       output\_dim=768,

&nbsp;       max\_length=512

&nbsp;   )

&nbsp;

&nbsp;   # Create dummy fused representations and uncertainty

&nbsp;   batch\_size = 2

&nbsp;   fused\_dim = 512

&nbsp;   fused\_rep = torch.randn(batch\_size, fused\_dim).to(device)

&nbsp;   uncertainty = torch.rand(batch\_size).to(device)

&nbsp;

&nbsp;   # Create prompts

&nbsp;   prompts = \[

&nbsp;       "Patient presents with a suspicious mass in the upper outer quadrant of the right breast.",

&nbsp;       "Screening mammogram shows microcalcifications in the left breast."

&nbsp;   ]

&nbsp;

&nbsp;   # Generate reports

&nbsp;   reports = pipeline.generate\_reports(

&nbsp;       fused\_rep=fused\_rep,

&nbsp;       uncertainty=uncertainty,

&nbsp;       prompts=prompts,

&nbsp;       verify=True

&nbsp;   )

&nbsp;

&nbsp;   # Print generated reports

&nbsp;   for i, report in enumerate(reports):

&nbsp;       print(f"\\nReport {i+1}:")

&nbsp;       print(f"Uncertainty: {report\['uncertainty']:.2f}")

&nbsp;       print(f"Report: {report\['report']}")

&nbsp;

&nbsp;       if "highlighted\_report" in report:

&nbsp;           print(f"Highlighted Report: {report\['highlighted\_report']}")

&nbsp;

&nbsp;       if "verification" in report:

&nbsp;           print(f"Verification: {report\['verification']}")

&nbsp;

&nbsp;   # Create dummy training data for fine-tuning

&nbsp;   train\_data = \[

&nbsp;       {

&nbsp;           "input": "Patient presents with a suspicious mass in the upper outer quadrant of the right breast.",

&nbsp;           "target": "The patient presents with a suspicious mass in the upper outer quadrant of the right breast. Biopsy is recommended for further evaluation."

&nbsp;       },

&nbsp;       {

&nbsp;           "input": "Screening mammogram shows microcalcifications in the left breast.",

&nbsp;           "target": "Screening mammogram demonstrates scattered microcalcifications in the left breast. These are likely benign, but follow-up in 6 months is recommended."

&nbsp;       }

&nbsp;   ]

&nbsp;

&nbsp;   # Fine-tune the model (commented out to avoid long training time)

&nbsp;   # pipeline.fine\_tune(train\_data, output\_dir="output/report\_generator")

&nbsp;

&nbsp;   print("\\nLLM-based Report Generation completed!")
\# Set device

device = torch.device('cuda' if torch.cuda.is\_available() else 'cpu')



class ClinicianFeedbackInterface:

&nbsp;   """Interface for collecting clinician feedback on model predictions"""

&nbsp;   def \_\_init\_\_(self, feedback\_db\_path: str = "feedback\_db.json"):

&nbsp;       self.feedback\_db\_path = feedback\_db\_path

&nbsp;       self.feedback\_db = self.\_load\_feedback\_db()

&nbsp;       self.feedback\_queue = queue.Queue()

&nbsp;       self.active = False

&nbsp;       self.feedback\_thread = None

&nbsp;

&nbsp;   def \_load\_feedback\_db(self) -> Dict:

&nbsp;       """Load feedback database from file"""

&nbsp;       if os.path.exists(self.feedback\_db\_path):

&nbsp;           with open(self.feedback\_db\_path, 'r') as f:

&nbsp;               return json.load(f)

&nbsp;       return {

&nbsp;           "feedbacks": \[],

&nbsp;           "statistics": {

&nbsp;               "total\_feedbacks": 0,

&nbsp;               "agreement\_rate": 0.0,

&nbsp;               "common\_corrections": {}

&nbsp;           }

&nbsp;       }

&nbsp;

&nbsp;   def \_save\_feedback\_db(self):

&nbsp;       """Save feedback database to file"""

&nbsp;       with open(self.feedback\_db\_path, 'w') as f:

&nbsp;           json.dump(self.feedback\_db, f, indent=2)

&nbsp;

&nbsp;   def start\_feedback\_collection(self):

&nbsp;       """Start the feedback collection thread"""

&nbsp;       self.active = True

&nbsp;       self.feedback\_thread = threading.Thread(target=self.\_collect\_feedback)

&nbsp;       self.feedback\_thread.daemon = True

&nbsp;       self.feedback\_thread.start()

&nbsp;

&nbsp;   def stop\_feedback\_collection(self):

&nbsp;       """Stop the feedback collection thread"""

&nbsp;       self.active = False

&nbsp;       if self.feedback\_thread:

&nbsp;           self.feedback\_thread.join()

&nbsp;

&nbsp;   def \_collect\_feedback(self):

&nbsp;       """Simulate collecting feedback from clinicians"""

&nbsp;       while self.active:

&nbsp;           # In a real implementation, this would connect to a UI or API

&nbsp;           # For simulation, we'll generate random feedback

&nbsp;           if random.random() < 0.1:  # 10% chance of feedback per iteration

&nbsp;               # Generate random feedback

&nbsp;               feedback\_id = str(uuid.uuid4())

&nbsp;               case\_id = f"case\_{random.randint(1000, 9999)}"

&nbsp;

&nbsp;               # Random prediction and correction

&nbsp;               predictions = \["malignant", "benign", "suspicious"]

&nbsp;               prediction = random.choice(predictions)

&nbsp;               correction = random.choice(predictions)

&nbsp;

&nbsp;               # Confidence level

&nbsp;               confidence = random.uniform(0.5, 1.0)

&nbsp;

&nbsp;               # Feedback type

&nbsp;               feedback\_types = \["correction", "confirmation", "additional\_info"]

&nbsp;               feedback\_type = random.choice(feedback\_types)

&nbsp;

&nbsp;               # Create feedback object

&nbsp;               feedback = {

&nbsp;                   "id": feedback\_id,

&nbsp;                   "case\_id": case\_id,

&nbsp;                   "timestamp": datetime.now().isoformat(),

&nbsp;                   "prediction": prediction,

&nbsp;                   "correction": correction,

&nbsp;                   "confidence": confidence,

&nbsp;                   "feedback\_type": feedback\_type,

&nbsp;                   "processed": False

&nbsp;               }

&nbsp;

&nbsp;               # Add to queue

&nbsp;               self.feedback\_queue.put(feedback)

&nbsp;

&nbsp;               # Add to database

&nbsp;               self.feedback\_db\["feedbacks"].append(feedback)

&nbsp;               self.feedback\_db\["statistics"]\["total\_feedbacks"] += 1

&nbsp;

&nbsp;               # Update statistics

&nbsp;               if prediction == correction:

&nbsp;                   self.feedback\_db\["statistics"]\["agreement\_rate"] = (

&nbsp;                       (self.feedback\_db\["statistics"]\["agreement\_rate"] \*

&nbsp;                        (self.feedback\_db\["statistics"]\["total\_feedbacks"] - 1) + 1) /

&nbsp;                       self.feedback\_db\["statistics"]\["total\_feedbacks"]

&nbsp;                   )

&nbsp;               else:

&nbsp;                   self.feedback\_db\["statistics"]\["agreement\_rate"] = (

&nbsp;                       (self.feedback\_db\["statistics"]\["agreement\_rate"] \*

&nbsp;                        (self.feedback\_db\["statistics"]\["total\_feedbacks"] - 1)) /

&nbsp;                       self.feedback\_db\["statistics"]\["total\_feedbacks"]

&nbsp;                   )

&nbsp;

&nbsp;               # Update common corrections

&nbsp;               correction\_key = f"{prediction}\_to\_{correction}"

&nbsp;               if correction\_key not in self.feedback\_db\["statistics"]\["common\_corrections"]:

&nbsp;                   self.feedback\_db\["statistics"]\["common\_corrections"]\[correction\_key] = 0

&nbsp;               self.feedback\_db\["statistics"]\["common\_corrections"]\[correction\_key] += 1

&nbsp;

&nbsp;               # Save database

&nbsp;               self.\_save\_feedback\_db()

&nbsp;

&nbsp;               print(f"Collected feedback: {feedback}")

&nbsp;

&nbsp;           # Sleep for a while

&nbsp;           time.sleep(1)

&nbsp;

&nbsp;   def get\_feedback(self) -> List\[Dict]:

&nbsp;       """Get all feedback from the queue"""

&nbsp;       feedbacks = \[]

&nbsp;       while not self.feedback\_queue.empty():

&nbsp;           feedbacks.append(self.feedback\_queue.get())

&nbsp;       return feedbacks

&nbsp;

&nbsp;   def get\_feedback\_statistics(self) -> Dict:

&nbsp;       """Get feedback statistics"""

&nbsp;       return self.feedback\_db\["statistics"]

&nbsp;

&nbsp;   def add\_feedback(self, feedback: Dict):

&nbsp;       """Add feedback to the database"""

&nbsp;       feedback\["id"] = str(uuid.uuid4())

&nbsp;       feedback\["timestamp"] = datetime.now().isoformat()

&nbsp;       feedback\["processed"] = False

&nbsp;

&nbsp;       self.feedback\_db\["feedbacks"].append(feedback)

&nbsp;       self.feedback\_db\["statistics"]\["total\_feedbacks"] += 1

&nbsp;

&nbsp;       # Update statistics

&nbsp;       if feedback\["prediction"] == feedback\["correction"]:

&nbsp;           self.feedback\_db\["statistics"]\["agreement\_rate"] = (

&nbsp;               (self.feedback\_db\["statistics"]\["agreement\_rate"] \*

&nbsp;                (self.feedback\_db\["statistics"]\["total\_feedbacks"] - 1) + 1) /

&nbsp;               self.feedback\_db\["statistics"]\["total\_feedbacks"]

&nbsp;           )

&nbsp;       else:

&nbsp;           self.feedback\_db\["statistics"]\["agreement\_rate"] = (

&nbsp;               (self.feedback\_db\["statistics"]\["agreement\_rate"] \*

&nbsp;                (self.feedback\_db\["statistics"]\["total\_feedbacks"] - 1)) /

&nbsp;               self.feedback\_db\["statistics"]\["total\_feedbacks"]

&nbsp;           )

&nbsp;

&nbsp;       # Update common corrections

&nbsp;       correction\_key = f"{feedback\['prediction']}\_to\_{feedback\['correction']}"

&nbsp;       if correction\_key not in self.feedback\_db\["statistics"]\["common\_corrections"]:

&nbsp;           self.feedback\_db\["statistics"]\["common\_corrections"]\[correction\_key] = 0

&nbsp;       self.feedback\_db\["statistics"]\["common\_corrections"]\[correction\_key] += 1

&nbsp;

&nbsp;       # Save database

&nbsp;       self.\_save\_feedback\_db()

&nbsp;

&nbsp;       # Add to queue

&nbsp;       self.feedback\_queue.put(feedback)



class FederatedClient:

&nbsp;   """Represents a client in federated learning"""

&nbsp;   def \_\_init\_\_(self, client\_id: str, model: nn.Module, data\_loader: DataLoader):

&nbsp;       self.client\_id = client\_id

&nbsp;       self.model = model

&nbsp;       self.data\_loader = data\_loader

&nbsp;       self.local\_epochs = 1

&nbsp;       self.batch\_size = data\_loader.batch\_size

&nbsp;       self.learning\_rate = 0.01

&nbsp;

&nbsp;   def train(self, global\_model: nn.Module) -> Dict\[str, torch.Tensor]:

&nbsp;       """Train the local model and return model updates"""

&nbsp;       # Copy global model weights

&nbsp;       self.model.load\_state\_dict(global\_model.state\_dict())

&nbsp;

&nbsp;       # Set model to training mode

&nbsp;       self.model.train()

&nbsp;

&nbsp;       # Define optimizer

&nbsp;       optimizer = SGD(self.model.parameters(), lr=self.learning\_rate)

&nbsp;       criterion = nn.CrossEntropyLoss()

&nbsp;

&nbsp;       # Train for local epochs

&nbsp;       for epoch in range(self.local\_epochs):

&nbsp;           total\_loss = 0.0

&nbsp;           correct = 0

&nbsp;           total = 0

&nbsp;

&nbsp;           for batch\_idx, (data, target) in enumerate(self.data\_loader):

&nbsp;               # Move data to device

&nbsp;               data, target = data.to(device), target.to(device)

&nbsp;

&nbsp;               # Zero gradients

&nbsp;               optimizer.zero\_grad()

&nbsp;

&nbsp;               # Forward pass

&nbsp;               output = self.model(data)

&nbsp;

&nbsp;               # Compute loss

&nbsp;               loss = criterion(output, target)

&nbsp;

&nbsp;               # Backward pass

&nbsp;               loss.backward()

&nbsp;               optimizer.step()

&nbsp;

&nbsp;               # Statistics

&nbsp;               total\_loss += loss.item()

&nbsp;               \_, predicted = output.max(1)

&nbsp;               total += target.size(0)

&nbsp;               correct += predicted.eq(target).sum().item()

&nbsp;

&nbsp;           print(f"Client {self.client\_id}, Epoch {epoch+1}, Loss: {total\_loss/len(self.data\_loader):.4f}, Accuracy: {100.\*correct/total:.2f}%")

&nbsp;

&nbsp;       # Compute model updates (difference between local and global model)

&nbsp;       global\_state\_dict = global\_model.state\_dict()

&nbsp;       local\_state\_dict = self.model.state\_dict()

&nbsp;

&nbsp;       updates = {}

&nbsp;       for name in global\_state\_dict:

&nbsp;           updates\[name] = local\_state\_dict\[name] - global\_state\_dict\[name]

&nbsp;

&nbsp;       return updates

&nbsp;

&nbsp;   def evaluate(self, global\_model: nn.Module) -> Dict\[str, float]:

&nbsp;       """Evaluate the global model on local data"""

&nbsp;       # Copy global model weights

&nbsp;       self.model.load\_state\_dict(global\_model.state\_dict())

&nbsp;

&nbsp;       # Set model to evaluation mode

&nbsp;       self.model.eval()

&nbsp;

&nbsp;       criterion = nn.CrossEntropyLoss()

&nbsp;       total\_loss = 0.0

&nbsp;       correct = 0

&nbsp;       total = 0

&nbsp;

&nbsp;       with torch.no\_grad():

&nbsp;           for data, target in self.data\_loader:

&nbsp;               # Move data to device

&nbsp;               data, target = data.to(device), target.to(device)

&nbsp;

&nbsp;               # Forward pass

&nbsp;               output = self.model(data)

&nbsp;

&nbsp;               # Compute loss

&nbsp;               loss = criterion(output, target)

&nbsp;

&nbsp;               # Statistics

&nbsp;               total\_loss += loss.item()

&nbsp;               \_, predicted = output.max(1)

&nbsp;               total += target.size(0)

&nbsp;               correct += predicted.eq(target).sum().item()

&nbsp;

&nbsp;       return {

&nbsp;           "loss": total\_loss / len(self.data\_loader),

&nbsp;           "accuracy": 100. \* correct / total

&nbsp;       }



class FederatedServer:

&nbsp;   """Server for federated learning"""

&nbsp;   def \_\_init\_\_(self, model: nn.Module, num\_rounds: int = 10):

&nbsp;       self.model = model

&nbsp;       self.num\_rounds = num\_rounds

&nbsp;       self.clients = \[]

&nbsp;       self.client\_updates = \[]

&nbsp;       self.round\_metrics = \[]

&nbsp;

&nbsp;   def add\_client(self, client: FederatedClient):

&nbsp;       """Add a client to the federation"""

&nbsp;       self.clients.append(client)

&nbsp;

&nbsp;   def aggregate\_updates(self, updates: List\[Dict\[str, torch.Tensor]]) -> Dict\[str, torch.Tensor]:

&nbsp;       """Aggregate model updates from clients using FedAvg"""

&nbsp;       if not updates:

&nbsp;           return {}

&nbsp;

&nbsp;       # Initialize aggregated updates

&nbsp;       aggregated\_updates = {}

&nbsp;

&nbsp;       # Get the first update to initialize

&nbsp;       first\_update = updates\[0]

&nbsp;       for name in first\_update:

&nbsp;           aggregated\_updates\[name] = torch.zeros\_like(first\_update\[name])

&nbsp;

&nbsp;       # Sum all updates

&nbsp;       for update in updates:

&nbsp;           for name in update:

&nbsp;               aggregated\_updates\[name] += update\[name]

&nbsp;

&nbsp;       # Average the updates

&nbsp;       for name in aggregated\_updates:

&nbsp;           aggregated\_updates\[name] /= len(updates)

&nbsp;

&nbsp;       return aggregated\_updates

&nbsp;

&nbsp;   def train(self):

&nbsp;       """Run federated learning for multiple rounds"""

&nbsp;       for round\_num in range(1, self.num\_rounds + 1):

&nbsp;           print(f"\\n--- Federated Learning Round {round\_num}/{self.num\_rounds} ---")

&nbsp;

&nbsp;           # Select clients for this round (in a real implementation, this would be random)

&nbsp;           selected\_clients = self.clients

&nbsp;

&nbsp;           # Send global model to clients and collect updates

&nbsp;           updates = \[]

&nbsp;           client\_metrics = \[]

&nbsp;

&nbsp;           for client in selected\_clients:

&nbsp;               print(f"Training client {client.client\_id}...")

&nbsp;               update = client.train(self.model)

&nbsp;               updates.append(update)

&nbsp;

&nbsp;               # Evaluate client

&nbsp;               metrics = client.evaluate(self.model)

&nbsp;               client\_metrics.append(metrics)

&nbsp;               print(f"Client {client.client\_id} - Loss: {metrics\['loss']:.4f}, Accuracy: {metrics\['accuracy']:.2f}%")

&nbsp;

&nbsp;           # Aggregate updates

&nbsp;           aggregated\_updates = self.aggregate\_updates(updates)

&nbsp;

&nbsp;           # Update global model

&nbsp;           global\_state\_dict = self.model.state\_dict()

&nbsp;           for name in aggregated\_updates:

&nbsp;               global\_state\_dict\[name] += aggregated\_updates\[name]

&nbsp;           self.model.load\_state\_dict(global\_state\_dict)

&nbsp;

&nbsp;           # Calculate average metrics

&nbsp;           avg\_loss = sum(m\["loss"] for m in client\_metrics) / len(client\_metrics)

&nbsp;           avg\_accuracy = sum(m\["accuracy"] for m in client\_metrics) / len(client\_metrics)

&nbsp;

&nbsp;           # Store round metrics

&nbsp;           round\_metrics = {

&nbsp;               "round": round\_num,

&nbsp;               "avg\_loss": avg\_loss,

&nbsp;               "avg\_accuracy": avg\_accuracy,

&nbsp;               "num\_clients": len(selected\_clients)

&nbsp;           }

&nbsp;           self.round\_metrics.append(round\_metrics)

&nbsp;

&nbsp;           print(f"Round {round\_num} - Avg Loss: {avg\_loss:.4f}, Avg Accuracy: {avg\_accuracy:.2f}%")

&nbsp;

&nbsp;       return self.round\_metrics



class ModelUpdater:

&nbsp;   """Updates model and knowledge based on feedback and federated learning"""

&nbsp;   def \_\_init\_\_(self, model: nn.Module, kg: Graph, output\_dir: str = "model\_updates"):

&nbsp;       self.model = model

&nbsp;       self.kg = kg

&nbsp;       self.output\_dir = output\_dir

&nbsp;       self.update\_history = \[]

&nbsp;

&nbsp;       # Create output directory

&nbsp;       os.makedirs(output\_dir, exist\_ok=True)

&nbsp;

&nbsp;       # Initialize optimizer

&nbsp;       self.optimizer = Adam(self.model.parameters(), lr=1e-4)

&nbsp;

&nbsp;       # Initialize loss function

&nbsp;       self.criterion = nn.CrossEntropyLoss()

&nbsp;

&nbsp;       # Initialize tensorboard writer

&nbsp;       self.writer = SummaryWriter(log\_dir=os.path.join(output\_dir, "logs"))

&nbsp;

&nbsp;   def update\_from\_feedback(self, feedbacks: List\[Dict], data\_loader: DataLoader = None):

&nbsp;       """Update model based on clinician feedback"""

&nbsp;       if not feedbacks:

&nbsp;           return

&nbsp;

&nbsp;       print(f"Updating model with {len(feedbacks)} feedbacks...")

&nbsp;

&nbsp;       # Set model to training mode

&nbsp;       self.model.train()

&nbsp;

&nbsp;       # Update model for each feedback

&nbsp;       for feedback in feedbacks:

&nbsp;           # Skip if already processed

&nbsp;           if feedback.get("processed", False):

&nbsp;               continue

&nbsp;

&nbsp;           # In a real implementation, we would use the feedback to create training examples

&nbsp;           # For simulation, we'll use a dummy update

&nbsp;

&nbsp;           # Create dummy data for training

&nbsp;           if data\_loader:

&nbsp;               # Train on a batch of data

&nbsp;               for batch\_idx, (data, target) in enumerate(data\_loader):

&nbsp;                   if batch\_idx >= 1:  # Only use one batch for simulation

&nbsp;                       break

&nbsp;

&nbsp;                   # Move data to device

&nbsp;                   data, target = data.to(device), target.to(device)

&nbsp;

&nbsp;                   # Zero gradients

&nbsp;                   self.optimizer.zero\_grad()

&nbsp;

&nbsp;                   # Forward pass

&nbsp;                   output = self.model(data)

&nbsp;

&nbsp;                   # Compute loss

&nbsp;                   loss = self.criterion(output, target)

&nbsp;

&nbsp;                   # Backward pass

&nbsp;                   loss.backward()

&nbsp;                   self.optimizer.step()

&nbsp;

&nbsp;                   # Log to tensorboard

&nbsp;                   self.writer.add\_scalar("feedback\_update/loss", loss.item(), len(self.update\_history))

&nbsp;

&nbsp;           # Update knowledge graph based on feedback

&nbsp;           self.\_update\_knowledge\_graph(feedback)

&nbsp;

&nbsp;           # Mark feedback as processed

&nbsp;           feedback\["processed"] = True

&nbsp;

&nbsp;           # Add to update history

&nbsp;           self.update\_history.append({

&nbsp;               "timestamp": datetime.now().isoformat(),

&nbsp;               "type": "feedback",

&nbsp;               "feedback\_id": feedback\["id"],

&nbsp;               "description": f"Updated model based on feedback: {feedback\['prediction']} -> {feedback\['correction']}"

&nbsp;           })

&nbsp;

&nbsp;       # Save updated model

&nbsp;       self.\_save\_model("feedback\_update")

&nbsp;

&nbsp;       print("Model updated based on feedback.")

&nbsp;

&nbsp;   def \_update\_knowledge\_graph(self, feedback: Dict):

&nbsp;       """Update knowledge graph based on feedback"""

&nbsp;       # Extract entities from feedback

&nbsp;       prediction = feedback\["prediction"]

&nbsp;       correction = feedback\["correction"]

&nbsp;

&nbsp;       # Create new triples based on feedback

&nbsp;       # This is a simplified example - in practice, you would have more sophisticated logic

&nbsp;

&nbsp;       # Create entities

&nbsp;       prediction\_entity = URIRef(f"http://example.org/medical/{prediction}")

&nbsp;       correction\_entity = URIRef(f"http://example.org/medical/{correction}")

&nbsp;

&nbsp;       # Add correction relationship

&nbsp;       self.kg.add((prediction\_entity, URIRef("http://example.org/medical/correctedTo"), correction\_entity))

&nbsp;

&nbsp;       # Add timestamp

&nbsp;       self.kg.add((

&nbsp;           prediction\_entity,

&nbsp;           URIRef("http://example.org/medical/correctedAt"),

&nbsp;           Literal(datetime.now().isoformat())

&nbsp;       ))

&nbsp;

&nbsp;       # Save updated knowledge graph

&nbsp;       kg\_path = os.path.join(self.output\_dir, "updated\_knowledge\_graph.ttl")

&nbsp;       self.kg.serialize(destination=kg\_path, format='turtle')

&nbsp;

&nbsp;       print(f"Knowledge graph updated based on feedback: {prediction} -> {correction}")

&nbsp;

&nbsp;   def update\_from\_federated\_learning(self, round\_metrics: List\[Dict]):

&nbsp;       """Update model based on federated learning results"""

&nbsp;       if not round\_metrics:

&nbsp;           return

&nbsp;

&nbsp;       print("Updating model based on federated learning results...")

&nbsp;

&nbsp;       # Log metrics to tensorboard

&nbsp;       for metrics in round\_metrics:

&nbsp;           self.writer.add\_scalar("federated\_learning/avg\_loss", metrics\["avg\_loss"], metrics\["round"])

&nbsp;           self.writer.add\_scalar("federated\_learning/avg\_accuracy", metrics\["avg\_accuracy"], metrics\["round"])

&nbsp;

&nbsp;       # Save updated model

&nbsp;       self.\_save\_model("federated\_update")

&nbsp;

&nbsp;       # Add to update history

&nbsp;       self.update\_history.append({

&nbsp;           "timestamp": datetime.now().isoformat(),

&nbsp;           "type": "federated\_learning",

&nbsp;           "description": f"Updated model based on federated learning round {metrics\['round']}"

&nbsp;       })

&nbsp;

&nbsp;       print("Model updated based on federated learning.")

&nbsp;

&nbsp;   def \_save\_model(self, update\_type: str):

&nbsp;       """Save model with timestamp"""

&nbsp;       timestamp = datetime.now().strftime("%Y%m%d\_%H%M%S")

&nbsp;       model\_path = os.path.join(self.output\_dir, f"model\_{update\_type}\_{timestamp}.pt")

&nbsp;

&nbsp;       torch.save({

&nbsp;           "model\_state\_dict": self.model.state\_dict(),

&nbsp;           "optimizer\_state\_dict": self.optimizer.state\_dict(),

&nbsp;           "timestamp": timestamp,

&nbsp;           "update\_type": update\_type

&nbsp;       }, model\_path)

&nbsp;

&nbsp;       print(f"Model saved to {model\_path}")

&nbsp;

&nbsp;   def get\_update\_history(self) -> List\[Dict]:

&nbsp;       """Get model update history"""

&nbsp;       return self.update\_history

&nbsp;

&nbsp;   def plot\_training\_progress(self):

&nbsp;       """Plot training progress"""

&nbsp;       if not self.update\_history:

&nbsp;           print("No update history available.")

&nbsp;           return

&nbsp;

&nbsp;       # Create figure

&nbsp;       fig, ax = plt.subplots(figsize=(10, 6))

&nbsp;

&nbsp;       # Extract data

&nbsp;       timestamps = \[]

&nbsp;       feedback\_updates = \[]

&nbsp;       federated\_updates = \[]

&nbsp;

&nbsp;       for update in self.update\_history:

&nbsp;           timestamps.append(update\["timestamp"])

&nbsp;           if update\["type"] == "feedback":

&nbsp;               feedback\_updates.append(len(feedback\_updates) + 1)

&nbsp;               federated\_updates.append(None)

&nbsp;           elif update\["type"] == "federated\_learning":

&nbsp;               federated\_updates.append(len(federated\_updates) + 1)

&nbsp;               feedback\_updates.append(None)

&nbsp;

&nbsp;       # Plot

&nbsp;       ax.plot(timestamps, feedback\_updates, 'o-', label='Feedback Updates')

&nbsp;       ax.plot(timestamps, federated\_updates, 'o-', label='Federated Learning Updates')

&nbsp;

&nbsp;       # Formatting

&nbsp;       ax.set\_xlabel('Timestamp')

&nbsp;       ax.set\_ylabel('Update Count')

&nbsp;       ax.set\_title('Model Update History')

&nbsp;       ax.legend()

&nbsp;       ax.grid(True)

&nbsp;

&nbsp;       # Rotate x-axis labels

&nbsp;       plt.xticks(rotation=45)

&nbsp;

&nbsp;       # Save plot

&nbsp;       plot\_path = os.path.join(self.output\_dir, "training\_progress.png")

&nbsp;       plt.tight\_layout()

&nbsp;       plt.savefig(plot\_path)

&nbsp;       plt.close()

&nbsp;

&nbsp;       print(f"Training progress plot saved to {plot\_path}")



class FeedbackLoop:

&nbsp;   """Implements the feedback loop between dataset processing and model components"""

&nbsp;   def \_\_init\_\_(self, feedback\_interface: ClinicianFeedbackInterface,

&nbsp;                federated\_server: FederatedServer, model\_updater: ModelUpdater,

&nbsp;                dataset\_processor=None, model\_components=None):

&nbsp;       self.feedback\_interface = feedback\_interface

&nbsp;       self.federated\_server = federated\_server

&nbsp;       self.model\_updater = model\_updater

&nbsp;       self.dataset\_processor = dataset\_processor

&nbsp;       self.model\_components = model\_components

&nbsp;

&nbsp;       self.active = False

&nbsp;       self.feedback\_thread = None

&nbsp;       self.federated\_thread = None

&nbsp;

&nbsp;       # Feedback loop parameters

&nbsp;       self.feedback\_interval = 60  # seconds

&nbsp;       self.federated\_interval = 3600  # seconds (1 hour)

&nbsp;       self.last\_feedback\_time = time.time()

&nbsp;       self.last\_federated\_time = time.time()

&nbsp;

&nbsp;   def start(self):

&nbsp;       """Start the feedback loop"""

&nbsp;       self.active = True

&nbsp;

&nbsp;       # Start feedback collection

&nbsp;       self.feedback\_interface.start\_feedback\_collection()

&nbsp;

&nbsp;       # Start feedback processing thread

&nbsp;       self.feedback\_thread = threading.Thread(target=self.\_process\_feedback\_loop)

&nbsp;       self.feedback\_thread.daemon = True

&nbsp;       self.feedback\_thread.start()

&nbsp;

&nbsp;       # Start federated learning thread

&nbsp;       self.federated\_thread = threading.Thread(target=self.\_federated\_learning\_loop)

&nbsp;       self.federated\_thread.daemon = True

&nbsp;       self.federated\_thread.start()

&nbsp;

&nbsp;       print("Feedback loop started.")

&nbsp;

&nbsp;   def stop(self):

&nbsp;       """Stop the feedback loop"""

&nbsp;       self.active = False

&nbsp;

&nbsp;       # Stop feedback collection

&nbsp;       self.feedback\_interface.stop\_feedback\_collection()

&nbsp;

&nbsp;       # Wait for threads to finish

&nbsp;       if self.feedback\_thread:

&nbsp;           self.feedback\_thread.join()

&nbsp;

&nbsp;       if self.federated\_thread:

&nbsp;           self.federated\_thread.join()

&nbsp;

&nbsp;       print("Feedback loop stopped.")

&nbsp;

&nbsp;   def \_process\_feedback\_loop(self):

&nbsp;       """Process feedback in a loop"""

&nbsp;       while self.active:

&nbsp;           current\_time = time.time()

&nbsp;

&nbsp;           # Check if it's time to process feedback

&nbsp;           if current\_time - self.last\_feedback\_time >= self.feedback\_interval:

&nbsp;               # Get feedback

&nbsp;               feedbacks = self.feedback\_interface.get\_feedback()

&nbsp;

&nbsp;               if feedbacks:

&nbsp;                   print(f"Processing {len(feedbacks)} feedbacks...")

&nbsp;

&nbsp;                   # Update model based on feedback

&nbsp;                   self.model\_updater.update\_from\_feedback(feedbacks)

&nbsp;

&nbsp;                   # Update dataset processor if available

&nbsp;                   if self.dataset\_processor:

&nbsp;                       self.\_update\_dataset\_processor(feedbacks)

&nbsp;

&nbsp;                   # Update model components if available

&nbsp;                   if self.model\_components:

&nbsp;                       self.\_update\_model\_components(feedbacks)

&nbsp;

&nbsp;               # Update last feedback time

&nbsp;               self.last\_feedback\_time = current\_time

&nbsp;

&nbsp;           # Sleep for a while

&nbsp;           time.sleep(5)

&nbsp;

&nbsp;   def \_federated\_learning\_loop(self):

&nbsp;       """Run federated learning in a loop"""

&nbsp;       while self.active:

&nbsp;           current\_time = time.time()

&nbsp;

&nbsp;           # Check if it's time to run federated learning

&nbsp;           if current\_time - self.last\_federated\_time >= self.federated\_interval:

&nbsp;               print("Running federated learning round...")

&nbsp;

&nbsp;               # Run federated learning

&nbsp;               round\_metrics = self.federated\_server.train()

&nbsp;

&nbsp;               # Update model based on federated learning

&nbsp;               self.model\_updater.update\_from\_federated\_learning(round\_metrics)

&nbsp;

&nbsp;               # Update last federated time

&nbsp;               self.last\_federated\_time = current\_time

&nbsp;

&nbsp;           # Sleep for a while

&nbsp;           time.sleep(5)

&nbsp;

&nbsp;   def \_update\_dataset\_processor(self, feedbacks: List\[Dict]):

&nbsp;       """Update dataset processor based on feedback"""

&nbsp;       # In a real implementation, this would update the dataset processing pipeline

&nbsp;       # based on the feedback received

&nbsp;

&nbsp;       print("Updating dataset processor based on feedback...")

&nbsp;

&nbsp;       # For simulation, we'll just print a message

&nbsp;       for feedback in feedbacks:

&nbsp;           print(f"  - Feedback: {feedback\['prediction']} -> {feedback\['correction']}")

&nbsp;

&nbsp;       # In practice, you might:

&nbsp;       # 1. Add new data to the dataset

&nbsp;       # 2. Update data preprocessing steps

&nbsp;       # 3. Adjust data augmentation strategies

&nbsp;       # 4. Update data balancing techniques

&nbsp;

&nbsp;   def \_update\_model\_components(self, feedbacks: List\[Dict]):

&nbsp;       """Update model components based on feedback"""

&nbsp;       # In a real implementation, this would update various model components

&nbsp;       # based on the feedback received

&nbsp;

&nbsp;       print("Updating model components based on feedback...")

&nbsp;

&nbsp;       # For simulation, we'll just print a message

&nbsp;       for feedback in feedbacks:

&nbsp;           print(f"  - Feedback: {feedback\['prediction']} -> {feedback\['correction']}")

&nbsp;

&nbsp;       # In practice, you might:

&nbsp;       # 1. Update feature extraction methods

&nbsp;       # 2. Adjust model architecture

&nbsp;       # 3. Update knowledge graph integration

&nbsp;       # 4. Modify uncertainty estimation techniques

&nbsp;       # 5. Update explainability methods

&nbsp;

&nbsp;   def get\_status(self) -> Dict:

&nbsp;       """Get the current status of the feedback loop"""

&nbsp;       current\_time = time.time()

&nbsp;

&nbsp;       return {

&nbsp;           "active": self.active,

&nbsp;           "last\_feedback\_time": self.last\_feedback\_time,

&nbsp;           "last\_federated\_time": self.last\_federated\_time,

&nbsp;           "time\_since\_feedback": current\_time - self.last\_feedback\_time,

&nbsp;           "time\_since\_federated": current\_time - self.last\_federated\_time,

&nbsp;           "feedback\_statistics": self.feedback\_interface.get\_feedback\_statistics(),

&nbsp;           "update\_history": self.model\_updater.get\_update\_history()

&nbsp;       }



\# Example usage

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   # Create a simple model for demonstration

&nbsp;   class SimpleModel(nn.Module):

&nbsp;       def \_\_init\_\_(self, input\_dim=512, hidden\_dim=256, output\_dim=2):

&nbsp;           super(SimpleModel, self).\_\_init\_\_()

&nbsp;           self.fc1 = nn.Linear(input\_dim, hidden\_dim)

&nbsp;           self.fc2 = nn.Linear(hidden\_dim, output\_dim)

&nbsp;

&nbsp;       def forward(self, x):

&nbsp;           x = F.relu(self.fc1(x))

&nbsp;           x = self.fc2(x)

&nbsp;           return x

&nbsp;

&nbsp;   # Create model

&nbsp;   model = SimpleModel().to(device)

&nbsp;

&nbsp;   # Create knowledge graph

&nbsp;   kg = Graph()

&nbsp;

&nbsp;   # Create feedback interface

&nbsp;   feedback\_interface = ClinicianFeedbackInterface()

&nbsp;

&nbsp;   # Create federated clients

&nbsp;   # In a real implementation, each client would have its own data

&nbsp;   # For simulation, we'll create dummy data loaders

&nbsp;   class DummyDataset(Dataset):

&nbsp;       def \_\_init\_\_(self, size=100):

&nbsp;           self.size = size

&nbsp;

&nbsp;       def \_\_len\_\_(self):

&nbsp;           return self.size

&nbsp;

&nbsp;       def \_\_getitem\_\_(self, idx):

&nbsp;           # Generate random data

&nbsp;           data = torch.randn(512)

&nbsp;           target = torch.randint(0, 2, (1,)).item()

&nbsp;           return data, target

&nbsp;

&nbsp;   # Create data loaders for clients

&nbsp;   client1\_data = DummyDataset(100)

&nbsp;   client2\_data = DummyDataset(100)

&nbsp;   client3\_data = DummyDataset(100)

&nbsp;

&nbsp;   client1\_loader = DataLoader(client1\_data, batch\_size=10, shuffle=True)

&nbsp;   client2\_loader = DataLoader(client2\_data, batch\_size=10, shuffle=True)

&nbsp;   client3\_loader = DataLoader(client3\_data, batch\_size=10, shuffle=True)

&nbsp;

&nbsp;   # Create federated clients

&nbsp;   client1 = FederatedClient("client1", model, client1\_loader)

&nbsp;   client2 = FederatedClient("client2", model, client2\_loader)

&nbsp;   client3 = FederatedClient("client3", model, client3\_loader)

&nbsp;

&nbsp;   # Create federated server

&nbsp;   federated\_server = FederatedServer(model, num\_rounds=3)

&nbsp;   federated\_server.add\_client(client1)

&nbsp;   federated\_server.add\_client(client2)

&nbsp;   federated\_server.add\_client(client3)

&nbsp;

&nbsp;   # Create model updater

&nbsp;   model\_updater = ModelUpdater(model, kg)

&nbsp;

&nbsp;   # Create feedback loop

&nbsp;   feedback\_loop = FeedbackLoop(

&nbsp;       feedback\_interface=feedback\_interface,

&nbsp;       federated\_server=federated\_server,

&nbsp;       model\_updater=model\_updater

&nbsp;   )

&nbsp;

&nbsp;   # Start feedback loop

&nbsp;   feedback\_loop.start()

&nbsp;

&nbsp;   # Simulate adding feedback

&nbsp;   for i in range(5):

&nbsp;       feedback = {

&nbsp;           "prediction": random.choice(\["malignant", "benign"]),

&nbsp;           "correction": random.choice(\["malignant", "benign"]),

&nbsp;           "confidence": random.uniform(0.5, 1.0),

&nbsp;           "feedback\_type": random.choice(\["correction", "confirmation", "additional\_info"])

&nbsp;       }

&nbsp;       feedback\_interface.add\_feedback(feedback)

&nbsp;       time.sleep(2)

&nbsp;

&nbsp;   # Let it run for a while

&nbsp;   time.sleep(10)

&nbsp;

&nbsp;   # Get status

&nbsp;   status = feedback\_loop.get\_status()

&nbsp;   print("\\nFeedback Loop Status:")

&nbsp;   print(f"Active: {status\['active']}")

&nbsp;   print(f"Feedback Statistics: {status\['feedback\_statistics']}")

&nbsp;   print(f"Update History: {len(status\['update\_history'])} updates")

&nbsp;

&nbsp;   # Stop feedback loop

&nbsp;   feedback\_loop.stop()

&nbsp;

&nbsp;   # Plot training progress

&nbsp;   model\_updater.plot\_training\_progress()

&nbsp;

&nbsp;   print("\\nInteractive Learning Component completed!")

\# Set style

plt.style.use('seaborn-v0\_8-whitegrid')

sns.set\_palette("viridis")



class KnowledgeGraphVisualizer:

&nbsp;   def \_\_init\_\_(self):

&nbsp;       # Define update cycles

&nbsp;       self.updates = \[

&nbsp;           {"date": datetime(2021, 1, 1), "description": "Initial Knowledge Graph"},

&nbsp;           {"date": datetime(2021, 6, 15), "description": "Update 1: New Biomarkers"},

&nbsp;           {"date": datetime(2022, 3, 10), "description": "Update 2: Treatment Relationships"},

&nbsp;           {"date": datetime(2022, 9, 5), "description": "Update 3: Clinical Guidelines"}

&nbsp;       ]

&nbsp;

&nbsp;       # Define entities added in each update

&nbsp;       self.entities = {

&nbsp;           "Initial": \["Breast\_Cancer", "Biopsy", "DCIS", "Invasive\_Carcinoma", "Mammogram"],

&nbsp;           "Update1": \["HER2", "HER2\_low", "Breast\_Lump", "Nipple\_Discharge"],

&nbsp;           "Update2": \["Endocrine\_Therapy", "Trastuzumab", "PIK3CA", "Alpelisib"],

&nbsp;           "Update3": \["Trastuzumab\_deruxtecan", "Skin\_Changes", "Benign\_Tumor"]

&nbsp;       }

&nbsp;

&nbsp;       # Define relationships added in each update

&nbsp;       self.relationships = {

&nbsp;           "Initial": \[

&nbsp;               ("Breast\_Cancer", "Biopsy", "REQUIRES"),

&nbsp;               ("DCIS", "Breast\_Cancer", "SUBTYPE\_OF"),

&nbsp;               ("Invasive\_Carcinoma", "Breast\_Cancer", "SUBTYPE\_OF"),

&nbsp;               ("Mammogram", "Breast\_Cancer", "DETECTS")

&nbsp;           ],

&nbsp;           "Update1": \[

&nbsp;               ("HER2", "Breast\_Cancer", "BIOMARKER\_OF"),

&nbsp;               ("HER2\_low", "Breast\_Cancer", "BIOMARKER\_OF"),

&nbsp;               ("Breast\_Lump", "Breast\_Cancer", "SYMPTOM\_OF"),

&nbsp;               ("Nipple\_Discharge", "Breast\_Cancer", "SYMPTOM\_OF")

&nbsp;           ],

&nbsp;           "Update2": \[

&nbsp;               ("Endocrine\_Therapy", "Breast\_Cancer", "TREATMENT\_FOR"),

&nbsp;               ("Trastuzumab", "HER2", "TARGETS"),

&nbsp;               ("PIK3CA", "Breast\_Cancer", "MUTATION\_IN"),

&nbsp;               ("Alpelisib", "PIK3CA", "TARGETS")

&nbsp;           ],

&nbsp;           "Update3": \[

&nbsp;               ("Trastuzumab\_deruxtecan", "HER2\_low", "TARGETS"),

&nbsp;               ("Skin\_Changes", "Breast\_Cancer", "SYMPTOM\_OF"),

&nbsp;               ("Benign\_Tumor", "Breast\_Lump", "DIFFERENTIAL\_DIAGNOSIS")

&nbsp;           ]

&nbsp;       }

&nbsp;

&nbsp;       # Define contradictions

&nbsp;       self.contradictions = \[

&nbsp;           {

&nbsp;               "id": "C001",

&nbsp;               "entities": ("HER2", "Trastuzumab"),

&nbsp;               "detected": datetime(2021, 6, 15),

&nbsp;               "resolved": datetime(2021, 7, 1),

&nbsp;               "description": "Efficacy controversy"

&nbsp;           },

&nbsp;           {

&nbsp;               "id": "C002",

&nbsp;               "entities": ("ESR1", "Endocrine\_Therapy"),

&nbsp;               "detected": datetime(2022, 3, 10),

&nbsp;               "resolved": datetime(2022, 3, 25),

&nbsp;               "description": "Resistance evidence"

&nbsp;           },

&nbsp;           {

&nbsp;               "id": "C003",

&nbsp;               "entities": ("PIK3CA", "Alpelisib"),

&nbsp;               "detected": datetime(2022, 9, 5),

&nbsp;               "resolved": datetime(2022, 9, 20),

&nbsp;               "description": "Guideline update"

&nbsp;           }

&nbsp;       ]

&nbsp;

&nbsp;       # Define provenance data

&nbsp;       self.sources = \[

&nbsp;           {"id": "S001", "name": "Clinical Guidelines", "type": "Guideline", "reliability": 0.95},

&nbsp;           {"id": "S002", "name": "PubMed Research", "type": "Research", "reliability": 0.85},

&nbsp;           {"id": "S003", "name": "Clinical Trials", "type": "Trial", "reliability": 0.90},

&nbsp;           {"id": "S004", "name": "Expert Consensus", "type": "Expert", "reliability": 0.80},

&nbsp;           {"id": "S005", "name": "FDA Approvals", "type": "Regulatory", "reliability": 0.98}

&nbsp;       ]

&nbsp;

&nbsp;       # Define provenance links

&nbsp;       self.provenance\_links = \[

&nbsp;           # Initial entities

&nbsp;           {"source": "S001", "element": "Breast\_Cancer", "type": "entity", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "Biopsy", "type": "entity", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "DCIS", "type": "entity", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "Invasive\_Carcinoma", "type": "entity", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "Mammogram", "type": "entity", "confidence": 0.95},

&nbsp;

&nbsp;           # Update1 entities

&nbsp;           {"source": "S002", "element": "HER2", "type": "entity", "confidence": 0.85},

&nbsp;           {"source": "S002", "element": "HER2\_low", "type": "entity", "confidence": 0.85},

&nbsp;           {"source": "S003", "element": "Breast\_Lump", "type": "entity", "confidence": 0.90},

&nbsp;           {"source": "S003", "element": "Nipple\_Discharge", "type": "entity", "confidence": 0.90},

&nbsp;

&nbsp;           # Update2 entities

&nbsp;           {"source": "S005", "element": "Endocrine\_Therapy", "type": "entity", "confidence": 0.98},

&nbsp;           {"source": "S005", "element": "Trastuzumab", "type": "entity", "confidence": 0.98},

&nbsp;           {"source": "S002", "element": "PIK3CA", "type": "entity", "confidence": 0.85},

&nbsp;           {"source": "S005", "element": "Alpelisib", "type": "entity", "confidence": 0.98},

&nbsp;

&nbsp;           # Update3 entities

&nbsp;           {"source": "S005", "element": "Trastuzumab\_deruxtecan", "type": "entity", "confidence": 0.98},

&nbsp;           {"source": "S003", "element": "Skin\_Changes", "type": "entity", "confidence": 0.90},

&nbsp;           {"source": "S004", "element": "Benign\_Tumor", "type": "entity", "confidence": 0.80},

&nbsp;

&nbsp;           # Relationships

&nbsp;           {"source": "S001", "element": "Breast\_Cancer-Biopsy-REQUIRES", "type": "relationship", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "DCIS-Breast\_Cancer-SUBTYPE\_OF", "type": "relationship", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "Invasive\_Carcinoma-Breast\_Cancer-SUBTYPE\_OF", "type": "relationship", "confidence": 0.95},

&nbsp;           {"source": "S001", "element": "Mammogram-Breast\_Cancer-DETECTS", "type": "relationship", "confidence": 0.95},

&nbsp;           {"source": "S002", "element": "HER2-Breast\_Cancer-BIOMARKER\_OF", "type": "relationship", "confidence": 0.85},

&nbsp;           {"source": "S002", "element": "HER2\_low-Breast\_Cancer-BIOMARKER\_OF", "type": "relationship", "confidence": 0.85},

&nbsp;           {"source": "S003", "element": "Breast\_Lump-Breast\_Cancer-SYMPTOM\_OF", "type": "relationship", "confidence": 0.90},

&nbsp;           {"source": "S003", "element": "Nipple\_Discharge-Breast\_Cancer-SYMPTOM\_OF", "type": "relationship", "confidence": 0.90},

&nbsp;           {"source": "S005", "element": "Endocrine\_Therapy-Breast\_Cancer-TREATMENT\_FOR", "type": "relationship", "confidence": 0.98},

&nbsp;           {"source": "S005", "element": "Trastuzumab-HER2-TARGETS", "type": "relationship", "confidence": 0.98},

&nbsp;           {"source": "S002", "element": "PIK3CA-Breast\_Cancer-MUTATION\_IN", "type": "relationship", "confidence": 0.85},

&nbsp;           {"source": "S005", "element": "Alpelisib-PIK3CA-TARGETS", "type": "relationship", "confidence": 0.98},

&nbsp;           {"source": "S005", "element": "Trastuzumab\_deruxtecan-HER2\_low-TARGETS", "type": "relationship", "confidence": 0.98},

&nbsp;           {"source": "S003", "element": "Skin\_Changes-Breast\_Cancer-SYMPTOM\_OF", "type": "relationship", "confidence": 0.90},

&nbsp;           {"source": "S004", "element": "Benign\_Tumor-Breast\_Lump-DIFFERENTIAL\_DIAGNOSIS", "type": "relationship", "confidence": 0.80}

&nbsp;       ]

&nbsp;

&nbsp;   def create\_temporal\_evolution\_visualization(self):

&nbsp;       """Create comprehensive temporal evolution visualization"""

&nbsp;       fig = plt.figure(figsize=(20, 12))

&nbsp;       gs = fig.add\_gridspec(3, 3, height\_ratios=\[1, 1, 1], hspace=0.3, wspace=0.3)

&nbsp;

&nbsp;       # 1. Timeline of updates

&nbsp;       ax1 = fig.add\_subplot(gs\[0, :])

&nbsp;

&nbsp;       # Create timeline data

&nbsp;       events = \[]

&nbsp;       for i, update in enumerate(self.updates):

&nbsp;           events.append({

&nbsp;               "date": update\["date"],

&nbsp;               "event": update\["description"],

&nbsp;               "type": "update",

&nbsp;               "entities": len(self.entities\[list(self.entities.keys())\[i]]),

&nbsp;               "relationships": len(self.relationships\[list(self.relationships.keys())\[i]])

&nbsp;           })

&nbsp;

&nbsp;       # Add contradiction events

&nbsp;       for contradiction in self.contradictions:

&nbsp;           events.append({

&nbsp;               "date": contradiction\["detected"],

&nbsp;               "event": f"Contradiction: {contradiction\['id']}",

&nbsp;               "type": "contradiction\_detected",

&nbsp;               "entities": 0,

&nbsp;               "relationships": 0

&nbsp;           })

&nbsp;           events.append({

&nbsp;               "date": contradiction\["resolved"],

&nbsp;               "event": f"Resolved: {contradiction\['id']}",

&nbsp;               "type": "contradiction\_resolved",

&nbsp;               "entities": 0,

&nbsp;               "relationships": 0

&nbsp;           })

&nbsp;

&nbsp;       # Convert to DataFrame

&nbsp;       df = pd.DataFrame(events)

&nbsp;       df = df.sort\_values("date")

&nbsp;

&nbsp;       # Create timeline

&nbsp;       colors = plt.cm.Set3(np.linspace(0, 1, len(df\["type"].unique())))

&nbsp;       type\_color\_map = {t: colors\[i] for i, t in enumerate(df\["type"].unique())}

&nbsp;

&nbsp;       for i, event in df.iterrows():

&nbsp;           ax1.scatter(event\["date"], i, color=type\_color\_map\[event\["type"]], s=100, alpha=0.7)

&nbsp;           ax1.text(event\["date"] + timedelta(days=5), i,

&nbsp;                   f"{event\['event']}\\n(+{event\['entities']} entities, +{event\['relationships']} rels)",

&nbsp;                   verticalalignment='center', fontsize=9)

&nbsp;

&nbsp;       # Format x-axis

&nbsp;       ax1.xaxis.set\_major\_formatter(mdates.DateFormatter('%Y-%m'))

&nbsp;       plt.setp(ax1.get\_xticklabels(), rotation=45)

&nbsp;       ax1.set\_title("Knowledge Graph Evolution Timeline", fontsize=14, fontweight='bold')

&nbsp;       ax1.set\_yticks(\[])

&nbsp;       ax1.grid(True, alpha=0.3)

&nbsp;

&nbsp;       # Create legend

&nbsp;       legend\_elements = \[plt.Line2D(\[0], \[0], marker='o', color='w',

&nbsp;                         markerfacecolor=type\_color\_map\[t], markersize=10, label=t) for t in df\["type"].unique()]

&nbsp;       ax1.legend(handles=legend\_elements, loc='upper left')

&nbsp;

&nbsp;       # 2. Growth metrics

&nbsp;       ax2 = fig.add\_subplot(gs\[1, 0])

&nbsp;

&nbsp;       # Calculate cumulative growth

&nbsp;       cumulative\_entities = \[]

&nbsp;       cumulative\_relationships = \[]

&nbsp;       dates = \[]

&nbsp;

&nbsp;       for update in self.updates:

&nbsp;           dates.append(update\["date"])

&nbsp;           if update\["description"] == "Initial Knowledge Graph":

&nbsp;               cumulative\_entities.append(len(self.entities\["Initial"]))

&nbsp;               cumulative\_relationships.append(len(self.relationships\["Initial"]))

&nbsp;           elif update\["description"] == "Update 1: New Biomarkers":

&nbsp;               cumulative\_entities.append(len(self.entities\["Initial"]) + len(self.entities\["Update1"]))

&nbsp;               cumulative\_relationships.append(len(self.relationships\["Initial"]) + len(self.relationships\["Update1"]))

&nbsp;           elif update\["description"] == "Update 2: Treatment Relationships":

&nbsp;               cumulative\_entities.append(len(self.entities\["Initial"]) + len(self.entities\["Update1"]) + len(self.entities\["Update2"]))

&nbsp;               cumulative\_relationships.append(len(self.relationships\["Initial"]) + len(self.relationships\["Update1"]) + len(self.relationships\["Update2"]))

&nbsp;           elif update\["description"] == "Update 3: Clinical Guidelines":

&nbsp;               cumulative\_entities.append(len(self.entities\["Initial"]) + len(self.entities\["Update1"]) + len(self.entities\["Update2"]) + len(self.entities\["Update3"]))

&nbsp;               cumulative\_relationships.append(len(self.relationships\["Initial"]) + len(self.relationships\["Update1"]) + len(self.relationships\["Update2"]) + len(self.relationships\["Update3"]))

&nbsp;

&nbsp;       # Plot growth metrics

&nbsp;       ax2.plot(dates, cumulative\_entities, 'o-', label='Entities', color='#4e79a7')

&nbsp;       ax2.plot(dates, cumulative\_relationships, 'o-', label='Relationships', color='#f28e2c')

&nbsp;       ax2.set\_title("Knowledge Graph Growth", fontsize=12, fontweight='bold')

&nbsp;       ax2.set\_xlabel("Date")

&nbsp;       ax2.set\_ylabel("Count")

&nbsp;       ax2.legend()

&nbsp;       ax2.grid(True, alpha=0.3)

&nbsp;       ax2.xaxis.set\_major\_formatter(mdates.DateFormatter('%Y-%m'))

&nbsp;       plt.setp(ax2.get\_xticklabels(), rotation=45)

&nbsp;

&nbsp;       # 3. Entity type distribution

&nbsp;       ax3 = fig.add\_subplot(gs\[1, 1])

&nbsp;

&nbsp;       # Count entity types by update

&nbsp;       entity\_types = {

&nbsp;           "Initial": {"Disease": 1, "Procedure": 1, "Subtype": 2, "Test": 1},

&nbsp;           "Update1": {"Biomarker": 2, "Symptom": 2},

&nbsp;           "Update2": {"Treatment": 2, "Gene": 2},

&nbsp;           "Update3": {"Treatment": 1, "Symptom": 1, "Condition": 1}

&nbsp;       }

&nbsp;

&nbsp;       # Create a stacked bar chart

&nbsp;       entity\_df = pd.DataFrame(entity\_types).T

&nbsp;       entity\_df.plot(kind='bar', stacked=True, ax=ax3, colormap='viridis')

&nbsp;       ax3.set\_title("Entity Types by Update", fontsize=12, fontweight='bold')

&nbsp;       ax3.set\_xlabel("Update")

&nbsp;       ax3.set\_ylabel("Count")

&nbsp;       ax3.legend(title="Entity Type")

&nbsp;       ax3.grid(True, alpha=0.3)

&nbsp;

&nbsp;       # 4. Relationship type distribution

&nbsp;       ax4 = fig.add\_subplot(gs\[1, 2])

&nbsp;

&nbsp;       # Count relationship types by update

&nbsp;       rel\_types = {

&nbsp;           "Initial": {"REQUIRES": 1, "SUBTYPE\_OF": 2, "DETECTS": 1},

&nbsp;           "Update1": {"BIOMARKER\_OF": 2, "SYMPTOM\_OF": 2},

&nbsp;           "Update2": {"TREATMENT\_FOR": 1, "TARGETS": 2, "MUTATION\_IN": 1},

&nbsp;           "Update3": {"TARGETS": 1, "SYMPTOM\_OF": 1, "DIFFERENTIAL\_DIAGNOSIS": 1}

&nbsp;       }

&nbsp;

&nbsp;       # Create a stacked bar chart

&nbsp;       rel\_df = pd.DataFrame(rel\_types).T

&nbsp;       rel\_df.plot(kind='bar', stacked=True, ax=ax4, colormap='plasma')

&nbsp;       ax4.set\_title("Relationship Types by Update", fontsize=12, fontweight='bold')

&nbsp;       ax4.set\_xlabel("Update")

&nbsp;       ax4.set\_ylabel("Count")

&nbsp;       ax4.legend(title="Relationship Type")

&nbsp;       ax4.grid(True, alpha=0.3)

&nbsp;

&nbsp;       # 5. Contradiction timeline

&nbsp;       ax5 = fig.add\_subplot(gs\[2, 0])

&nbsp;

&nbsp;       # Create a timeline for contradictions

&nbsp;       for i, contradiction in enumerate(self.contradictions):

&nbsp;           # Draw line from detection to resolution

&nbsp;           ax5.plot(\[contradiction\["detected"], contradiction\["resolved"]], \[i, i],

&nbsp;                    color='red', linewidth=2, alpha=0.7)

&nbsp;           # Mark detection point

&nbsp;           ax5.scatter(contradiction\["detected"], i, color='red', s=100, alpha=0.7, zorder=3)

&nbsp;           # Mark resolution point

&nbsp;           ax5.scatter(contradiction\["resolved"], i, color='green', s=100, alpha=0.7, zorder=3)

&nbsp;           # Add text

&nbsp;           ax5.text(contradiction\["detected"] - timedelta(days=15), i,

&nbsp;                    f"{contradiction\['id']}\\n{contradiction\['description']}",

&nbsp;                    verticalalignment='center', horizontalalignment='right', fontsize=9)

&nbsp;

&nbsp;       ax5.set\_title("Contradiction Resolution Timeline", fontsize=12, fontweight='bold')

&nbsp;       ax5.set\_xlabel("Date")

&nbsp;       ax5.set\_yticks(range(len(self.contradictions)))

&nbsp;       ax5.set\_yticklabels(\[c\["id"] for c in self.contradictions])

&nbsp;       ax5.grid(True, alpha=0.3)

&nbsp;       ax5.xaxis.set\_major\_formatter(mdates.DateFormatter('%Y-%m'))

&nbsp;       plt.setp(ax5.get\_xticklabels(), rotation=45)

&nbsp;

&nbsp;       # 6. Knowledge graph snapshots

&nbsp;       # Initial graph

&nbsp;       ax6 = fig.add\_subplot(gs\[2, 1])

&nbsp;       G\_initial = nx.Graph()

&nbsp;

&nbsp;       # Add initial entities

&nbsp;       for entity in self.entities\["Initial"]:

&nbsp;           G\_initial.add\_node(entity)

&nbsp;

&nbsp;       # Add initial relationships

&nbsp;       for rel in self.relationships\["Initial"]:

&nbsp;           G\_initial.add\_edge(rel\[0], rel\[1], label=rel\[2])

&nbsp;

&nbsp;       # Draw the graph

&nbsp;       pos = nx.spring\_layout(G\_initial, seed=42)

&nbsp;       nx.draw(G\_initial, pos, ax=ax6, with\_labels=True, node\_size=500, node\_color='skyblue',

&nbsp;               font\_size=8, font\_weight='bold')

&nbsp;       ax6.set\_title("Initial Knowledge Graph", fontsize=12, fontweight='bold')

&nbsp;

&nbsp;       # Final graph

&nbsp;       ax7 = fig.add\_subplot(gs\[2, 2])

&nbsp;       G\_final = nx.Graph()

&nbsp;

&nbsp;       # Add all entities

&nbsp;       for entities\_list in self.entities.values():

&nbsp;           for entity in entities\_list:

&nbsp;               G\_final.add\_node(entity)

&nbsp;

&nbsp;       # Add all relationships

&nbsp;       for rels\_list in self.relationships.values():

&nbsp;           for rel in rels\_list:

&nbsp;               G\_final.add\_edge(rel\[0], rel\[1], label=rel\[2])

&nbsp;

&nbsp;       # Draw the graph

&nbsp;       pos = nx.spring\_layout(G\_final, seed=42)

&nbsp;       nx.draw(G\_final, pos, ax=ax7, with\_labels=True, node\_size=300, node\_color='lightgreen',

&nbsp;               font\_size=7, font\_weight='bold')

&nbsp;       ax7.set\_title("Final Knowledge Graph", fontsize=12, fontweight='bold')

&nbsp;

&nbsp;       # Add a main title to the figure

&nbsp;       plt.suptitle("Temporal Evolution of Breast Cancer Knowledge Graph", fontsize=16, fontweight='bold')

&nbsp;       plt.tight\_layout(rect=\[0, 0, 1, 0.96])  # Adjust for the suptitle

&nbsp;

&nbsp;       return fig

&nbsp;

&nbsp;   def create\_provenance\_visualization(self):

&nbsp;       """Create a clear and comprehensive provenance visualization"""

&nbsp;       fig = plt.figure(figsize=(20, 16))

&nbsp;       gs = fig.add\_gridspec(3, 2, height\_ratios=\[1, 1, 1], hspace=0.3, wspace=0.3)

&nbsp;

&nbsp;       # 1. Source contribution overview

&nbsp;       ax1 = fig.add\_subplot(gs\[0, 0])

&nbsp;

&nbsp;       # Count contributions by source

&nbsp;       source\_contributions = {}

&nbsp;       for source in self.sources:

&nbsp;           source\_id = source\["id"]

&nbsp;           source\_contributions\[source\_id] = {

&nbsp;               "name": source\["name"],

&nbsp;               "entities": 0,

&nbsp;               "relationships": 0,

&nbsp;               "avg\_confidence": 0

&nbsp;           }

&nbsp;

&nbsp;       # Calculate contributions

&nbsp;       for link in self.provenance\_links:

&nbsp;           source\_id = link\["source"]

&nbsp;           if link\["type"] == "entity":

&nbsp;               source\_contributions\[source\_id]\["entities"] += 1

&nbsp;           else:

&nbsp;               source\_contributions\[source\_id]\["relationships"] += 1

&nbsp;

&nbsp;       # Calculate average confidence

&nbsp;       for source\_id in source\_contributions:

&nbsp;           confidences = \[link\["confidence"] for link in self.provenance\_links if link\["source"] == source\_id]

&nbsp;           source\_contributions\[source\_id]\["avg\_confidence"] = np.mean(confidences)

&nbsp;

&nbsp;       # Create DataFrame

&nbsp;       df\_contributions = pd.DataFrame(\[

&nbsp;           {

&nbsp;               "Source": sc\["name"],

&nbsp;               "Entities": sc\["entities"],

&nbsp;               "Relationships": sc\["relationships"],

&nbsp;               "Total": sc\["entities"] + sc\["relationships"],

&nbsp;               "Avg Confidence": sc\["avg\_confidence"]

&nbsp;           }

&nbsp;           for sc in source\_contributions.values()

&nbsp;       ])

&nbsp;

&nbsp;       # Create bar chart

&nbsp;       df\_contributions.plot(x="Source", y=\["Entities", "Relationships"], kind="bar", ax=ax1,

&nbsp;                            color=\["#1f77b4", "#2ca02c"])

&nbsp;       ax1.set\_title("Source Contributions", fontsize=14, fontweight='bold')

&nbsp;       ax1.set\_ylabel("Count")

&nbsp;       ax1.set\_xlabel("")

&nbsp;       ax1.grid(True, alpha=0.3)

&nbsp;       ax1.legend(title="Element Type")

&nbsp;

&nbsp;       # Add total count labels

&nbsp;       for i, total in enumerate(df\_contributions\["Total"]):

&nbsp;           ax1.text(i, total + 0.2, str(total), ha='center', fontweight='bold')

&nbsp;

&nbsp;       # 2. Source reliability vs contribution

&nbsp;       ax2 = fig.add\_subplot(gs\[0, 1])

&nbsp;

&nbsp;       # Get source reliability

&nbsp;       source\_reliability = {source\["id"]: source\["reliability"] for source in self.sources}

&nbsp;

&nbsp;       # Create scatter plot

&nbsp;       for i, row in df\_contributions.iterrows():

&nbsp;           source\_name = row\["Source"]

&nbsp;           source\_id = \[s\["id"] for s in self.sources if s\["name"] == source\_name]\[0]

&nbsp;           reliability = source\_reliability\[source\_id]

&nbsp;           total = row\["Total"]

&nbsp;           avg\_conf = row\["Avg Confidence"]

&nbsp;

&nbsp;           # Size based on total contribution

&nbsp;           size = total \* 30 + 50

&nbsp;

&nbsp;           # Color based on average confidence

&nbsp;           color = plt.cm.RdYlGn(avg\_conf)

&nbsp;

&nbsp;           ax2.scatter(reliability, avg\_conf, s=size, color=color, alpha=0.7,

&nbsp;                      edgecolors='black', linewidth=1)

&nbsp;           ax2.text(reliability, avg\_conf, source\_name, ha='center', va='center',

&nbsp;                   fontsize=9, fontweight='bold')

&nbsp;

&nbsp;       ax2.set\_title("Source Reliability vs. Confidence", fontsize=14, fontweight='bold')

&nbsp;       ax2.set\_xlabel("Source Reliability")

&nbsp;       ax2.set\_ylabel("Average Confidence of Contributions")

&nbsp;       ax2.grid(True, alpha=0.3)

&nbsp;       ax2.set\_xlim(0.7, 1.0)

&nbsp;       ax2.set\_ylim(0.7, 1.0)

&nbsp;

&nbsp;       # Add colorbar for confidence

&nbsp;       sm = plt.cm.ScalarMappable(cmap=plt.cm.RdYlGn, norm=plt.Normalize(vmin=0.7, vmax=1.0))

&nbsp;       sm.set\_array(\[])

&nbsp;       cbar = plt.colorbar(sm, ax=ax2)

&nbsp;       cbar.set\_label('Average Confidence')

&nbsp;

&nbsp;       # 3. Provenance network (simplified)

&nbsp;       ax3 = fig.add\_subplot(gs\[1, :])

&nbsp;

&nbsp;       # Create a simplified graph showing only sources and their direct connections

&nbsp;       G = nx.Graph()

&nbsp;

&nbsp;       # Add source nodes

&nbsp;       for source in self.sources:

&nbsp;           G.add\_node(source\["id"],

&nbsp;                     label=source\["name"],

&nbsp;                     type="source",

&nbsp;                     reliability=source\["reliability"])

&nbsp;

&nbsp;       # Add element nodes (only a subset for clarity)

&nbsp;       # We'll group elements by source and type

&nbsp;       element\_groups = {}

&nbsp;       for link in self.provenance\_links:

&nbsp;           source\_id = link\["source"]

&nbsp;           element\_type = link\["type"]

&nbsp;

&nbsp;           if source\_id not in element\_groups:

&nbsp;               element\_groups\[source\_id] = {"entities": 0, "relationships": 0}

&nbsp;

&nbsp;           if element\_type == "entity":

&nbsp;               element\_groups\[source\_id]\["entities"] += 1

&nbsp;           else:

&nbsp;               element\_groups\[source\_id]\["relationships"] += 1

&nbsp;

&nbsp;       # Add group nodes

&nbsp;       for source\_id, groups in element\_groups.items():

&nbsp;           if groups\["entities"] > 0:

&nbsp;               group\_id = f"{source\_id}\_entities"

&nbsp;               G.add\_node(group\_id,

&nbsp;                         label=f"Entities ({groups\['entities']})",

&nbsp;                         type="entity\_group",

&nbsp;                         count=groups\["entities"])

&nbsp;               G.add\_edge(source\_id, group\_id)

&nbsp;

&nbsp;           if groups\["relationships"] > 0:

&nbsp;               group\_id = f"{source\_id}\_relationships"

&nbsp;               G.add\_node(group\_id,

&nbsp;                         label=f"Relationships ({groups\['relationships']})",

&nbsp;                         type="relationship\_group",

&nbsp;                         count=groups\["relationships"])

&nbsp;               G.add\_edge(source\_id, group\_id)

&nbsp;

&nbsp;       # Create a circular layout

&nbsp;       pos = nx.circular\_layout(G)

&nbsp;

&nbsp;       # Draw the graph

&nbsp;       # Draw sources

&nbsp;       sources = \[n for n in G.nodes if G.nodes\[n]\['type'] == 'source']

&nbsp;       nx.draw\_networkx\_nodes(G, pos, nodelist=sources, node\_size=2000,

&nbsp;                             node\_color='#ff7f0e', ax=ax3)

&nbsp;

&nbsp;       # Draw entity groups

&nbsp;       entity\_groups = \[n for n in G.nodes if G.nodes\[n]\['type'] == 'entity\_group']

&nbsp;       nx.draw\_networkx\_nodes(G, pos, nodelist=entity\_groups, node\_size=1000,

&nbsp;                             node\_color='#1f77b4', ax=ax3)

&nbsp;

&nbsp;       # Draw relationship groups

&nbsp;       relationship\_groups = \[n for n in G.nodes if G.nodes\[n]\['type'] == 'relationship\_group']

&nbsp;       nx.draw\_networkx\_nodes(G, pos, nodelist=relationship\_groups, node\_size=1000,

&nbsp;                             node\_color='#2ca02c', ax=ax3, node\_shape='s')

&nbsp;

&nbsp;       # Draw edges

&nbsp;       nx.draw\_networkx\_edges(G, pos, width=2, alpha=0.5, ax=ax3)

&nbsp;

&nbsp;       # Draw labels

&nbsp;       labels = {n: G.nodes\[n]\['label'] for n in G.nodes}

&nbsp;       nx.draw\_networkx\_labels(G, pos, labels=labels, font\_size=10, ax=ax3)

&nbsp;

&nbsp;       # Add a legend

&nbsp;       legend\_elements = \[

&nbsp;           plt.Line2D(\[0], \[0], marker='o', color='w', markerfacecolor='#ff7f0e',

&nbsp;                     markersize=15, label='Source'),

&nbsp;           plt.Line2D(\[0], \[0], marker='o', color='w', markerfacecolor='#1f77b4',

&nbsp;                     markersize=12, label='Entity Group'),

&nbsp;           plt.Line2D(\[0], \[0], marker='s', color='w', markerfacecolor='#2ca02c',

&nbsp;                     markersize=12, label='Relationship Group')

&nbsp;       ]

&nbsp;       ax3.legend(handles=legend\_elements, loc='upper right')

&nbsp;

&nbsp;       ax3.set\_title("Provenance Network Overview", fontsize=14, fontweight='bold')

&nbsp;       ax3.axis('off')

&nbsp;

&nbsp;       # 4. Confidence distribution by source

&nbsp;       ax4 = fig.add\_subplot(gs\[2, 0])

&nbsp;

&nbsp;       # Prepare data for boxplot

&nbsp;       confidence\_data = \[]

&nbsp;       source\_names = \[]

&nbsp;

&nbsp;       for source in self.sources:

&nbsp;           source\_id = source\["id"]

&nbsp;           source\_name = source\["name"]

&nbsp;           confidences = \[link\["confidence"] for link in self.provenance\_links if link\["source"] == source\_id]

&nbsp;

&nbsp;           if confidences:

&nbsp;               confidence\_data.append(confidences)

&nbsp;               source\_names.append(source\_name)

&nbsp;

&nbsp;       # Create boxplot

&nbsp;       bp = ax4.boxplot(confidence\_data, labels=source\_names, patch\_artist=True)

&nbsp;

&nbsp;       # Color the boxes

&nbsp;       colors = \['#ff7f0e', '#1f77b4', '#2ca02c', '#d62728', '#9467bd']

&nbsp;       for patch, color in zip(bp\['boxes'], colors):

&nbsp;           patch.set\_facecolor(color)

&nbsp;           patch.set\_alpha(0.7)

&nbsp;

&nbsp;       ax4.set\_title("Confidence Distribution by Source", fontsize=14, fontweight='bold')

&nbsp;       ax4.set\_ylabel("Confidence Score")

&nbsp;       ax4.set\_ylim(0.7, 1.0)

&nbsp;       ax4.grid(True, alpha=0.3)

&nbsp;       plt.setp(ax4.get\_xticklabels(), rotation=45, ha='right')

&nbsp;

&nbsp;       # 5. Element type distribution by source

&nbsp;       ax5 = fig.add\_subplot(gs\[2, 1])

&nbsp;

&nbsp;       # Prepare data for stacked bar chart

&nbsp;       element\_counts = \[]

&nbsp;       for source in self.sources:

&nbsp;           source\_id = source\["id"]

&nbsp;           entities = sum(1 for link in self.provenance\_links

&nbsp;                         if link\["source"] == source\_id and link\["type"] == "entity")

&nbsp;           relationships = sum(1 for link in self.provenance\_links

&nbsp;                             if link\["source"] == source\_id and link\["type"] == "relationship")

&nbsp;           element\_counts.append(\[entities, relationships])

&nbsp;

&nbsp;       # Create DataFrame

&nbsp;       df\_element\_types = pd.DataFrame(

&nbsp;           element\_counts,

&nbsp;           index=\[source\["name"] for source in self.sources],

&nbsp;           columns=\["Entities", "Relationships"]

&nbsp;       )

&nbsp;

&nbsp;       # Create stacked bar chart

&nbsp;       df\_element\_types.plot(kind="bar", stacked=True, ax=ax5, color=\["#1f77b4", "#2ca02c"])

&nbsp;       ax5.set\_title("Element Type Distribution by Source", fontsize=14, fontweight='bold')

&nbsp;       ax5.set\_ylabel("Count")

&nbsp;       ax5.set\_xlabel("")

&nbsp;       ax5.grid(True, alpha=0.3)

&nbsp;       ax5.legend(title="Element Type")

&nbsp;       plt.setp(ax5.get\_xticklabels(), rotation=45, ha='right')

&nbsp;

&nbsp;       # Add a main title to the figure

&nbsp;       plt.suptitle("Knowledge Graph Provenance Analysis", fontsize=18, fontweight='bold')

&nbsp;       plt.tight\_layout(rect=\[0, 0, 1, 0.97])  # Adjust for the suptitle

&nbsp;

&nbsp;       return fig

&nbsp;

&nbsp;   def display\_visualizations(self):

&nbsp;       """Display both temporal evolution and provenance visualizations"""

&nbsp;       # Display temporal evolution

&nbsp;       print("Displaying Temporal Evolution Visualization...")

&nbsp;       fig\_temporal = self.create\_temporal\_evolution\_visualization()

&nbsp;       plt.show(fig\_temporal)

&nbsp;

&nbsp;       # Display provenance

&nbsp;       print("\\nDisplaying Provenance Visualization...")

&nbsp;       fig\_provenance = self.create\_provenance\_visualization()

&nbsp;       plt.show(fig\_provenance)

&nbsp;

&nbsp;       return fig\_temporal, fig\_provenance



\# Main execution

if \_\_name\_\_ == "\_\_main\_\_":

&nbsp;   visualizer = KnowledgeGraphVisualizer()

&nbsp;   fig\_temporal, fig\_provenance = visualizer.display\_visualizations()

&nbsp;

































































```
